```python
import numpy as np
import matplotlib.pyplot as plt

def metodo_euler(f, x0, t0, tf, h):
    # Crear array de tiempos
    t = np.arange(t0, tf + h, h)
    n = len(t)
    
    # Inicializar array de soluciones
    x = np.zeros(n)
    x[0] = x0
    
    # Aplicar método de Euler
    for i in range(1, n):
        x[i] = x[i-1] + f(x[i-1], t[i-1]) * h
    
    return t, x

# Definir la función específica de tu ejemplo: f(P,t) = 3P
def f(P, t):
    return 3 * P

# Parámetros del problema
P0 = 0.5      # Condición inicial
t0 = 0        # Tiempo inicial
tf = 1.0      # Tiempo final
h = 0.1       # Paso temporal

# Resolver con método de Euler
t, P_euler = metodo_euler(f, P0, t0, tf, h)

# Solución analítica para comparar (P' = 3P, P(0)=0.5)
P_analitica = 0.5 * np.exp(3 * t)

# Crear tabla similar a la del pizarrón
print("Método de Euler - Tabla de resultados:")
print("t\t\tP(t) Euler\t\tP(t) Analítica\t\tError")
print("-" * 70)
for i in range(len(t)):
    error = abs(P_euler[i] - P_analitica[i])
    print(f"{t[i]:.1f}\t\t{P_euler[i]:.6f}\t\t{P_analitica[i]:.6f}\t\t{error:.6f}")

# Gráfico comparativo
plt.figure(figsize=(10, 6))
plt.plot(t, P_euler, 'bo-', label='Método de Euler', linewidth=2, markersize=6)
plt.plot(t, P_analitica, 'r-', label='Solución Analítica', linewidth=2)
plt.xlabel('Tiempo (t)')
plt.ylabel('P(t)')
plt.title('Método de Euler: $\dot{P} = 3P$, $P(0) = 0.5$')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Paso 1
P0 = 0.5
f_P0 = f(P0, 0)
P1 = P0 + f_P0 * h
print(f"Paso 1:")
print(f"t₀ = 0.0, P₀ = {P0}")
print(f"f(P₀) = 3 × {P0} = {f_P0}")
print(f"P₁ = P₀ + f(P₀)×h = {P0} + {f_P0}×{h} = {P1}")

# Paso 2
f_P1 = f(P1, 0.1)
P2 = P1 + f_P1 * h
print(f"\nPaso 2:")
print(f"t₁ = 0.1, P₁ = {P1}")
print(f"f(P₁) = 3 × {P1} = {f_P1}")
print(f"P₂ = P₁ + f(P₁)×h = {P1} + {f_P1}×{h} = {P2}")
```
