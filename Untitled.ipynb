```python
import numpy as np
import matplotlib.pyplot as plt

def metodo_euler(f, x0, t0, tf, h):
    # Crear array de tiempos
    t = np.arange(t0, tf + h, h)
    n = len(t)
    
    # Inicializar array de soluciones
    x = np.zeros(n)
    x[0] = x0
    
    # Aplicar método de Euler
    for i in range(1, n):
        x[i] = x[i-1] + f(x[i-1], t[i-1]) * h
    
    return t, x

# Definir la función específica de tu ejemplo: f(P,t) = 3P
def f(P, t):
    return 3 * P

# Parámetros del problema
P0 = 0.5      # Condición inicial
t0 = 0        # Tiempo inicial
tf = 1.0      # Tiempo final
h = 0.1       # Paso temporal

# Resolver con método de Euler
t, P_euler = metodo_euler(f, P0, t0, tf, h)

# Solución analítica para comparar (P' = 3P, P(0)=0.5)
P_analitica = 0.5 * np.exp(3 * t)

# Crear tabla similar a la del pizarrón
print("Método de Euler - Tabla de resultados:")
print("t\t\tP(t) Euler\t\tP(t) Analítica\t\tError")
print("-" * 70)
for i in range(len(t)):
    error = abs(P_euler[i] - P_analitica[i])
    print(f"{t[i]:.1f}\t\t{P_euler[i]:.6f}\t\t{P_analitica[i]:.6f}\t\t{error:.6f}")

# Gráfico comparativo
plt.figure(figsize=(10, 6))
plt.plot(t, P_euler, 'bo-', label='Método de Euler', linewidth=2, markersize=6)
plt.plot(t, P_analitica, 'r-', label='Solución Analítica', linewidth=2)
plt.xlabel('Tiempo (t)')
plt.ylabel('P(t)')
plt.title('Método de Euler: $\dot{P} = 3P$, $P(0) = 0.5$')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Paso 1
P0 = 0.5
f_P0 = f(P0, 0)
P1 = P0 + f_P0 * h
print(f"Paso 1:")
print(f"t₀ = 0.0, P₀ = {P0}")
print(f"f(P₀) = 3 × {P0} = {f_P0}")
print(f"P₁ = P₀ + f(P₀)×h = {P0} + {f_P0}×{h} = {P1}")

# Paso 2
f_P1 = f(P1, 0.1)
P2 = P1 + f_P1 * h
print(f"\nPaso 2:")
print(f"t₁ = 0.1, P₁ = {P1}")
print(f"f(P₁) = 3 × {P1} = {f_P1}")
print(f"P₂ = P₁ + f(P₁)×h = {P1} + {f_P1}×{h} = {P2}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def RungeKutta2D(f, x0, t0, tf, h):
    """
    2nd-order Runge-Kutta para sistemas 2D
    
    Parameters:
    f: function f(x, t) que retorna [dx1/dt, dx2/dt]
    x0: condición inicial [x1_0, x2_0]
    t0: tiempo inicial
    tf: tiempo final
    h: paso de tiempo
    """
    T = []
    X1 = []
    X2 = []
    
    # Condiciones iniciales
    X1.append(x0[0])
    X2.append(x0[1])
    T.append(t0)
    
    t = t0
    x = x0.copy()  # [x1, x2]
    
    while t < tf:
        # Calcular los pasos de RK2 - ¡PASANDO EL TIEMPO t!
        k1 = f(x, t) * h
        k2 = f(x + k1/2, t + h/2) * h
        
        # Actualizar estado
        x = x + k2
        t = t + h
        
        # Guardar resultados
        X1.append(x[0])
        X2.append(x[1])
        T.append(t)
    
    return X1, X2, T

def oscilador_armonico(state, t):
    """
    Oscilador armónico: 
    dx/dt = v
    dv/dt = -ω²x
    """
    x, v = state
    omega = 1.0  # frecuencia
    return np.array([v, -omega**2 * x])

# Condiciones iniciales: x=1, v=0
x0 = [1.0, 0.0]
t0, tf, h = 0, 10, 0.01

# Resolver con RK2
X1, X2, T = RungeKutta2D(oscilador_armonico, x0, t0, tf, h)

# Graficar
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(T, X1, label='Posición (x)')
plt.plot(T, X2, label='Velocidad (v)')
plt.xlabel('Tiempo')
plt.ylabel('Estado')
plt.legend()
plt.grid(True)
plt.show()
```
