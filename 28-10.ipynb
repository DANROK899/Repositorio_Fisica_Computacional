```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos el rango de x e y
x = np.linspace(0, 1, 10)  # 10 puntos entre 0 y 1
y = np.linspace(0, 1, 10)  # 10 puntos entre 0 y 1# Creamos una malla de puntos

X, Y = np.meshgrid(x, y)

# E_x = -4*cos(x)*cos(y)
Ex = -4 * np.cos(X) * np.cos(Y)
# E_y = 4*sin(x)*sin(y)
Ey = 4 * np.sin(X) * np.sin(Y)

plt.figure(figsize=(8, 6))
plt.quiver(X, Y, Ex, Ey, color='purple', scale=20)

plt.title('Campo Eléctrico E(x,y) = [4 sen(x) cos(y)]')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

```python
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

X, Y = np.meshgrid(x, y)

Ex = -4 / (np.cos(X)**2) * np.exp(Y)
# E_x = -4 * sec^2(x) * e^y
Ey = -4 * np.tan(X) * np.exp(Y)
# E_y = -4 * tan(x) * e^y

plt.figure(figsize=(8, 6))

plt.quiver(X, Y, Ex, Ey, color='purple', scale=100)
plt.title('Campo Eléctrico E(x,y) = [4 tan(x) exp(y)]')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.show()
```
