```python
import numpy as np
import matplotlib.pyplot as plt

def psi(x, N=10):
    suma = 0
    cte = 4*np.sqrt(6)/(np.pi**2)

    for n in range(N):
        ec = ((-1)**n)/((2*n+1)**2)*np.sin((2*n+1)*np.pi*x)
        suma += ec

    return cte*suma

x = np.linspace(0, 1, 1000)
y = psi(x, N=50)

plt.figure(figsize=(10,10))
plt.plot(x,y,"r",linewidth=1)
plt.xlabel("x")
plt.ylabel("Ψ(x)")
plt.title("Ψ(x) = 4√9/π² Σ (-1)ⁿ/(2n+1)² sen((2n+1)πx)")
plt.grid(True, alpha=0.1)
plt.show()
```
