### 2. Integración numérica.

**Inciso a)**

Integración por regla del trapecio

```python
def funcion(x):
   
    return x**4 - 2*x + 1  

def integración_trapecio(a, b, tramos):
    h = (b - a) / tramos 
    suma = funcion(a)  
    for i in range(1, tramos):
        x_i = a + i * h  
        suma += 2 * funcion(x_i)  
    suma += funcion(b)  
    area = h * (suma / 2)  
    
    error = area-4.4
    
    print("El área bajo la curva obtenida fue de", area)
    
    print( "El error en este método utilizando", tramos, "numero de tramos, comparado con el valor esperado es de:", error)

    
```

```python
integración_trapecio(0,2,10)
```

```python
def funcion(x):
   
    return x**4 - 2*x + 1 

def integracion_simpson(a,b, n):
    h= (b-a)/n
    
    x=list()
    fx = list()
    i = 0
    while i<= n: 
        x.append(a + i * h) 
        fx.append(funcion(x[i])) 
        i += 1
  
    res = 0
    i = 0
    while i<= n: 
        if i == 0 or i == n: 
            res+= fx[i] 
        elif i % 2 != 0: 
            res+= 4 * fx[i] 
        else: 
            res+= 2 * fx[i] 
        i+= 1
    res = res * (h / 3) 
    error = res-4.4
    print("El área bajo la curva obtenida fue de", res)
    
    print( "El error en este método utilizando", n, "numero de tramos, comparado con el valor esperado es de:", error)

  
    
```

### 3. El lımite de difraccion de un telescopio.

**Inciso a)**

```python
import numpy as np
import matplotlib.pyplot as plt

def J(m, x):
    def integrando(theta):
        return np.cos(m * theta - x * np.sin(theta))
    
    N = 1000
    a = 0
    b = np.pi
    h = (b - a) / N
    suma = integrando(a) + integrando(b)
    for i in range(1, N, 2):
        suma += 4 * integrando(a + i * h)
    for i in range(2, N-1, 2):
        suma += 2 * integrando(a + i * h)
    return suma * h / 3

valores_x = np.linspace(0, 20, 1000)

valores_J0 = [J(0, x) for x in valores_x]
valores_J1 = [J(1, x) for x in valores_x]
valores_J2 = [J(2, x) for x in valores_x]

plt.figure(figsize=(10, 6))
plt.plot(valores_x, valores_J0, label='$J_0(x)$')
plt.plot(valores_x, valores_J1, label='$J_1(x)$')
plt.plot(valores_x, valores_J2, label='$J_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$J_m(x)$')
plt.title('Funciones de Bessel')
plt.legend()
plt.grid(True)
plt.show()

```

### Capacidad calorífica de un sólido

**inciso a) y b)**

```python
import math
from scipy import integrate
import matplotlib.pyplot as plt

def capacidad_calorifica(T):
    V = 0.001
    rho = 6.022e28
    Td = 428
    kb = 1.38064852e-23

    f = lambda x: (x**4 * math.exp(x)) / (math.exp(x) - 1)**2
    INT, _ = integrate.quad(f, 0, Td / T)
    valor = 9 * V * rho * kb * ((T / Td) ** 3) * INT
    return valor

cv = []
T = list(range(5, 500))

for temperatura in T:
    cv.append(capacidad_calorifica(temperatura))

plt.plot(T, cv, 'ro')
plt.xlabel('Temperatura (K)')
plt.ylabel('Cv (J/K)')
plt.title('Capacidad Calorífica del Aluminio con la Temperatura')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import dblquad

G = 6.67430e-11
M = 1.0e4
longitud_lado = 10
densidad = M / longitud_lado**2

def integrando(y, x, z):
    r = np.sqrt(x**2 + y**2 + z**2)
    return -G * densidad * z / r**3

def fuerza(z):
    integral, _ = dblquad(integrando, -longitud_lado/2, longitud_lado/2,
                          lambda x: -longitud_lado/2, lambda x: longitud_lado/2,
                          args=(z,))
    return integral

valores_z = np.linspace(0.1, 10, 100)
valores_fuerza = [fuerza(z) for z in valores_z]

plt.plot(valores_z, valores_fuerza)
plt.xlabel('Distancia desde la lámina (m)')
plt.ylabel('Fuerza (N)')
plt.title('Fuerza Gravitacional vs Distancia desde la Lámina')
plt.grid(True)
plt.show()
```
