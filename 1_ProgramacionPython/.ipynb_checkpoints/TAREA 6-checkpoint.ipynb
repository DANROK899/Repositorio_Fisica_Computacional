### 1. Productos vectoriales y matriciales

Escribe el pseudocódigo de los algoritmos que realicen las siguientes operaciones vectoriales; así como los respectivos programas en python, sin usar bibliotecas como numpy ni cualquier otra (i.e. usando sólo listas ) que implementen dichos algoritmos:

_a)_ El _producto interior_ (producto punto) de dos vectores en $\mathbb{R}^3$. En este caso, para la implementación el usuario del programa debe insertar las entradas de los vectores.

```python
def producto_punto_r3():
    
    # Leer primer vector
    print("\nIngrese las coordenadas del primer vector:")
    v1 = []
    for i in range(3):
        coord = float(input(f"Coordenada {i+1}: "))
        v1.append(coord)
    
    # Leer segundo vector
    print("\nIngrese las coordenadas del segundo vector:")
    v2 = []
    for i in range(3):
        coord = float(input(f"Coordenada {i+1}: "))
        v2.append(coord)
    
    # Calcular producto punto
    resultado = 0
    for i in range(3):
        resultado += v1[i] * v2[i]
    
    # Mostrar resultado
    print(f"\nVector 1: {v1}")
    print(f"Vector 2: {v2}")
    print(f"Producto punto: {resultado}")
    
    return resultado

# Ejecutar función
if __name__ == "__main__":
    producto_punto_r3()
```

_b)_ El producto interior de dos vectores en $\mathbb{R}^n$. Para la implementación, en este caso, encuentra una manera de generar dos vectores diferentes en $\mathbb{R}^{n-1}$.

```python
import random

def producto_punto_rn(n):
    v1 = [round(random.uniform(-10, 10), 2) for _ in range(n)]
    v2 = [round(random.uniform(-10, 10), 2) for _ in range(n)]
    
    resultado = 0
    for i in range(n):
        resultado += v1[i] * v2[i]
    
    print(f"v1: {v1}")
    print(f"v2: {v2}")
    print(f"Producto punto: {resultado:.2f}")
    
    return resultado

if __name__ == "__main__":
    n = int(input("Dimensión n: "))
    producto_punto_rn(n)
```

_c)_ El producto de una matriz en $\mathbb{A} \in \mathbb{R}^{nxn}$, por un vector de $\mathbb{R}^n$. Para la implementación las entradas tanto del vector como de la matriz deben leerse de un archivo de datos (que tú puedes generar).

```python
def leer_matriz_archivo(nombre_archivo):
    matriz = []
    with open(nombre_archivo, 'r') as f:
        for linea in f:
            fila = [float(x) for x in linea.split()]
            matriz.append(fila)
    return matriz

def leer_vector_archivo(nombre_archivo):
    with open(nombre_archivo, 'r') as f:
        linea = f.readline()
        vector = [float(x) for x in linea.split()]
    return vector

def matriz_por_vector():
    matriz = leer_matriz_archivo('matriz.txt')
    vector = leer_vector_archivo('vector.txt')
    
    n = len(matriz)
    resultado = [0] * n
    
    for i in range(n):
        for j in range(n):
            resultado[i] += matriz[i][j] * vector[j]
    
    print(f"Resultado: {resultado}")
    return resultado

if __name__ == "__main__":
    matriz_por_vector()
```

_d)_ El producto de dos matrices $\mathbb{A}, \mathbb{B} \in \mathbb{R}^{nxn}$. En la implementación las entradas de ambas matrices deben leerse de un archivo de datos (que tú generes).

```python
def leer_matriz_archivo(nombre_archivo):
    matriz = []
    with open(nombre_archivo, 'r') as f:
        for linea in f:
            fila = [float(x) for x in linea.split()]
            matriz.append(fila)
    return matriz

def matriz_por_matriz():
    A = leer_matriz_archivo('matriz_A.txt')
    B = leer_matriz_archivo('matriz_B.txt')
    
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    
    print("Resultado:")
    for fila in C:
        print(fila)
    
    return C

if __name__ == "__main__":
    matriz_por_matriz()
```

_e)_ Convierte tus códigos en funciones y generaliza todo en un única función que realice el producto de dos matrices $\mathbb{A}, \in \mathbb{R}^{nm}$ y $\mathbb{B} \in \mathbb{R}^{mn}$, así como el cálculo de cualquiera de los incisos anteriores, dependiendo de la entrada que se le de a la función. En todos los casos, las entradas deben leerse de un archivo de datos.

```python
def leer_datos_archivo(nombre_archivo):
    with open(nombre_archivo, 'r') as f:
        lineas = f.readlines()
    
    if len(lineas) == 1:
        return [float(x) for x in lineas[0].split()]
    else:
        matriz = []
        for linea in lineas:
            fila = [float(x) for x in linea.split()]
            matriz.append(fila)
        return matriz

def producto_generalizado(archivo_A, archivo_B):
    A = leer_datos_archivo(archivo_A)
    B = leer_datos_archivo(archivo_B)
    
    if isinstance(A[0], list) and isinstance(B[0], list):
        # Matriz × Matriz
        n = len(A)
        m = len(B[0])
        p = len(B)
        
        C = [[0] * m for _ in range(n)]
        for i in range(n):
            for j in range(m):
                for k in range(p):
                    C[i][j] += A[i][k] * B[k][j]
        return C
    
    elif isinstance(A[0], list) and not isinstance(B[0], list):
        # Matriz × Vector
        n = len(A)
        resultado = [0] * n
        for i in range(n):
            for j in range(len(B)):
                resultado[i] += A[i][j] * B[j]
        return resultado
    
    else:
        # Vector × Vector
        resultado = 0
        for i in range(len(A)):
            resultado += A[i] * B[i]
        return resultado

if __name__ == "__main__":
    # Ejemplo de uso
    resultado = producto_generalizado('matriz_A.txt', 'matriz_B.txt')
    print(resultado)
```

### 2. Eliminación Gaussiana.

Considera el siguiente sistema de ecuaciones lineales:

$$
\begin{aligned}
2t + x + 4y + z &= -4 \\
3t + 4x - y - z &= 3 \\
t - 4x + y + 5z &= 9 \\
2t - 2x + y + 3z &= 7
\end{aligned}
$$

_a)_ Resuelve el sistema ("a mano") usando el método de Eliminación Gaussiana. Realiza el pivoteo que consideres más adecuado y muestra cada una de las operaciones elementales realizadas en cada paso hasta que logres diagonalizar el sistema para aplicar la sustitución hacia atrás (backsubstitution).

$$
\begin{aligned}
\text{[1]}\quad & 2t + x + 4y + z = -4 \\
\text{[2]}\quad & 3t + 4x - y - z = 3 \\
\text{[3]}\quad & t - 4x + y + 5z = 9 \\
\text{[4]}\quad & 2t - 2x + y + 3z = 7
\end{aligned}
$$


$$
\left[\begin{array}{cccc|c}
2 & 1 & 4 & 1 & -4 \\
3 & 4 & -1 & -1 & 3 \\
1 & -4 & 1 & 5 & 9 \\
2 & -2 & 1 & 3 & 7
\end{array}\right]
$$


Intercambiamos F1 ↔ F2 para tener pivote más grande:

$$
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\
2 & 1 & 4 & 1 & -4 \\
1 & -4 & 1 & 5 & 9 \\
2 & -2 & 1 & 3 & 7
\end{array}\right]
$$

**F2 ← F2 - $\frac{2}{3}$F1:**
$$
\begin{aligned}
&[2,1,4,1|-4] - \frac{2}{3}[3,4,-1,-1|3] \\
&= [2,1,4,1|-4] - [2,\frac{8}{3},-\frac{2}{3},-\frac{2}{3}|2] \\
&= [0,-\frac{5}{3},\frac{14}{3},\frac{5}{3}|-6]
\end{aligned}
$$

**F3 ← F3 - $\frac{1}{3}$F1:**
$$
\begin{aligned}
&[1,-4,1,5|9] - \frac{1}{3}[3,4,-1,-1|3] \\
&= [1,-4,1,5|9] - [1,\frac{4}{3},-\frac{1}{3},-\frac{1}{3}|1] \\
&= [0,-\frac{16}{3},\frac{4}{3},\frac{16}{3}|8]
\end{aligned}
$$

**F4 ← F4 - $\frac{2}{3}$F1:**
$$
\begin{aligned}
&[2,-2,1,3|7] - \frac{2}{3}[3,4,-1,-1|3] \\
&= [2,-2,1,3|7] - [2,\frac{8}{3},-\frac{2}{3},-\frac{2}{3}|2] \\
&= [0,-\frac{14}{3},\frac{5}{3},\frac{11}{3}|5]
\end{aligned}
$$

**Matriz resultante:**
$$
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\
0 & -\frac{5}{3} & \frac{14}{3} & \frac{5}{3} & -6 \\
0 & -\frac{16}{3} & \frac{4}{3} & \frac{16}{3} & 8 \\
0 & -\frac{14}{3} & \frac{5}{3} & \frac{11}{3} & 5
\end{array}\right]
$$

F2 ↔ F3 (valor absoluto mayor):

$$
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\
0 & -\frac{16}{3} & \frac{4}{3} & \frac{16}{3} & 8 \\
0 & -\frac{5}{3} & \frac{14}{3} & \frac{5}{3} & -6 \\
0 & -\frac{14}{3} & \frac{5}{3} & \frac{11}{3} & 5
\end{array}\right]
$$


**F3 ← F3 - $\frac{5}{16}$F2:**
$$
\begin{aligned}
&\left[0,-\frac{5}{3},\frac{14}{3},\frac{5}{3}|-6\right] - \frac{5}{16}\left[0,-\frac{16}{3},\frac{4}{3},\frac{16}{3}|8\right] \\
&= \left[0,-\frac{5}{3},\frac{14}{3},\frac{5}{3}|-6\right] - \left[0,-\frac{5}{3},\frac{5}{12},\frac{5}{3}|2.5\right] \\
&= \left[0,0,\frac{51}{12},0|-8.5\right]

$$

**F4 ← F4 - $\frac{7}{8}$F2:**
$$
\begin{aligned}
&\left[0,-\frac{14}{3},\frac{5}{3},\frac{11}{3}|5\right] - \frac{7}{8}\left[0,-\frac{16}{3},\frac{4}{3},\frac{16}{3}|8\right] \\
&= \left[0,-\frac{14}{3},\frac{5}{3},\frac{11}{3}|5\right] - \left[0,-\frac{14}{3},\frac{7}{6},\frac{14}{3}|7\right] \\
&= \left[0,0,\frac{1}{2},-1|-2\right]
\end{aligned}
$$

**Matriz resultante:**
$$
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\
0 & -\frac{16}{3} & \frac{4}{3} & \frac{16}{3} & 8 \\
0 & 0 & \frac{51}{12} & 0 & -8.5 \\
0 & 0 & \frac{1}{2} & -1 & -2
\end{array}\right]
$$


F3 ↔ F4:

$$
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\
0 & -\frac{16}{3} & \frac{4}{3} & \frac{16}{3} & 8 \\
0 & 0 & \frac{1}{2} & -1 & -2 \\
0 & 0 & \frac{51}{12} & 0 & -8.5
\end{array}\right]
$$


**F4 ← F4 - $\frac{17}{2}$F3:**
$$
\begin{aligned}
&\left[0,0,\frac{51}{12},0|-8.5\right] - \frac{17}{2}\left[0,0,\frac{1}{2},-1|-2\right] \\
&= \left[0,0,\frac{51}{12},0|-8.5\right] - \left[0,0,\frac{17}{4},-\frac{17}{2}|-17\right] \\
&= \left[0,0,0,\frac{17}{2}|8.5\right]
\end{aligned}
$$

$$
\left[\begin{array}{cccc|c}

\end{array}\right]
$$

**De F4:**
$$
\frac{17}{2}z = 8.5 \quad \Rightarrow \quad z = 8.5 \times \frac{2}{17} = 1
$$

**De F3:**
$$
\frac{1}{2}y - z = -2 \quad \Rightarrow \quad \frac{1}{2}y - 1 = -2 \quad \Rightarrow \quad \frac{1}{2}y = -1 \quad \Rightarrow \quad y = -2
$$

**De F2:**
$$
\begin{aligned}
&-\frac{16}{3}x + \frac{4}{3}y + \frac{16}{3}z = 8 \\
&-\frac{16}{3}x + \frac{4}{3}(-2) + \frac{16}{3}(1) = 8 \\
&-\frac{16}{3}x - \frac{8}{3} + \frac{16}{3} = 8 \\
&-\frac{16}{3}x + \frac{8}{3} = 8 \\
&-\frac{16}{3}x = 8 - \frac{8}{3} = \frac{16}{3} \\

\end{aligned}
$$

**De F1:**
$$
\begin{aligned}
&3t + 4x - y - z = 3 \\
&3t + 4(-1) - (-2) - 1 = 3 \\
&3t - 4 + 2 - 1 = 3 \\
&3t - 3 = 3 \\
&3t = 6 \quad \Rightarrow \quad t = 2
\end{aligned}
$$

$$
\therefore \boxed{t = 2,\quad x = -1,\quad y = -2,\quad z = 1}
$$

 _b)_ A partir de lo anterior escribe un programa para realizar la Eliminación Gaussiana de cualquier sistema de ecuaciones lineales

```python
def eliminacion_gaussiana(A, b):
    n = len(A)
    
    # Matriz aumentada
    for i in range(n):
        A[i].append(b[i])
    
    # Eliminación hacia adelante
    for i in range(n):
        # Pivoteo
        max_fila = i
        for k in range(i + 1, n):
            if abs(A[k][i]) > abs(A[max_fila][i]):
                max_fila = k
        A[i], A[max_fila] = A[max_fila], A[i]
        
        # Eliminación
        for k in range(i + 1, n):
            factor = A[k][i] / A[i][i]
            for j in range(i, n + 1):
                A[k][j] -= factor * A[i][j]
    
    # Sustitución hacia atrás
    x = [0] * n
    for i in range(n - 1, -1, -1):
        x[i] = A[i][n]
        for j in range(i + 1, n):
            x[i] -= A[i][j] * x[j]
        x[i] /= A[i][i]
    
    return x

# Sistema del problema
A = [
    [2, 1, 4, 1],
    [3, 4, -1, -1],
    [1, -4, 1, 5],
    [2, -2, 1, 3]
]
b = [-4, 3, 9, 7]

solucion = eliminacion_gaussiana(A, b)
print("Solución:", [f"{val:.2f}" for val in solucion])
```

### 3. Pozo cuántico asimétrico.

La mecánica cuántica se puede formular como un problema matricial y resolverse en una computadora usando métodos de álgebra lineal. Supongamos, por ejemplo, que tenemos una partícula de masa $M$ en un pozo cuántico unidimensional de ancho $L$, pero no un pozo cuadrado como los ejemplos que probablemente hayas visto antes. Supongamos en cambio que el potencial $V(x)$ varía de alguna manera dentro del pozo:

$$x = 0 \quad x = L$$

No es posible resolver analíticamente este problema de forma general, pero podemos resolverlo en la computadora.

En un estado puro de energía $E$, la parte espacial de la función de onda obedece a la ecuación de Schrödinger independiente del tiempo $\hat{H} \psi(x) = E \psi(x)$, donde el operador hamiltoniano $\hat{H}$ esta dado por:

$$\hat{H} = -\frac{\hbar^2}{2M} \frac{d^2}{dx^2} + V(x).$$

Para simplificar, supongamos que las paredes del pozo son infinitamente altas, de modo que la función de onda es cero fuera del pozo, lo que significa que debe ir a cero en $x = 0$ y $x = L$. En ese caso, la función de onda se puede expresar como una serie sinusoidal de Fourier de la siguiente forma:

$$\psi(x) = \sum_{n=1}^{\infty} \psi_n \sin \frac{\pi nx}{L},$$

donde $\psi_1, \psi_2, \ldots$ son los coeficientes de Fourier.

_a)_ Observando que, para $m, n$ enteros positivos

$$\int_{0}^{L} \sin \frac{\pi mx}{L} \sin \frac{\pi nx}{L}  dx = 
\begin{cases} 
L/2 & \text{si } m = n, \\
0 & \text{de lo contrario},
\end{cases}$$

demuestra que la ecuación de Schrödinger $\hat{H}\psi = E\psi$ implica que:

$$\sum_{n=1}^{\infty} \psi_n \int_{0}^{L} \sin \frac{\pi mx}{L} \hat{H} \sin \frac{\pi nx}{L}  dx = \frac{1}{2} LE \psi_m.$$

Así, definiendo una matriz $H$ con elementos:

$$H_{mn} = \frac{2}{L} \int_{0}^{L} \sin \left( \frac{\pi mx}{L} \right) \hat{H} \sin \left( \frac{\pi nx}{L} \right)  dx$$

$$= \frac{2}{L} \int_{0}^{L} \sin \left( \frac{\pi mx}{L} \right) \left[ -\frac{\hbar^2}{2M} \frac{d^2}{dx^2} + V(x) \right] \sin \left( \frac{\pi nx}{L} \right)  dx,$$

demuestra que la ecuación de Schrödinger se puede escribir en forma matricial como $H\psi = E\psi$, donde $\psi$ es el vector $(\psi_1, \psi_2, \ldots)$. Por lo tanto, $\psi$ es un vector propio de la _matriz hamiltoniana_ $H$ con valor propio $E$. Si podemos calcular los valores propios de esta matriz, entonces conoceremos las energías permitidas de la partícula en el pozo.

**Demostración:**

Partimos de la ecuación de Schrödinger:

$$
\hat{H}\psi = E\psi
$$

Sustituimos la expansión de Fourier:
$$
\hat{H}\sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right) = E\sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right)
$$

Multiplicamos por $\sin\left(\frac{\pi m x}{L}\right)$ e integramos:
$$
\int_0^L \sin\left(\frac{\pi m x}{L}\right)\hat{H}\sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right) dx = E\int_0^L \sin\left(\frac{\pi m x}{L}\right)\sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right) dx
$$

Intercambiamos suma e integral:
$$
\sum_{n=1}^{\infty} \psi_n \int_0^L \sin\left(\frac{\pi m x}{L}\right)\hat{H}\sin\left(\frac{\pi n x}{L}\right) dx = E\sum_{n=1}^{\infty} \psi_n \int_0^L \sin\left(\frac{\pi m x}{L}\right)\sin\left(\frac{\pi n x}{L}\right) dx
$$

Usamos la ortogonalidad:
$$
\sum_{n=1}^{\infty} \psi_n \int_0^L \sin\left(\frac{\pi m x}{L}\right)\hat{H}\sin\left(\frac{\pi n x}{L}\right) dx = E\psi_m \cdot \frac{L}{2}
$$

Definimos:
$$
H_{mn} = \frac{2}{L}\int_0^L \sin\left(\frac{\pi m x}{L}\right)\hat{H}\sin\left(\frac{\pi n x}{L}\right) dx
$$

Multiplicando por $\frac{2}{L}$:
$$
\sum_{n=1}^{\infty} H_{mn}\psi_n = E\psi_m
$$

En forma matricial:
$$
\mathbf{H}\boldsymbol{\psi} = E\boldsymbol{\psi}
$$

donde $\mathbf{H}$ es la matriz hamiltoniana y $\boldsymbol{\psi} = (\psi_1, \psi_2, \ldots)^T$.

_b)_ Para el caso $V(x) = ax/L$, evalúa la integral en $H_{mn}$ analíticamente y encuentra una expresión general para el elemento de la matriz $H_{mn}$. Demuestra que la matriz es real y simétrica. Probablemente te resulte útil saber que

$$\int_0^L x \sin \left( \frac{\pi mx}{L} \right) \sin \left( \frac{\pi nx}{L} \right)  dx = 
\begin{cases} 
0 & \text{si } m \neq n \text{ ambos pares o impares,} \\
-\left( \frac{2L}{\pi} \right)^2 \frac{mn}{(m^2 - n^2)^2} & \text{si } m \neq n \text{ uno par y otro impar,} \\
L^2/4 & \text{si } m = n.
\end{cases}$$

Escribe un programa para evaluar tu expresión para $H_{mn}$ con $m$ y $n$ arbitrarios; cuando la partícula en el pozo es un electrón, el pozo tiene un ancho 5 Å y $a = 10 \, \text{eV}$. (La masa y la carga de un electrón son $9.1094 \times 10^{-31} \, \text{kg}$ y $1.6022 \times 10^{-19} \, \text{C}$ respectivamente.)

```python
import numpy as np

# Constantes físicas
hbar = 1.0545718e-34  # J·s
M = 9.1094e-31        # kg
e_charge = 1.6022e-19 # C
L = 5e-10             # m
a = 10 * e_charge     # J

def H_mn(m, n):
    """Elemento H_{mn} de la matriz Hamiltoniana"""
    # Parte cinética
    T = (hbar**2 * np.pi**2 * n**2) / (2 * M * L**2)
    
    # Parte potencial
    if m == n:
        V = a / 2
    elif (m % 2 == 0 and n % 2 == 1) or (m % 2 == 1 and n % 2 == 0):
        V = - (8 * a / np.pi**2) * (m * n) / ((m**2 - n**2)**2)
    else:
        V = 0
    
    return (T if m == n else 0) + V

# Construir matriz Hamiltoniana 10x10
N = 10
H = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        H[i, j] = H_mn(i+1, j+1)

# Convertir a eV para mostrar
H_eV = H / e_charge
print("Matriz Hamiltoniana (eV):")
print(np.round(H_eV, 4))

# Verificar simetría
print(f"\nEs simétrica: {np.allclose(H, H.T)}")
```

_c)_ En teoría, la matriz $H$ es infinitamente grande, por lo que no podemos calcular todos sus valores propios. Pero podemos obtener una solución bastante precisa de los primeros recortando la matriz. Modifica el programa que escribiste para el inciso anterior para crear una matriz de $10 \times 10$ de los elementos de $H$ (i.e. $m, n = 10$). Calcula los valores propios de esta matriz usando la función apropiada de la biblioteca `numpy.linalg` y entonces imprime (en unidades de electronvolts) los primeros diez niveles de energía del pozo cuántico, dentro de esta aproximación. Deberías encontrar, por ejemplo, que la energía del estado fundamental del sistema es de alrededor de 5,84 eV. (Hint: ten en cuenta que los índices matriciales en Python comienzan en cero, mientras que los índices en expresiones algebraicas estándar, como las anteriores, comienzan en uno.
```python
import numpy as np

# Constantes físicas
hbar = 1.0545718e-34  # J·s
M = 9.1094e-31        # kg
e_charge = 1.6022e-19 # C
L = 5e-10             # m
a = 10 * e_charge     # J

def H_mn(m, n):
    """Elemento H_{mn} de la matriz Hamiltoniana"""
    # Parte cinética
    T = (hbar**2 * np.pi**2 * n**2) / (2 * M * L**2)
    
    # Parte potencial
    if m == n:
        V = a / 2
    elif (m % 2 == 0 and n % 2 == 1) or (m % 2 == 1 and n % 2 == 0):
        V = - (8 * a / np.pi**2) * (m * n) / ((m**2 - n**2)**2)
    else:
        V = 0
    
    return (T if m == n else 0) + V

# Construir matriz Hamiltoniana 10x10
N = 10
H = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        H[i, j] = H_mn(i+1, j+1)  # +1 porque índices físicos empiezan en 1

# Calcular valores propios
energias = np.linalg.eigvalsh(H) / e_charge

print("Primeros 10 niveles de energía (eV):")
for i, E in enumerate(energias):
    print(f"E_{i+1} = {E:.4f} eV")
```

_e)_ Ahora modifica tu programa, una vez más, para calcular la función de onda $\psi(x)$ para el estado fundamental y los dos primeros estados excitados del pozo. Utiliza tus resultados para hacer una gráfica con tres curvas que muestren la densidad de probabilidad $|\psi(x)|^2$ en función de $x$ para cada uno de estos tres estados. Presta mucha atención a la normalización de la función de onda: debe satisfacer la condición $\int_0^L |\psi(x)|^2 dx = 1$. ¿Esto es cierto para tu función de onda?

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes físicas
hbar = 1.0545718e-34
M = 9.1094e-31
e_charge = 1.6022e-19
L = 5e-10
a = 10 * e_charge

def H_mn(m, n):
    T = (hbar**2 * np.pi**2 * n**2) / (2 * M * L**2)
    if m == n:
        V = a / 2
    elif (m % 2 == 0 and n % 2 == 1) or (m % 2 == 1 and n % 2 == 0):
        V = - (8 * a / np.pi**2) * (m * n) / ((m**2 - n**2)**2)
    else:
        V = 0
    return (T if m == n else 0) + V

# Construir matriz 100x100
N = 100
H = np.zeros((N, N))
for i in range(N):
    for j in range(N):
        H[i, j] = H_mn(i+1, j+1)

# Calcular autovalores y autovectores
energias, autovectores = np.linalg.eigh(H)
energias_eV = energias / e_charge

# Puntos espaciales
x = np.linspace(0, L, 1000)
x_angstrom = x * 1e10

# Funciones de onda para primeros 3 estados
psi_funciones = []
for i in range(3):
    def psi_i(x_val, coefs=autovectores[:, i]):
        resultado = 0
        for n in range(1, N+1):
            resultado += coefs[n-1] * np.sin(n * np.pi * x_val / L)
        return resultado
    psi_funciones.append(psi_i)

# Normalizar
for i in range(3):
    psi_vals = psi_funciones[i](x)
    norma = np.sqrt(np.trapz(psi_vals**2, x))
    def psi_norm(x_val, psi_orig=psi_funciones[i], norm=norma):
        return psi_orig(x_val) / norm
    psi_funciones[i] = psi_norm

# Verificar normalización
print("Verificación de normalización:")
for i in range(3):
    psi_vals = psi_funciones[i](x)
    norma = np.trapz(psi_vals**2, x)
    print(f"Estado {i+1}: ∫|ψ|²dx = {norma:.6f}")

# Graficar densidades de probabilidad
plt.figure(figsize=(10, 6))
for i in range(3):
    psi_vals = psi_funciones[i](x)
    densidad = psi_vals**2
    plt.plot(x_angstrom, densidad, linewidth=2, label=f'Estado {i+1} (E={energias_eV[i]:.2f} eV)')

plt.xlabel('x (Å)')
plt.ylabel('|ψ(x)|²')
plt.title('Densidades de Probabilidad - Pozo Cuántico Asimétrico')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
