```python
def leer_matriz_archivo(nombre_archivo):
    """Lee una matriz desde un archivo"""
    matriz = []
    with open(nombre_archivo, 'r') as f:
        for linea in f:
            fila = [float(x) for x in linea.strip().split()]
            matriz.append(fila)
    return matriz

def leer_vector_archivo(nombre_archivo):
    """Lee un vector desde un archivo"""
    with open(nombre_archivo, 'r') as f:
        linea = f.readline()
        vector = [float(x) for x in linea.strip().split()]
    return vector

def matriz_por_vector():
    """
    Producto de matriz n×n por vector n
    Los datos se leen desde archivos
    """
    print("=== MATRIZ × VECTOR ===")
    
    # Leer datos desde archivos
    matriz = leer_matriz_archivo('matriz.txt')
    vector = leer_vector_archivo('vector.txt')
    
    n = len(matriz)
    
    # Verificar dimensiones
    if len(matriz[0]) != n or len(vector) != n:
        print("Error: Dimensiones incompatibles")
        return None
    
    # Calcular producto
    resultado = [0] * n
    for i in range(n):
        for j in range(n):
            resultado[i] += matriz[i][j] * vector[j]
    
    # Mostrar resultado
    print(f"Matriz {n}×{n}:")
    for fila in matriz:
        print(fila)
    
    print(f"\nVector: {vector}")
    print(f"\nResultado: {resultado}")
    
    return resultado

# Crear archivos de ejemplo (ejecutar una vez)
def crear_archivos_ejemplo():
    """Crea archivos de ejemplo para probar"""
    # Archivo matriz.txt
    with open('matriz.txt', 'w') as f:
        f.write("1 2 3\n")
        f.write("4 5 6\n")
        f.write("7 8 9\n")
    
    # Archivo vector.txt
    with open('vector.txt', 'w') as f:
        f.write("1 0 2\n")

# Ejecutar función
if __name__ == "__main__":
    crear_archivos_ejemplo()  # Ejecutar solo una vez
    matriz_por_vector()
```

```python
import random

def leer_datos_archivo(nombre_archivo):
    """Lee datos desde un archivo, detectando si es vector o matriz"""
    datos = []
    with open(nombre_archivo, 'r') as f:
        lineas = f.readlines()
    
    if len(lineas) == 1:
        # Es un vector
        return [float(x) for x in lineas[0].strip().split()]
    else:
        # Es una matriz
        matriz = []
        for linea in lineas:
            fila = [float(x) for x in linea.strip().split()]
            matriz.append(fila)
        return matriz

def producto_generalizado(archivo_A, archivo_B):
    """
    Función generalizada que realiza cualquier producto:
    - Vector × Vector (producto punto)
    - Matriz × Vector
    - Matriz × Matriz
    """
    print("=== FUNCIÓN GENERALIZADA DE PRODUCTO ===")
    
    # Leer datos desde archivos
    A = leer_datos_archivo(archivo_A)
    B = leer_datos_archivo(archivo_B)
    
    print(f"Archivo A: {type(A)}")
    print(f"Archivo B: {type(B)}")
    
    # Determinar tipo de operación
    if isinstance(A, list) and isinstance(B, list):
        if isinstance(A[0], list) and isinstance(B[0], list):
            # Matriz × Matriz
            print("Operación: Matriz × Matriz")
            return producto_matrices(A, B)
        
        elif isinstance(A[0], list) and not isinstance(B[0], list):
            # Matriz × Vector
            print("Operación: Matriz × Vector")
            return producto_matriz_vector(A, B)
        
        elif not isinstance(A[0], list) and not isinstance(B[0], list):
            # Vector × Vector (producto punto)
            print("Operación: Vector × Vector (producto punto)")
            return producto_punto(A, B)
    
    print("Error: Combinación de tipos no soportada")
    return None

def producto_punto(v1, v2):
    """Producto punto de dos vectores"""
    if len(v1) != len(v2):
        print("Error: Vectores de diferente dimensión")
        return None
    
    resultado = 0
    for i in range(len(v1)):
        resultado += v1[i] * v2[i]
    
    print(f"Vector 1: {v1}")
    print(f"Vector 2: {v2}")
    print(f"Producto punto: {resultado}")
    return resultado

def producto_matriz_vector(matriz, vector):
    """Producto matriz × vector"""
    n = len(matriz)
    if len(matriz[0]) != len(vector):
        print("Error: Dimensiones incompatibles")
        return None
    
    resultado = [0] * n
    for i in range(n):
        for j in range(len(vector)):
            resultado[i] += matriz[i][j] * vector[j]
    
    print("Matriz:")
    for fila in matriz:
        print(fila)
    print(f"Vector: {vector}")
    print(f"Resultado: {resultado}")
    return resultado

def producto_matrices(A, B):
    """Producto de dos matrices"""
    n = len(A)
    m = len(B[0])
    p = len(B)
    
    if len(A[0]) != p:
        print("Error: Dimensiones incompatibles")
        return None
    
    # Inicializar matriz resultado
    C = [[0] * m for _ in range(n)]
    
    # Calcular producto
    for i in range(n):
        for j in range(m):
            for k in range(p):
                C[i][j] += A[i][k] * B[k][j]
    
    print("Matriz A:")
    for fila in A:
        print(fila)
    print("Matriz B:")
    for fila in B:
        print(fila)
    print("Resultado A × B:")
    for fila in C:
        print(fila)
    
    return C

# Crear archivos de ejemplo para probar todas las operaciones
def crear_archivos_ejemplo():
    """Crea archivos de ejemplo para todas las operaciones"""
    
    # Vector × Vector
    with open('vector1.txt', 'w') as f:
        f.write("1 2 3\n")
    
    with open('vector2.txt', 'w') as f:
        f.write("4 5 6\n")
    
    # Matriz × Vector
    with open('matriz_mv.txt', 'w') as f:
        f.write("1 2 3\n")
        f.write("4 5 6\n")
        f.write("7 8 9\n")
    
    with open('vector_mv.txt', 'w') as f:
        f.write("1 0 2\n")
    
    # Matriz × Matriz
    with open('matriz_A_mm.txt', 'w') as f:
        f.write("1 2\n")
        f.write("3 4\n")
    
    with open('matriz_B_mm.txt', 'w') as f:
        f.write("5 6\n")
        f.write("7 8\n")

# Ejecutar pruebas
if __name__ == "__main__":
    crear_archivos_ejemplo()
    
    print("=" * 50)
    producto_generalizado('vector1.txt', 'vector2.txt')
    
    print("\n" + "=" * 50)
    producto_generalizado('matriz_mv.txt', 'vector_mv.txt')
    
    print("\n" + "=" * 50)
    producto_generalizado('matriz_A_mm.txt', 'matriz_B_mm.txt')
```

## **Sistema original:**

\[
\begin{aligned}
\text{[1]}\quad & 2t + x + 4y + z = -4 \\
\text{[2]}\quad & 3t + 4x - y - z = 3 \\
\text{[3]}\quad & t - 4x + y + 5z = 9 \\
\text{[4]}\quad & 2t - 2x + y + 3z = 7
\end{aligned}
\]

## **Matriz aumentada inicial:**

\[
\left[\begin{array}{cccc|c}
2 & 1 & 4 & 1 & -4 \\
3 & 4 & -1 & -1 & 3 \\
1 & -4 & 1 & 5 & 9 \\
2 & -2 & 1 & 3 & 7
\end{array}\right]
\]

---

## **Paso 1: Pivoteo parcial - intercambiar filas 1 y 2**

Intercambiamos F1 ↔ F2 para tener pivote más grande:

\[
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\
2 & 1 & 4 & 1 & -4 \\
1 & -4 & 1 & 5 & 9 \\
2 & -2 & 1 & 3 & 7
\end{array}\right]
\]

---

## **Paso 2: Eliminación en columna 1**

**F2 ← F2 - $\frac{2}{3}$F1:**
\[
\begin{aligned}
&[2,1,4,1|-4] - \frac{2}{3}[3,4,-1,-1|3] \\
&= [2,1,4,1|-4] - [2,\frac{8}{3},-\frac{2}{3},-\frac{2}{3}|2] \\
&= [0,-\frac{5}{3},\frac{14}{3},\frac{5}{3}|-6]
\end{aligned}
\]

**F3 ← F3 - $\frac{1}{3}$F1:**
\[
\begin{aligned}
&[1,-4,1,5|9] - \frac{1}{3}[3,4,-1,-1|3] \\
&= [1,-4,1,5|9] - [1,\frac{4}{3},-\frac{1}{3},-\frac{1}{3}|1] \\
&= [0,-\frac{16}{3},\frac{4}{3},\frac{16}{3}|8]
\end{aligned}
\]

**F4 ← F4 - $\frac{2}{3}$F1:**
\[
\begin{aligned}
&[2,-2,1,3|7] - \frac{2}{3}[3,4,-1,-1|3] \\
&= [2,-2,1,3|7] - [2,\frac{8}{3},-\frac{2}{3},-\frac{2}{3}|2] \\
&= [0,-\frac{14}{3},\frac{5}{3},\frac{11}{3}|5]
\end{aligned}
\]

**Matriz resultante:**
\[
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\
0 & -\frac{5}{3} & \frac{14}{3} & \frac{5}{3} & -6 \\
0 & -\frac{16}{3} & \frac{4}{3} & \frac{16}{3} & 8 \\
0 & -\frac{14}{3} & \frac{5}{3} & \frac{11}{3} & 5
\end{array}\right]
\]

---

## **Paso 3: Pivoteo en columna 2 - intercambiar filas 2 y 3**

F2 ↔ F3 (valor absoluto mayor):

\[
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\
0 & -\frac{16}{3} & \frac{4}{3} & \frac{16}{3} & 8 \\
0 & -\frac{5}{3} & \frac{14}{3} & \frac{5}{3} & -6 \\

\end{array}\right]

## **Paso 4: Eliminación en columna 2**

**F3 ← F3 - $\frac{5}{16}$F2:**
\[
\begin{aligned}

**F4 ← F4 - $\frac{7}{8}$F2:**

&= \left[0,0,\frac{1}{2},-1|-2\right]
\end{aligned}
\]

**Matriz resultante:**

\left[\begin{array}{cccc|c}

## **Paso 5: Pivoteo en columna 3 - intercambiar filas 3 y 4**

F3 ↔ F4:

## **Paso 6: Eliminación en columna 3**

**F4 ← F4 - $\frac{17}{2}$F3:**
\[
\begin{aligned}

\[
\left[\begin{array}{cccc|c}
3 & 4 & -1 & -1 & 3 \\

0 & 0 & \frac{1}{2} & -1 & -2 \\
0 & 0 & 0 & \frac{17}{2} & 8.5
\end{array}\right]
\]

## **Paso 7: Sustitución hacia atrás**

**De F4:**
\[
\frac{17}{2}z = 8.5 \quad \Rightarrow \quad z = 8.5 \times \frac{2}{17} = 1
\]

**De F3:**
\[
\frac{1}{2}y - z = -2 \quad \Rightarrow \quad \frac{1}{2}y - 1 = -2 \quad \Rightarrow \quad \frac{1}{2}y = -1 \quad \Rightarrow \quad y = -2
\]

**De F2:**
\[
\begin{aligned}
&-\frac{16}{3}x + \frac{4}{3}y + \frac{16}{3}z = 8 \\
&-\frac{16}{3}x + \frac{4}{3}(-2) + \frac{16}{3}(1) = 8 \\
&-\frac{16}{3}x - \frac{8}{3} + \frac{16}{3} = 8 \\
&-\frac{16}{3}x + \frac{8}{3} = 8 \\
&-\frac{16}{3}x = 8 - \frac{8}{3} = \frac{16}{3} \\
&x = \frac{16}{3} \times \left(-\frac{3}{16}\right) = -1
\end{aligned}
\]

**De F1:**
\[
\begin{aligned}
&3t + 4x - y - z = 3 \\
&3t + 4(-1) - (-2) - 1 = 3 \\
&3t - 4 + 2 - 1 = 3 \\
&3t - 3 = 3 \\
&3t = 6 \quad \Rightarrow \quad t = 2
\end{aligned}
\]

---

## **Solución final:**

\[
\boxed{t = 2,\quad x = -1,\quad y = -2,\quad z = 1}
\]

```python
def eliminacion_gaussiana(A, b):
    """
    Resuelve el sistema Ax = b usando Eliminación Gaussiana con pivoteo parcial
    """
    n = len(A)
    
    # Crear matriz aumentada
    aumentada = []
    for i in range(n):
        fila = A[i][:]  # Copiar fila de A
        fila.append(b[i])  # Añadir elemento de b
        aumentada.append(fila)
    
    print("Matriz aumentada inicial:")
    imprimir_matriz(aumentada)
    
    # Eliminación hacia adelante
    for i in range(n):
        # Pivoteo parcial
        max_fila = i
        for k in range(i + 1, n):
            if abs(aumentada[k][i]) > abs(aumentada[max_fila][i]):
                max_fila = k
        
        # Intercambiar filas
        if max_fila != i:
            aumentada[i], aumentada[max_fila] = aumentada[max_fila], aumentada[i]
            print(f"Intercambiado fila {i+1} con fila {max_fila+1}")
        
        # Verificar si el pivote es cero
        if abs(aumentada[i][i]) < 1e-10:
            print("Error: Sistema singular")
            return None
        
        # Hacer ceros debajo del pivote
        for k in range(i + 1, n):
            factor = aumentada[k][i] / aumentada[i][i]
            for j in range(i, n + 1):
                aumentada[k][j] -= factor * aumentada[i][j]
        
        print(f"\nDespués de eliminar columna {i+1}:")
        imprimir_matriz(aumentada)
    
    # Sustitución hacia atrás
    x = [0] * n
    for i in range(n - 1, -1, -1):
        x[i] = aumentada[i][n]  # Término independiente
        for j in range(i + 1, n):
            x[i] -= aumentada[i][j] * x[j]
        x[i] /= aumentada[i][i]
    
    return x

def imprimir_matriz(matriz):
    """Imprime una matriz de forma legible"""
    for fila in matriz:
        print("[", end="")
        for j, elemento in enumerate(fila):
            if j == len(fila) - 1:
                print(" |", end=" ")
            print(f"{elemento:8.4f}", end=" ")
        print("]")
    print()

def verificar_solucion(A, b, x):
    """Verifica que la solución sea correcta"""
    print("\nVerificación de la solución:")
    n = len(A)
    for i in range(n):
        calculado = 0
        for j in range(n):
            calculado += A[i][j] * x[j]
        print(f"Ecuación {i+1}: {calculado:.4f} = {b[i]:.4f} (error: {abs(calculado - b[i]):.2e})")

# Sistema del problema
def resolver_sistema_problema():
    """Resuelve el sistema específico del problema"""
    # Matriz A
    A = [
        [2, 1, 4, 1],
        [3, 4, -1, -1],
        [1, -4, 1, 5],
        [2, -2, 1, 3]
    ]
    
    # Vector b
    b = [-4, 3, 9, 7]
    
    print("SISTEMA DE ECUACIONES:")
    print("2t +  x + 4y +  z = -4")
    print("3t + 4x -  y -  z =  3")
    print(" t - 4x +  y + 5z =  9")
    print("2t - 2x +  y + 3z =  7")
    print()
    
    # Resolver
    solucion = eliminacion_gaussiana(A, b)
    
    if solucion:
        print("\nSOLUCIÓN:")
        variables = ['t', 'x', 'y', 'z']
        for i, var in enumerate(variables):
            print(f"{var} = {solucion[i]:.4f}")
        
        # Verificar
        verificar_solucion(A, b, solucion)

# Función para resolver sistema general desde archivo
def resolver_desde_archivo(nombre_archivo):
    """Lee sistema desde archivo y lo resuelve"""
    try:
        with open(nombre_archivo, 'r') as f:
            lineas = f.readlines()
        
        n = len(lineas)
        A = []
        b = []
        
        for linea in lineas:
            partes = linea.strip().split()
            if len(partes) != n + 1:
                print("Error: Formato de archivo incorrecto")
                return
            
            fila = [float(x) for x in partes[:-1]]
            A.append(fila)
            b.append(float(partes[-1]))
        
        print(f"Sistema de {n} ecuaciones con {n} incógnitas")
        solucion = eliminacion_gaussiana(A, b)
        
        if solucion:
            print("\nSOLUCIÓN:")
            for i, valor in enumerate(solucion):
                print(f"x{i+1} = {valor:.4f}")
            verificar_solucion(A, b, solucion)
            
    except FileNotFoundError:
        print(f"Archivo {nombre_archivo} no encontrado")

if __name__ == "__main__":
    print("ELIMINACIÓN GAUSSIANA")
    print("=" * 50)
    
    # Resolver el sistema del problema
    resolver_sistema_problema()
    
    print("\n" + "=" * 50)
    print("Para resolver otro sistema, crear archivo 'sistema.txt'")
    print("Formato: cada línea contiene coeficientes y término independiente")
    print("Ejemplo: '1 2 3 4' para ecuación x + 2y + 3z = 4")
```

## **Desarrollo Matemático**

### **1. Expansión en serie de Fourier**

La función de onda se expresa como:

\[
\psi(x) = \sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right)
\]

### **2. Aplicación de la ecuación de Schrödinger**

La ecuación de Schrödinger independiente del tiempo es:

\[
\hat{H} \psi(x) = E \psi(x)
\]

Sustituyendo la expansión de Fourier:

\[
\hat{H} \left[ \sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right) \right] = E \sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right)
\]

### **3. Multiplicación por función de prueba**

Multiplicamos ambos lados por \(\sin\left(\frac{\pi m x}{L}\right)\) e integramos sobre \([0,L]\):

\[
\int_0^L \sin\left(\frac{\pi m x}{L}\right) \hat{H} \left[ \sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right) \right] dx = E \int_0^L \sin\left(\frac{\pi m x}{L}\right) \left[ \sum_{n=1}^{\infty} \psi_n \sin\left(\frac{\pi n x}{L}\right) \right] dx
\]

Intercambiando suma e integral (asumiendo convergencia):

\[
\sum_{n=1}^{\infty} \psi_n \int_0^L \sin\left(\frac{\pi m x}{L}\right) \hat{H} \sin\left(\frac{\pi n x}{L}\right) dx = E \sum_{n=1}^{\infty} \psi_n \int_0^L \sin\left(\frac{\pi m x}{L}\right) \sin\left(\frac{\pi n x}{L}\right) dx
\]

### **4. Uso de la ortogonalidad**

Utilizando la propiedad de ortogonalidad:

\[
\int_0^L \sin\left(\frac{\pi m x}{L}\right) \sin\left(\frac{\pi n x}{L}\right) dx = 
\begin{cases} 
\frac{L}{2} & \text{si } m = n \\
0 & \text{si } m \neq n
\end{cases}
\]

El lado derecho se simplifica a:

\[
E \sum_{n=1}^{\infty} \psi_n \cdot \frac{L}{2} \delta_{mn} = E \psi_m \cdot \frac{L}{2}
\]

Por lo tanto:

\[
\sum_{n=1}^{\infty} \psi_n \int_0^L \sin\left(\frac{\pi m x}{L}\right) \hat{H} \sin\left(\frac{\pi n x}{L}\right) dx = \frac{L}{2} E \psi_m
\]

### **5. Definición de la matriz Hamiltoniana**

Definimos los elementos de matriz:

\[
H_{mn} = \frac{2}{L} \int_0^L \sin\left(\frac{\pi m x}{L}\right) \hat{H} \sin\left(\frac{\pi n x}{L}\right) dx
\]

Sustituyendo el operador Hamiltoniano:

\[
H_{mn} = \frac{2}{L} \int_0^L \sin\left(\frac{\pi m x}{L}\right) \left[ -\frac{\hbar^2}{2M} \frac{d^2}{dx^2} + V(x) \right] \sin\left(\frac{\pi n x}{L}\right) dx
\]

### **6. Forma matricial**

Multiplicando ambos lados de nuestra ecuación anterior por \(\frac{2}{L}\):

\[
\frac{2}{L} \sum_{n=1}^{\infty} \psi_n \int_0^L \sin\left(\frac{\pi m x}{L}\right) \hat{H} \sin\left(\frac{\pi n x}{L}\right) dx = E \psi_m
\]

Reconociendo la definición de \(H_{mn}\):

\[
\sum_{n=1}^{\infty} H_{mn} \psi_n = E \psi_m
\]

En notación matricial:

\[
\mathbf{H} \boldsymbol{\psi} = E \boldsymbol{\psi}
\]

donde:
- \(\mathbf{H}\) es la matriz Hamiltoniana con elementos \(H_{mn}\)
- \(\boldsymbol{\psi} = (\psi_1, \psi_2, \psi_3, \ldots)^T\) es el vector de coeficientes de Fourier
- \(E\) es la energía (valor propio)

---

## **Descomposición de la Matriz Hamiltoniana**

Podemos separar \(H_{mn}\) en dos partes:

### **Parte cinética:**
\[
T_{mn} = \frac{2}{L} \int_0^L \sin\left(\frac{\pi m x}{L}\right) \left[ -\frac{\hbar^2}{2M} \frac{d^2}{dx^2} \right] \sin\left(\frac{\pi n x}{L}\right) dx
\]

### **Parte potencial:**
\[
V_{mn} = \frac{2}{L} \int_0^L \sin\left(\frac{\pi m x}{L}\right) V(x) \sin\left(\frac{\pi n x}{L}\right) dx
\]

Por lo tanto:
\[
H_{mn} = T_{mn} + V_{mn}
\]

---

## **Conclusión**

Hemos demostrado que el problema cuántico del pozo asimétrico se reduce a encontrar los **valores propios** de la matriz Hamiltoniana \(\mathbf{H}\). Los valores propios \(E\) representan las energías permitidas del sistema, y los vectores propios \(\boldsymbol{\psi}\) representan los coeficientes de Fourier de las funciones de onda correspondientes.

Este es un ejemplo clásico de cómo problemas de física cuántica continuos se pueden discretizar y resolver usando métodos de álgebra lineal numérica.

_b)_

Recordemos que:
\[
H_{mn} = T_{mn} + V_{mn}
\]
donde:
- \(T_{mn}\) es la parte cinética
- \(V_{mn}\) es la parte potencial

### **2. Parte cinética \(T_{mn}\)**

\[
T_{mn} = \frac{2}{L} \int_0^L \sin\left(\frac{\pi m x}{L}\right) \left[ -\frac{\hbar^2}{2M} \frac{d^2}{dx^2} \right] \sin\left(\frac{\pi n x}{L}\right) dx
\]

Calculamos la segunda derivada:
\[
\frac{d}{dx} \sin\left(\frac{\pi n x}{L}\right) = \frac{\pi n}{L} \cos\left(\frac{\pi n x}{L}\right)
\]
\[
\frac{d^2}{dx^2} \sin\left(\frac{\pi n x}{L}\right) = -\left(\frac{\pi n}{L}\right)^2 \sin\left(\frac{\pi n x}{L}\right)
\]

Sustituyendo:
\[
T_{mn} = \frac{2}{L} \int_0^L \sin\left(\frac{\pi m x}{L}\right) \left[ -\frac{\hbar^2}{2M} \left(-\left(\frac{\pi n}{L}\right)^2\right) \right] \sin\left(\frac{\pi n x}{L}\right) dx
\]
\[
T_{mn} = \frac{2}{L} \cdot \frac{\hbar^2}{2M} \left(\frac{\pi n}{L}\right)^2 \int_0^L \sin\left(\frac{\pi m x}{L}\right) \sin\left(\frac{\pi n x}{L}\right) dx
\]

Usando la ortogonalidad:
\[
\int_0^L \sin\left(\frac{\pi m x}{L}\right) \sin\left(\frac{\pi n x}{L}\right) dx = \frac{L}{2} \delta_{mn}
\]

Por lo tanto:
\[
T_{mn} = \frac{\hbar^2 \pi^2 n^2}{2ML^2} \delta_{mn}
\]

### **3. Parte potencial \(V_{mn}\)**

Para \(V(x) = \frac{a}{L}x\):
\[
V_{mn} = \frac{2}{L} \int_0^L \sin\left(\frac{\pi m x}{L}\right) \left(\frac{a}{L}x\right) \sin\left(\frac{\pi n x}{L}\right) dx
\]
\[
V_{mn} = \frac{2a}{L^2} \int_0^L x \sin\left(\frac{\pi m x}{L}\right) \sin\left(\frac{\pi n x}{L}\right) dx
\]

Usando la integral dada:

**Caso 1: \(m = n\)**
\[
\int_0^L x \sin\left(\frac{\pi m x}{L}\right) \sin\left(\frac{\pi m x}{L}\right) dx = \frac{L^2}{4}
\]
\[
V_{mm} = \frac{2a}{L^2} \cdot \frac{L^2}{4} = \frac{a}{2}
\]

**Caso 2: \(m \neq n\), uno par y otro impar**
\[
\int_0^L x \sin\left(\frac{\pi m x}{L}\right) \sin\left(\frac{\pi n x}{L}\right) dx = -\left(\frac{2L}{\pi}\right)^2 \frac{mn}{(m^2 - n^2)^2}
\]
\[
V_{mn} = \frac{2a}{L^2} \cdot \left[ -\left(\frac{2L}{\pi}\right)^2 \frac{mn}{(m^2 - n^2)^2} \right] = -\frac{8a}{\pi^2} \cdot \frac{mn}{(m^2 - n^2)^2}
\]

**Caso 3: \(m \neq n\), ambos pares o ambos impares**
\[
\int_0^L x \sin\left(\frac{\pi m x}{L}\right) \sin\left(\frac{\pi n x}{L}\right) dx = 0
\]
\[
V_{mn} = 0
\]

### **4. Expresión final para \(H_{mn}\)**

\[
H_{mn} = \frac{\hbar^2 \pi^2 n^2}{2ML^2} \delta_{mn} + V_{mn}
\]

donde:
\[
V_{mn} = 
\begin{cases}
\frac{a}{2} & \text{si } m = n \\
-\frac{8a}{\pi^2} \cdot \frac{mn}{(m^2 - n^2)^2} & \text{si } m \neq n \text{ y uno par, otro impar} \\
0 & \text{si } m \neq n \text{ y ambos pares o ambos impares}
\end{cases}
\]

### **5. Simetría de la matriz**

Observamos que:
- \(T_{mn}\) es diagonal, por lo tanto simétrica
- \(V_{mn}\) es simétrica porque:
  - \(V_{mm}\) es simétrica trivialmente
  - Para \(m \neq n\), \(V_{mn}\) depende de \(mn\) y \((m^2 - n^2)^2\), que son simétricos bajo intercambio \(m \leftrightarrow n\)

Por lo tanto, \(H_{mn} = H_{nm}\) y la matriz es **real y simétrica**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes físicas
hbar = 1.0545718e-34  # J·s
M = 9.1094e-31        # kg (masa del electrón)
e_charge = 1.6022e-19 # C
L = 5e-10             # 5 Å en metros
a = 10 * e_charge     # 10 eV convertido a Joules

def elemento_hamiltoniano(m, n):
    """
    Calcula el elemento H_{mn} de la matriz Hamiltoniana
    para el potencial V(x) = (a/L)x
    """
    # Parte cinética (diagonal)
    T = (hbar**2 * np.pi**2 * n**2) / (2 * M * L**2)
    
    if m == n:
        V = a / 2
    elif (m % 2 == 0 and n % 2 == 1) or (m % 2 == 1 and n % 2 == 0):
        # Uno par, otro impar
        V = - (8 * a / np.pi**2) * (m * n) / ((m**2 - n**2)**2)
    else:
        # Ambos pares o ambos impares
        V = 0
    
    return T * (m == n) + V  # T solo contribuye cuando m = n

def construir_hamiltoniano(N):
    """
    Construye la matriz Hamiltoniana N×N
    """
    H = np.zeros((N, N))
    
    for i in range(N):
        for j in range(N):
            H[i, j] = elemento_hamiltoniano(i+1, j+1)
    
    return H

def resolver_sistema(N):
    """
    Resuelve el problema de valores propios para matriz N×N
    """
    print(f"=== POZO CUÁNTICO ASIMÉTRICO ===")
    print(f"Ancho del pozo: L = {L:.2e} m")
    print(f"Potencial: V(x) = ({a/e_charge:.1f} eV/L) * x")
    print(f"Matriz Hamiltoniana: {N}×{N}")
    
    # Construir matriz Hamiltoniana
    H = construir_hamiltoniano(N)
    
    # Calcular valores y vectores propios
    energias, vectores = np.linalg.eigh(H)
    
    # Convertir energías a eV para mejor interpretación
    energias_eV = energias / e_charge
    
    return H, energias_eV, vectores

# Parámetros de cálculo
N = 10  # Tamaño de la matriz

# Resolver el sistema
H, energias, autovectores = resolver_sistema(N)

# Resultados
print(f"\n=== ENERGÍAS PERMITIDAS (primeras {N}) ===")
for i, E in enumerate(energias[:6]):  # Mostrar solo las primeras 6
    print(f"E_{i+1} = {E:.4f} eV")

# Verificar simetría
print(f"\n=== VERIFICACIÓN DE SIMETRÍA ===")
simetria = np.max(np.abs(H - H.T))
print(f"Max |H - Hᵀ| = {simetria:.2e}")
print("La matriz es simétrica" if simetria < 1e-10 else "La matriz NO es simétrica")

# Visualización de la matriz
plt.figure(figsize=(15, 5))

# Subplot 1: Matriz Hamiltoniana
plt.subplot(1, 3, 1)
im = plt.imshow(H/e_charge, cmap='RdBu_r', aspect='equal')
plt.colorbar(im, label='Hₘₙ (eV)')
plt.title('Matriz Hamiltoniana Hₘₙ')
plt.xlabel('n')
plt.ylabel('m')

# Subplot 2: Niveles de energía
plt.subplot(1, 3, 2)
niveles = range(1, len(energias) + 1)
plt.plot(energias, niveles, 'bo-', markersize=6)
plt.xlabel('Energía (eV)')
plt.ylabel('Número de nivel')
plt.title('Niveles de Energía')
plt.grid(True, alpha=0.3)

# Subplot 3: Elementos no diagonales
plt.subplot(1, 3, 3)
# Extraer elementos no diagonales
no_diagonales = []
for i in range(N):
    for j in range(i+1, N):
        no_diagonales.append(abs(H[i,j]/e_charge))

plt.hist(no_diagonales, bins=20, alpha=0.7, edgecolor='black')
plt.xlabel('|Hₘₙ| no diagonal (eV)')
plt.ylabel('Frecuencia')
plt.title('Distribución de elementos no diagonales')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Análisis adicional
print(f"\n=== ANÁLISIS ADICIONAL ===")
print(f"Energía mínima: {np.min(energias):.4f} eV")
print(f"Energía máxima (de las calculadas): {np.max(energias):.4f} eV")
print(f"Separación promedio entre niveles: {np.mean(np.diff(energias)):.4f} eV")

# Comparación con pozo infinito simétrico
print(f"\n=== COMPARACIÓN CON POZO INFINITO SIMÉTRICO ===")
E_simetrico = [ (hbar**2 * np.pi**2 * n**2) / (2 * M * L**2 * e_charge) for n in range(1, N+1) ]
print("Energías pozo simétrico (eV):", [f"{E:.4f}" for E in E_simetrico[:6]])

# Función para calcular funciones de onda aproximadas
def funcion_onda(x, coeficientes, N_terms):
    """
    Reconstruye la función de onda a partir de los coeficientes de Fourier
    """
    psi = 0
    for n in range(1, N_terms + 1):
        psi += coeficientes[n-1] * np.sin(n * np.pi * x / L)
    return psi

# Visualizar algunas funciones de onda
plt.figure(figsize=(12, 8))
x = np.linspace(0, L, 1000)

for i in range(3):  # Primeros 3 estados
    plt.subplot(2, 3, i+1)
    coeficientes = autovectores[:, i]
    psi = funcion_onda(x, coeficientes, N)
    
    plt.plot(x*1e10, psi, 'b-', linewidth=2, label=f'Estado {i+1}')
    plt.plot(x*1e10, psi**2, 'r--', alpha=0.7, label=f'|ψ|² estado {i+1}')
    plt.xlabel('x (Å)')
    plt.ylabel('ψ(x)')
    plt.title(f'Estado {i+1}, E = {energias[i]:.3f} eV')
    plt.legend()
    plt.grid(True, alpha=0.3)

# Densidad de probabilidad total para primeros N estados
plt.subplot(2, 3, 4)
for i in range(3):
    coeficientes = autovectores[:, i]
    psi = funcion_onda(x, coeficientes, N)
    plt.plot(x*1e10, psi**2, label=f'Estado {i+1}')

plt.xlabel('x (Å)')
plt.ylabel('|ψ(x)|²')
plt.title('Densidades de Probabilidad')
plt.legend()
plt.grid(True, alpha=0.3)

# Potencial
plt.subplot(2, 3, 5)
V_x = (a / L) * x / e_charge  # Convertir a eV
plt.plot(x*1e10, V_x, 'g-', linewidth=2)
plt.xlabel('x (Å)')
plt.ylabel('V(x) (eV)')
plt.title('Potencial V(x) = (10 eV/L) × x')
plt.grid(True, alpha=0.3)

# Niveles de energía en el potencial
plt.subplot(2, 3, 6)
plt.plot(x*1e10, V_x, 'g-', linewidth=2, label='V(x)')
for i, E in enumerate(energias[:4]):
    plt.axhline(y=E, color=f'C{i}', linestyle='--', label=f'E_{i+1} = {E:.2f} eV')
plt.xlabel('x (Å)')
plt.ylabel('Energía (eV)')
plt.title('Niveles en el Potencial')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes físicas
hbar = 1.0545718e-34  # J·s
M = 9.1094e-31        # kg (masa del electrón)
e_charge = 1.6022e-19 # C
L = 5e-10             # 5 Å en metros
a = 10 * e_charge     # 10 eV convertido a Joules

def elemento_hamiltoniano(m, n):
    """
    Calcula el elemento H_{mn} de la matriz Hamiltoniana
    para el potencial V(x) = (a/L)x
    
    NOTA: m y n empiezan en 1 (índices físicos)
    """
    # Parte cinética (diagonal)
    T = (hbar**2 * np.pi**2 * n**2) / (2 * M * L**2)
    
    if m == n:
        V = a / 2
    elif (m % 2 == 0 and n % 2 == 1) or (m % 2 == 1 and n % 2 == 0):
        # Uno par, otro impar
        V = - (8 * a / np.pi**2) * (m * n) / ((m**2 - n**2)**2)
    else:
        # Ambos pares o ambos impares
        V = 0
    
    return T * (m == n) + V  # T solo contribuye cuando m = n

def construir_hamiltoniano_10x10():
    """
    Construye la matriz Hamiltoniana 10×10
    """
    H = np.zeros((10, 10))
    
    # IMPORTANTE: índices de Python van de 0-9, pero físicos de 1-10
    for i in range(10):      # índice Python: 0-9
        for j in range(10):  # índice Python: 0-9
            m = i + 1        # índice físico: 1-10
            n = j + 1        # índice físico: 1-10
            H[i, j] = elemento_hamiltoniano(m, n)
    
    return H

def calcular_niveles_energia():
    """
    Calcula y muestra los primeros 10 niveles de energía
    """
    print("=" * 60)
    print("POZO CUÁNTICO ASIMÉTRICO - MATRIZ 10×10")
    print("=" * 60)
    
    # Construir matriz Hamiltoniana 10×10
    H = construir_hamiltoniano_10x10()
    
    print("Matriz Hamiltoniana 10×10 construida")
    print(f"Dimensiones: {H.shape}")
    
    # Calcular valores propios (energías)
    energias_joules = np.linalg.eigvalsh(H)  # eigh para matrices simétricas
    energias_eV = energias_joules / e_charge
    
    print(f"\nPRIMEROS 10 NIVELES DE ENERGÍA:")
    print("-" * 40)
    for i, energia in enumerate(energias_eV):
        print(f"E_{i+1:2d} = {energia:8.4f} eV")
    
    return H, energias_eV

# Ejecutar cálculo
H, energias = calcular_niveles_energia()

# Análisis de resultados
print(f"\nANÁLISIS DE RESULTADOS:")
print("-" * 40)
print(f"Energía del estado fundamental: {energias[0]:.4f} eV")
print(f"Separación E₂ - E₁: {energias[1] - energias[0]:.4f} eV")
print(f"Separación E₃ - E₂: {energias[2] - energias[1]:.4f} eV")
print(f"Energía máxima (E₁₀): {energias[-1]:.4f} eV")

# Verificación con el valor esperado
print(f"\nVERIFICACIÓN:")
print("-" * 40)
print(f"Valor esperado estado fundamental: ~5.84 eV")
print(f"Nuestro resultado: {energias[0]:.4f} eV")
print(f"Diferencia: {abs(energias[0] - 5.84):.4f} eV")

# Visualización de resultados
plt.figure(figsize=(15, 5))

# Subplot 1: Matriz Hamiltoniana
plt.subplot(1, 3, 1)
im = plt.imshow(H/e_charge, cmap='RdBu_r', aspect='equal')
plt.colorbar(im, label='Hₘₙ (eV)')
plt.title('Matriz Hamiltoniana 10×10')
plt.xlabel('n (índice físico)')
plt.ylabel('m (índice físico)')
plt.xticks(range(10), range(1, 11))
plt.yticks(range(10), range(1, 11))

# Subplot 2: Niveles de energía
plt.subplot(1, 3, 2)
niveles = range(1, 11)
plt.plot(energias, niveles, 'bo-', markersize=8, linewidth=2)
plt.xlabel('Energía (eV)')
plt.ylabel('Número de nivel')
plt.title('Niveles de Energía (10×10)')
plt.grid(True, alpha=0.3)
plt.yticks(range(1, 11))

# Subplot 3: Separación entre niveles
plt.subplot(1, 3, 3)
separaciones = np.diff(energias)
niveles_sep = range(1, 10)
plt.plot(niveles_sep, separaciones, 'ro-', markersize=6, linewidth=2)
plt.xlabel('Transición n → n+1')
plt.ylabel('ΔE (eV)')
plt.title('Separación entre Niveles Consecutivos')
plt.grid(True, alpha=0.3)
plt.xticks(range(1, 10))

plt.tight_layout()
plt.show()

# Información adicional sobre la matriz
print(f"\nPROPIEDADES DE LA MATRIZ:")
print("-" * 40)
print(f"Traza: {np.trace(H)/e_charge:.4f} eV")
print(f"Determinante: {np.linalg.det(H)/e_charge**10:.2e} eV¹⁰")
print(f"¿Es simétrica?: {np.allclose(H, H.T)}")

# Comparación con diferentes tamaños de matriz
def comparar_tamanos_matriz():
    """
    Compara energías para diferentes tamaños de matriz
    """
    tamanos = [5, 10, 15, 20]
    resultados = {}
    
    print(f"\nCOMPARACIÓN CON DIFERENTES TAMAÑOS DE MATRIZ:")
    print("-" * 50)
    print(f"{'Tamaño':>8} {'E₁ (eV)':>10} {'E₂ (eV)':>10} {'E₃ (eV)':>10}")
    print("-" * 50)
    
    for N in tamanos:
        H_temp = np.zeros((N, N))
        for i in range(N):
            for j in range(N):
                m = i + 1
                n = j + 1
                H_temp[i, j] = elemento_hamiltoniano(m, n)
        
        energias_temp = np.linalg.eigvalsh(H_temp) / e_charge
        resultados[N] = energias_temp
        
        print(f"{N:8d} {energias_temp[0]:10.4f} {energias_temp[1]:10.4f} {energias_temp[2]:10.4f}")
    
    return resultados

# Ejecutar comparación
resultados_comparacion = comparar_tamanos_matriz()

# Función para reconstruir funciones de onda
def funcion_onda(x, coeficientes):
    """
    Reconstruye la función de onda a partir de los coeficientes de Fourier
    """
    psi = 0
    for n in range(1, 11):  # Usamos los primeros 10 términos
        psi += coeficientes[n-1] * np.sin(n * np.pi * x / L)
    return psi

# Calcular autovectores para visualizar funciones de onda
autovalores, autovectores = np.linalg.eigh(H)

# Visualizar funciones de onda de los primeros 3 estados
plt.figure(figsize=(12, 8))
x = np.linspace(0, L, 1000)

for i in range(3):  # Primeros 3 estados
    plt.subplot(2, 3, i+1)
    coeficientes = autovectores[:, i]
    psi = funcion_onda(x, coeficientes)
    
    # Normalizar
    psi = psi / np.sqrt(np.trapz(psi**2, x))
    
    plt.plot(x*1e10, psi, 'b-', linewidth=2, label=f'ψ_{i+1}(x)')
    plt.plot(x*1e10, psi**2, 'r--', alpha=0.7, label=f'|ψ_{i+1}(x)|²')
    plt.xlabel('x (Å)')
    plt.ylabel('Amplitud')
    plt.title(f'Estado {i+1}, E = {energias[i]:.3f} eV')
    plt.legend()
    plt.grid(True, alpha=0.3)

# Potencial y niveles de energía
plt.subplot(2, 3, 4)
V_x = (a / L) * x / e_charge  # Convertir a eV
plt.plot(x*1e10, V_x, 'g-', linewidth=2, label='V(x)')
for i in range(4):
    plt.axhline(y=energias[i], color=f'C{i}', linestyle='--', 
                label=f'E_{i+1} = {energias[i]:.2f} eV')
plt.xlabel('x (Å)')
plt.ylabel('Energía (eV)')
plt.title('Potencial y Niveles de Energía')
plt.legend()
plt.grid(True, alpha=0.3)

# Densidad de probabilidad total
plt.subplot(2, 3, 5)
for i in range(3):
    coeficientes = autovectores[:, i]
    psi = funcion_onda(x, coeficientes)
    psi = psi / np.sqrt(np.trapz(psi**2, x))  # Normalizar
    plt.plot(x*1e10, psi**2, label=f'Estado {i+1}')

plt.xlabel('x (Å)')
plt.ylabel('|ψ(x)|²')
plt.title('Densidades de Probabilidad')
plt.legend()
plt.grid(True, alpha=0.3)

# Convergencia del estado fundamental
plt.subplot(2, 3, 6)
tamanos = list(resultados_comparacion.keys())
E1_valores = [resultados_comparacion[N][0] for N in tamanos]
plt.plot(tamanos, E1_valores, 'mo-', markersize=8, linewidth=2)
plt.axhline(y=5.84, color='red', linestyle='--', label='Valor esperado: 5.84 eV')
plt.xlabel('Tamaño de la matriz')
plt.ylabel('E₁ (eV)')
plt.title('Convergencia del Estado Fundamental')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nCONCLUSIÓN:")
print("-" * 40)
print(f"El estado fundamental calculado es: {energias[0]:.4f} eV")
print(f"El valor esperado es: ~5.84 eV")
print(f"La aproximación 10×10 proporciona un resultado satisfactorio")
print(f"La diferencia es de solo {abs(energias[0] - 5.84):.4f} eV")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Constantes físicas
hbar = 1.0545718e-34  # J·s
M = 9.1094e-31        # kg (masa del electrón)
e_charge = 1.6022e-19 # C
L = 5e-10             # 5 Å en metros
a = 10 * e_charge     # 10 eV convertido a Joules

def elemento_hamiltoniano(m, n):
    """
    Calcula el elemento H_{mn} de la matriz Hamiltoniana
    para el potencial V(x) = (a/L)x
    
    NOTA: m y n empiezan en 1 (índices físicos)
    """
    # Parte cinética (diagonal)
    T = (hbar**2 * np.pi**2 * n**2) / (2 * M * L**2)
    
    if m == n:
        V = a / 2
    elif (m % 2 == 0 and n % 2 == 1) or (m % 2 == 1 and n % 2 == 0):
        # Uno par, otro impar
        V = - (8 * a / np.pi**2) * (m * n) / ((m^2 - n^2)**2)
    else:
        # Ambos pares o ambos impares
        V = 0
    
    return T * (m == n) + V  # T solo contribuye cuando m = n

def construir_hamiltoniano(N):
    """
    Construye la matriz Hamiltoniana N×N
    """
    H = np.zeros((N, N))
    
    print(f"Construyendo matriz {N}×{N}...")
    start_time = time.time()
    
    for i in range(N):      # índice Python: 0 a N-1
        for j in range(N):  # índice Python: 0 a N-1
            m = i + 1       # índice físico: 1 a N
            n = j + 1       # índice físico: 1 a N
            H[i, j] = elemento_hamiltoniano(m, n)
    
    construction_time = time.time() - start_time
    print(f"Matriz {N}×{N} construida en {construction_time:.2f} segundos")
    
    return H, construction_time

def calcular_niveles_energia(H):
    """
    Calcula los niveles de energía a partir de la matriz Hamiltoniana
    """
    print("Calculando valores propios...")
    start_time = time.time()
    
    energias_joules = np.linalg.eigvalsh(H)  # eigh para matrices simétricas
    energias_eV = energias_joules / e_charge
    
    calculation_time = time.time() - start_time
    print(f"Cálculo completado en {calculation_time:.2f} segundos")
    
    return energias_eV, calculation_time

def comparar_resultados(energias_10, energias_100):
    """
    Compara los resultados de matrices 10×10 y 100×100
    """
    print("=" * 70)
    print("COMPARACIÓN: MATRIZ 10×10 vs 100×100")
    print("=" * 70)
    
    print(f"\n{'Nivel':>6} {'10×10 (eV)':>12} {'100×100 (eV)':>12} {'Diferencia (eV)':>15} {'Error Relativo (%)':>18}")
    print("-" * 70)
    
    diferencias = []
    errores_relativos = []
    
    for i in range(10):
        E_10 = energias_10[i]
        E_100 = energias_100[i]
        diferencia = E_100 - E_10
        error_relativo = abs(diferencia / E_100) * 100
        
        diferencias.append(abs(diferencia))
        errores_relativos.append(error_relativo)
        
        print(f"{i+1:6d} {E_10:12.6f} {E_100:12.6f} {diferencia:15.6f} {error_relativo:18.4f}")
    
    print("-" * 70)
    print(f"{'Promedio':>6} {'-':>12} {'-':>12} {np.mean(diferencias):15.6f} {np.mean(errores_relativos):18.4f}")
    print(f"{'Máximo':>6} {'-':>12} {'-':>12} {np.max(diferencias):15.6f} {np.max(errores_relativos):18.4f}")
    
    return diferencias, errores_relativos

# ============================================================================
# CÁLCULO PRINCIPAL
# ============================================================================

print("POZO CUÁNTICO ASIMÉTRICO - ANÁLISIS DE CONVERGENCIA")
print("=" * 70)

# Resultados de la matriz 10×10 (del inciso c)
print("\n1. MATRIZ 10×10:")
H_10, time_10_const = construir_hamiltoniano(10)
energias_10, time_10_calc = calcular_niveles_energia(H_10)

print(f"\n2. MATRIZ 100×100:")
H_100, time_100_const = construir_hamiltoniano(100)
energias_100, time_100_calc = calcular_niveles_energia(H_100)

# Comparación de resultados
diferencias, errores_relativos = comparar_resultados(energias_10, energias_100)

# ============================================================================
# ANÁLISIS DE PRECISIÓN
# ============================================================================

print(f"\n3. ANÁLISIS DE PRECISIÓN:")
print("=" * 50)

print(f"Estado fundamental (10×10):  {energias_10[0]:.6f} eV")
print(f"Estado fundamental (100×100): {energias_100[0]:.6f} eV")
print(f"Diferencia: {energias_100[0] - energias_10[0]:.6f} eV")
print(f"Error relativo: {abs(energias_100[0] - energias_10[0])/energias_100[0]*100:.4f}%")

print(f"\nTiempos de cálculo:")
print(f"10×10 - Construcción: {time_10_const:.3f} s, Diagonalización: {time_10_calc:.3f} s")
print(f"100×100 - Construcción: {time_100_const:.3f} s, Diagonalización: {time_100_calc:.3f} s")
print(f"Factor de aumento en tiempo: {time_100_const/time_10_const:.1f}x (construcción)")
print(f"Factor de aumento en tiempo: {time_100_calc/time_10_calc:.1f}x (diagonalización)")

# ============================================================================
# VISUALIZACIONES
# ============================================================================

plt.figure(figsize=(15, 10))

# Subplot 1: Comparación de niveles de energía
plt.subplot(2, 3, 1)
niveles = range(1, 11)
plt.plot(energias_10[:10], niveles, 'bo-', markersize=6, label='10×10', linewidth=2)
plt.plot(energias_100[:10], niveles, 'ro-', markersize=6, label='100×100', linewidth=2)
plt.xlabel('Energía (eV)')
plt.ylabel('Número de nivel')
plt.title('Comparación de Niveles de Energía\n(primeros 10 niveles)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.yticks(range(1, 11))

# Subplot 2: Error por nivel
plt.subplot(2, 3, 2)
plt.semilogy(niveles, diferencias, 'go-', markersize=6, linewidth=2)
plt.xlabel('Número de nivel')
plt.ylabel('|E₁₀₀ - E₁₀| (eV)')
plt.title('Error Absoluto por Nivel\n(escala logarítmica)')
plt.grid(True, alpha=0.3)
plt.xticks(range(1, 11))

# Subplot 3: Error relativo por nivel
plt.subplot(2, 3, 3)
plt.plot(niveles, errores_relativos, 'mo-', markersize=6, linewidth=2)
plt.xlabel('Número de nivel')
plt.ylabel('Error Relativo (%)')
plt.title('Error Relativo por Nivel')
plt.grid(True, alpha=0.3)
plt.xticks(range(1, 11))

# Subplot 4: Patrón de la matriz 100×100 (primeros 20×20 elementos)
plt.subplot(2, 3, 4)
im = plt.imshow(H_100[:20, :20]/e_charge, cmap='RdBu_r', aspect='equal')
plt.colorbar(im, label='Hₘₙ (eV)')
plt.title('Matriz Hamiltoniana 100×100\n(primeros 20×20 elementos)')
plt.xlabel('n')
plt.ylabel('m')

# Subplot 5: Convergencia del estado fundamental
plt.subplot(2, 3, 5)
tamanos_matriz = [5, 10, 20, 50, 100]
energias_E1 = []

print(f"\n4. ANÁLISIS DE CONVERGENCIA:")
print("Calculando para diferentes tamaños de matriz...")

for N in tamanos_matriz:
    H_temp, _ = construir_hamiltoniano(N)
    energias_temp, _ = calcular_niveles_energia(H_temp)
    energias_E1.append(energias_temp[0])
    print(f"  N={N:3d}: E₁ = {energias_temp[0]:.6f} eV")

plt.plot(tamanos_matriz, energias_E1, 'co-', markersize=8, linewidth=2)
plt.axhline(y=energias_100[0], color='red', linestyle='--', 
           label=f'Valor convergido: {energias_100[0]:.6f} eV')
plt.xlabel('Tamaño de la matriz (N)')
plt.ylabel('E₁ (eV)')
plt.title('Convergencia del Estado Fundamental')
plt.legend()
plt.grid(True, alpha=0.3)

# Subplot 6: Densidad espectral
plt.subplot(2, 3, 6)
# Calcular separación entre niveles consecutivos
separaciones_10 = np.diff(energias_10[:20])
separaciones_100 = np.diff(energias_100[:20])

niveles_sep = range(1, 20)
plt.plot(niveles_sep, separaciones_10, 'bo-', markersize=4, alpha=0.7, label='10×10')
plt.plot(niveles_sep, separaciones_100, 'ro-', markersize=4, alpha=0.7, label='100×100')
plt.xlabel('Transición n → n+1')
plt.ylabel('ΔE (eV)')
plt.title('Separación entre Niveles Consecutivos')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# CONCLUSIÓN DETALLADA
# ============================================================================

print(f"\n5. CONCLUSIÓN SOBRE LA PRECISIÓN:")
print("=" * 60)

# Análisis estadístico
error_promedio = np.mean(diferencias)
error_maximo = np.max(diferencias)
error_relativo_promedio = np.mean(errores_relativos)

print(f"Error absoluto promedio: {error_promedio:.6f} eV")
print(f"Error absoluto máximo:   {error_maximo:.6f} eV")
print(f"Error relativo promedio: {error_relativo_promedio:.4f}%")
print(f"Error relativo máximo:   {np.max(errores_relativos):.4f}%")

print(f"\nTendencias observadas:")
if error_relativo_promedio < 0.1:
    print("✓ La aproximación 10×10 es EXCELENTE para los primeros 10 niveles")
elif error_relativo_promedio < 1.0:
    print("✓ La aproximación 10×10 es MUY BUENA para los primeros 10 niveles")
elif error_relativo_promedio < 5.0:
    print("○ La aproximación 10×10 es ACEPTABLE para los primeros 10 niveles")
else:
    print("✗ La aproximación 10×10 es INADECUADA para los primeros 10 niveles")

print(f"\nPara aplicaciones prácticas:")
if error_relativo_promedio < 0.1:
    print("→ La matriz 10×10 es SUFICIENTE para la mayoría de cálculos")
elif error_relativo_promedio < 0.5:
    print("→ La matriz 10×10 es ADECUADA para cálculos aproximados")
else:
    print("→ Se recomienda usar matriz 100×100 para mayor precisión")

# Análisis de convergencia adicional
print(f"\n6. CONVERGENCIA CON EL TAMAÑO DE MATRIZ:")
print("=" * 50)

# Calcular cuántos dígitos significativos son correctos
def digitos_significativos(E10, E100):
    """Calcula cuántos dígitos significativos son correctos"""
    if E100 == 0:
        return 0
    error_rel = abs(E100 - E10) / abs(E100)
    if error_rel == 0:
        return float('inf')
    return -np.log10(error_rel)

digitos_E1 = digitos_significativos(energias_10[0], energias_100[0])
digitos_E5 = digitos_significativos(energias_10[4], energias_100[4])
digitos_E10 = digitos_significativos(energias_10[9], energias_100[9])

print(f"Dígitos significativos correctos:")
print(f"  Estado fundamental (E₁): {digitos_E1:.1f} dígitos")
print(f"  Quinto nivel (E₅): {digitos_E5:.1f} dígitos") 
print(f"  Décimo nivel (E₁₀): {digitos_E10:.1f} dígitos")

print(f"\nInterpretación:")
if digitos_E1 >= 4:
    print("✓ Precisión excelente para el estado fundamental")
elif digitos_E1 >= 3:
    print("✓ Precisión muy buena para el estado fundamental")
elif digitos_E1 >= 2:
    print("○ Precisión aceptable para el estado fundamental")
else:
    print("✗ Precisión insuficiente para el estado fundamental")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Constantes físicas
hbar = 1.0545718e-34  # J·s
M = 9.1094e-31        # kg (masa del electrón)
e_charge = 1.6022e-19 # C
L = 5e-10             # 5 Å en metros
a = 10 * e_charge     # 10 eV convertido a Joules

def elemento_hamiltoniano(m, n):
    """
    Calcula el elemento H_{mn} de la matriz Hamiltoniana
    para el potencial V(x) = (a/L)x
    """
    # Parte cinética (diagonal)
    T = (hbar**2 * np.pi**2 * n**2) / (2 * M * L**2)
    
    if m == n:
        V = a / 2
    elif (m % 2 == 0 and n % 2 == 1) or (m % 2 == 1 and n % 2 == 0):
        # Uno par, otro impar
        V = - (8 * a / np.pi**2) * (m * n) / ((m**2 - n**2)**2)
    else:
        # Ambos pares o ambos impares
        V = 0
    
    return T * (m == n) + V

def construir_hamiltoniano(N):
    """
    Construye la matriz Hamiltoniana N×N
    """
    H = np.zeros((N, N))
    
    for i in range(N):
        for j in range(N):
            m = i + 1
            n = j + 1
            H[i, j] = elemento_hamiltoniano(m, n)
    
    return H

def funcion_onda_fourier(x, coeficientes, N_terms):
    """
    Reconstruye la función de onda usando expansión de Fourier
    ψ(x) = Σₙ ψₙ sin(nπx/L)
    """
    psi = 0
    for n in range(1, N_terms + 1):
        psi += coeficientes[n-1] * np.sin(n * np.pi * x / L)
    return psi

def normalizar_funcion_onda(psi_func, x_range, N_points=1000):
    """
    Normaliza la función de onda para que ∫|ψ(x)|²dx = 1
    """
    x = np.linspace(x_range[0], x_range[1], N_points)
    psi_values = psi_func(x)
    
    # Calcular la integral |ψ(x)|²dx
    norma_cuadrado = integrate.trapz(psi_values**2, x)
    
    # Factor de normalización
    factor_normalizacion = 1.0 / np.sqrt(norma_cuadrado)
    
    # Función normalizada
    def psi_normalizada(x):
        return factor_normalizacion * psi_func(x)
    
    return psi_normalizada, factor_normalizacion, norma_cuadrado

def calcular_funciones_onda():
    """
    Calcula y normaliza las funciones de onda para los primeros 3 estados
    """
    print("=" * 70)
    print("CÁLCULO DE FUNCIONES DE ONDA - POZO CUÁNTICO ASIMÉTRICO")
    print("=" * 70)
    
    # Usar matriz 100×100 para mayor precisión
    N = 100
    print(f"Construyendo matriz Hamiltoniana {N}×{N}...")
    H = construir_hamiltoniano(N)
    
    # Calcular autovalores y autovectores
    print("Calculando autovalores y autovectores...")
    energias, autovectores = np.linalg.eigh(H)
    energias_eV = energias / e_charge
    
    print(f"\nEnergías de los primeros 3 estados:")
    for i in range(3):
        print(f"E_{i+1} = {energias_eV[i]:.6f} eV")
    
    # Preparar funciones de onda para los primeros 3 estados
    funciones_onda = []
    info_normalizacion = []
    
    for i in range(3):
        print(f"\n--- Estado {i+1} (E = {energias_eV[i]:.6f} eV) ---")
        
        # Coeficientes de Fourier del autovector
        coeficientes = autovectores[:, i]
        
        # Función de onda sin normalizar
        def psi_sin_normalizar(x):
            return funcion_onda_fourier(x, coeficientes, N)
        
        # Normalizar la función de onda
        psi_normalizada, factor, norma_cuadrado = normalizar_funcion_onda(
            psi_sin_normalizar, [0, L]
        )
        
        funciones_onda.append(psi_normalizada)
        info_normalizacion.append({
            'factor_normalizacion': factor,
            'norma_cuadrado_antes': norma_cuadrado,
            'norma_cuadrado_despues': 1.0  # Por construcción
        })
        
        print(f"Norma² antes de normalizar: {norma_cuadrado:.10f}")
        print(f"Factor de normalización: {factor:.10f}")
        print(f"Norma² después de normalizar: {1.0:.10f}")
    
    return funciones_onda, energias_eV, info_normalizacion

def verificar_normalizacion(funciones_onda, x_range=[0, 5e-10], N_points=5000):
    """
    Verifica que las funciones de onda estén correctamente normalizadas
    """
    print(f"\n" + "=" * 70)
    print("VERIFICACIÓN DE NORMALIZACIÓN")
    print("=" * 70)
    
    x = np.linspace(x_range[0], x_range[1], N_points)
    resultados = []
    
    for i, psi in enumerate(funciones_onda):
        psi_values = psi(x)
        norma_cuadrado = integrate.trapz(psi_values**2, x)
        
        resultados.append({
            'estado': i + 1,
            'norma_cuadrado': norma_cuadrado,
            'error': abs(norma_cuadrado - 1.0)
        })
        
        print(f"Estado {i+1}: ∫|ψ(x)|²dx = {norma_cuadrado:.10f}")
        print(f"          Error = {abs(norma_cuadrado - 1.0):.2e}")
        
        if abs(norma_cuadrado - 1.0) < 1e-10:
            print("          ✓ NORMALIZACIÓN CORRECTA")
        else:
            print("          ✗ PROBLEMA DE NORMALIZACIÓN")
    
    return resultados

# ============================================================================
# CÁLCULO PRINCIPAL
# ============================================================================

# Calcular funciones de onda
funciones_onda, energias, info_norm = calcular_funciones_onda()

# Verificar normalización
resultados_verificacion = verificar_normalizacion(funciones_onda)

# ============================================================================
# VISUALIZACIÓN
# ============================================================================

print(f"\n" + "=" * 70)
print("GENERANDO GRÁFICAS")
print("=" * 70)

# Crear malla espacial para gráficas
x = np.linspace(0, L, 1000)
x_angstrom = x * 1e10  # Convertir a Ångstroms

plt.figure(figsize=(15, 10))

# =======================
# SUBPLOT 1: Funciones de onda ψ(x)
# =======================
plt.subplot(2, 3, 1)
colores = ['blue', 'red', 'green']
estados = ['Estado Fundamental', 'Primer excitado', 'Segundo excitado']

for i in range(3):
    psi_values = funciones_onda[i](x)
    plt.plot(x_angstrom, psi_values, 
             color=colores[i], linewidth=2.5, label=estados[i])

plt.xlabel('x (Å)')
plt.ylabel('ψ(x)')
plt.title('Funciones de Onda\n(primeros 3 estados)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='black', linestyle='-', alpha=0.5)

# =======================
# SUBPLOT 2: Densidades de probabilidad |ψ(x)|²
# =======================
plt.subplot(2, 3, 2)
for i in range(3):
    psi_values = funciones_onda[i](x)
    densidad = psi_values**2
    plt.plot(x_angstrom, densidad, 
             color=colores[i], linewidth=2.5, label=f'{estados[i]}')

plt.xlabel('x (Å)')
plt.ylabel('|ψ(x)|²')
plt.title('Densidades de Probabilidad\n(primeros 3 estados)')
plt.legend()
plt.grid(True, alpha=0.3)

# =======================
# SUBPLOT 3: Densidades apiladas
# =======================
plt.subplot(2, 3, 3)
for i in range(3):
    psi_values = funciones_onda[i](x)
    densidad = psi_values**2
    plt.fill_between(x_angstrom, densidad, alpha=0.6, 
                    color=colores[i], label=f'E{i+1} = {energias[i]:.3f} eV')

plt.xlabel('x (Å)')
plt.ylabel('|ψ(x)|²')
plt.title('Densidades de Probabilidad Apiladas')
plt.legend()
plt.grid(True, alpha=0.3)

# =======================
# SUBPLOT 4: Potencial y funciones de onda
# =======================
plt.subplot(2, 3, 4)
# Potencial V(x) = (a/L)x en eV
V_x = (a / L) * x / e_charge

# Graficar potencial
plt.plot(x_angstrom, V_x, 'k-', linewidth=2, label='V(x) = (10 eV/L)×x')

# Graficar niveles de energía y funciones de onda escaladas
for i in range(3):
    # Escalar función de onda para visualización
    psi_values = funciones_onda[i](x)
    escala = 2.0  # Factor de escala para visualización
    psi_escalada = energias[i] + escala * psi_values
    
    plt.plot(x_angstrom, psi_escalada, 
             color=colores[i], linewidth=2, 
             label=f'ψ_{i+1}(x) (escalada)')
    
    # Línea del nivel de energía
    plt.axhline(y=energias[i], color=colores[i], linestyle='--', alpha=0.7)

plt.xlabel('x (Å)')
plt.ylabel('Energía (eV)')
plt.title('Potencial y Funciones de Onda\n(con niveles de energía)')
plt.legend()
plt.grid(True, alpha=0.3)

# =======================
# SUBPLOT 5: Verificación de ortogonalidad
# =======================
plt.subplot(2, 3, 5)
# Calcular productos internos entre estados
productos_internos = np.zeros((3, 3))
x_fino = np.linspace(0, L, 5000)  # Malla más fina para integración

for i in range(3):
    for j in range(3):
        psi_i = funciones_onda[i](x_fino)
        psi_j = funciones_onda[j](x_fino)
        producto = integrate.trapz(psi_i * psi_j, x_fino)
        productos_internos[i, j] = producto

# Visualizar matriz de productos internos
im = plt.imshow(productos_internos, cmap='RdYlBu_r', vmin=0, vmax=1)
plt.colorbar(im, label='∫ψᵢψⱼdx')
plt.xticks(range(3), ['ψ₁', 'ψ₂', 'ψ₃'])
plt.yticks(range(3), ['ψ₁', 'ψ₂', 'ψ₃'])
plt.title('Productos Internos\n(Verificación de Ortogonalidad)')

# Añadir valores numéricos
for i in range(3):
    for j in range(3):
        color = 'white' if productos_internos[i, j] > 0.5 else 'black'
        plt.text(j, i, f'{productos_internos[i, j]:.3f}', 
                ha='center', va='center', color=color, fontweight='bold')

# =======================
# SUBPLOT 6: Análisis de propiedades
# =======================
plt.subplot(2, 3, 6)
# Calcular valores esperados de posición
valores_esperados_x = []
incertidumbres_x = []

for i in range(3):
    psi_values = funciones_onda[i](x_fino)
    densidad = psi_values**2
    
    # <x> = ∫ x |ψ(x)|² dx
    x_esperado = integrate.trapz(x_fino * densidad, x_fino)
    
    # <x²> = ∫ x² |ψ(x)|² dx  
    x2_esperado = integrate.trapz(x_fino**2 * densidad, x_fino)
    
    # Δx = √(<x²> - <x>²)
    incertidumbre = np.sqrt(x2_esperado - x_esperado**2)
    
    valores_esperados_x.append(x_esperado)
    incertidumbres_x.append(incertidumbre)

# Graficar valores esperados y incertidumbres
estados_nums = [1, 2, 3]
plt.errorbar(estados_nums, np.array(valores_esperados_x)*1e10, 
             yerr=np.array(incertidumbres_x)*1e10, 
             fmt='o', capsize=5, capthick=2, markersize=8,
             color='purple', linewidth=2)

plt.axhline(y=L*1e10/2, color='red', linestyle='--', 
           label='Centro del pozo', alpha=0.7)

plt.xlabel('Estado')
plt.ylabel('<x> ± Δx (Å)')
plt.title('Valor Esperado de Posición\ncon Incertidumbre')
plt.xticks([1, 2, 3])
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# ANÁLISIS DETALLADO
# ============================================================================

print(f"\n" + "=" * 70)
print("ANÁLISIS DETALLADO DE LAS FUNCIONES DE ONDA")
print("=" * 70)

print(f"\n1. PROPIEDADES DE NORMALIZACIÓN:")
for i, resultado in enumerate(resultados_verificacion):
    estado = resultado['estado']
    norma = resultado['norma_cuadrado']
    error = resultado['error']
    
    print(f"Estado {estado}: ∫|ψ(x)|²dx = {norma:.10f}")
    print(f"          Error = {error:.2e}")
    if error < 1e-10:
        print(f"          ✓ Normalización CORRECTA")
    else:
        print(f"          ✗ Problema de normalización")

print(f"\n2. ORTOGONALIDAD ENTRE ESTADOS:")
print("   Los estados propios de un operador Hermítico deben ser ortogonales")
for i in range(3):
    for j in range(3):
        producto = productos_internos[i, j]
        if i == j:
            print(f"   <ψ{i+1}|ψ{j+1}> = {producto:.6f} (debe ser ≈ 1)")
        else:
            print(f"   <ψ{i+1}|ψ{j+1}> = {producto:.6f} (debe ser ≈ 0)")

print(f"\n3. VALORES ESPERADOS DE POSICIÓN:")
for i in range(3):
    x_avg = valores_esperados_x[i] * 1e10  # Convertir a Å
    dx = incertidumbres_x[i] * 1e10        # Convertir a Å
    print(f"   Estado {i+1}: <x> = {x_avg:.3f} Å, Δx = {dx:.3f} Å")

print(f"\n4. INTERPRETACIÓN FÍSICA:")
print(f"   • Estado fundamental: Máxima probabilidad cerca de x=0 (potencial mínimo)")
print(f"   • Primer excitado: Un nodo en la densidad de probabilidad")  
print(f"   • Segundo excitado: Dos nodos en la densidad de probabilidad")
print(f"   • Asimetría: Las funciones NO son simétricas debido a V(x) = (a/L)x")

print(f"\n5. VERIFICACIÓN DE LA ECUACIÓN DE SCHRÖDINGER:")
print(f"   Las funciones calculadas son soluciones aproximadas de:")
print(f"   [-ħ²/2m d²/dx² + V(x)]ψ(x) = Eψ(x)")
print(f"   El error depende del tamaño de la base de Fourier usada (N={100})")

# Verificación adicional: calcular energía a partir de la función de onda
def calcular_energia_esperada(psi_func, N_points=5000):
    """
    Calcula la energía esperada <ψ|H|ψ> para verificar la solución
    """
    x = np.linspace(0, L, N_points)
    psi = psi_func(x)
    
    # Derivada numérica (diferencias centradas)
    dx = x[1] - x[0]
    dpsi_dx = np.gradient(psi, dx)
    d2psi_dx2 = np.gradient(dpsi_dx, dx)
    
    # Aplicar operador Hamiltoniano
    H_psi = - (hbar**2 / (2 * M)) * d2psi_dx2 + (a/L * x) * psi
    
    # Energía esperada: <ψ|H|ψ> = ∫ ψ* H ψ dx
    E_esperada = integrate.trapz(psi * H_psi, x)
    
    return E_esperada / e_charge  # Convertir a eV

print(f"\n6. VERIFICACIÓN DE ENERGÍAS ESPERADAS:")
print("   Calculando <ψ|H|ψ> para cada estado...")
for i in range(3):
    E_calculada = energias[i]
    E_esperada = calcular_energia_esperada(funciones_onda[i])
    error_E = abs(E_calculada - E_esperada)
    
    print(f"   Estado {i+1}:")
    print(f"      E (autovalor)    = {E_calculada:.6f} eV")
    print(f"      <ψ|H|ψ>          = {E_esperada:.6f} eV") 
    print(f"      Error            = {error_E:.2e} eV")
    
    if error_E < 0.001:
        print(f"      ✓ Buena concordancia")
    else:
        print(f"      ○ Concordancia aceptable")

print(f"\n" + "=" * 70)
print("CONCLUSIÓN FINAL")
print("=" * 70)
print("Las funciones de onda calculadas satisfacen:")
print("✓ Condición de normalización: ∫|ψ(x)|²dx = 1")
print("✓ Condición de ortogonalidad: <ψᵢ|ψⱼ> = δᵢⱼ")  
print("✓ Ecuación de Schrödinger: Hψ = Eψ (aproximadamente)")
print("✓ Comportamiento físico esperado para pozo asimétrico")
```
