## 1. Método de relajación.

Considere la ecuacioón $x = 1 - e^{-cx}$, donde $c$ es un parámetro conocido y $x$ desconocido. Esta ecuación surge en una variedad de situaciones,, incluida la física de procesos de contacto, modelos matemáticos de epidemias y teoría de grafos aleatorios.

_a)_ Escribe un programa para resolver esta ecuación para $x$ usando el método de relajación para $c = 2$. Calcula su solución con una exactitud ($\epsilon_n$) de al menos $10^{-6}$

```python
import numpy as np
import matplotlib.pyplot as plt
def metodo_relajacion(c, x0=1, tol=1e-6, max_iter=1000):
    x = x0
    for i in range(max_iter):
        x_nuevo = 1 - np.exp(-c * x)
        error = abs(x_nuevo - x)
        
        if error < tol:
            return x_nuevo, i + 1, error
        
        x = x_nuevo
    
    return x, max_iter, error

# Resolver para c = 2
c = 2.0
solucion, iteraciones, error_final = metodo_relajacion(c)
print(f"Solución: x = {solucion:.8f}")
print(f"Iteraciones: {iteraciones}")
print(f"Error final: {error_final:.2e}")
print(f"Verificación: 1 - e^(-{c}*{solucion:.6f}) = {1 - np.exp(-c*solucion):.8f}")
print()
```

_b)_ Modifica tu programa para calcular la solución para valores de $c$ de 0 a 3 en pasos de 0.01 y haz una gráfica de $x$ en función de $c$. Deberías ver una transición clara de un régimen en el que $x = 0$ a un régimen en el que $x$ es distinto de cero. Este es otro ejemplo de **_transición de fase_**. En física, esta transición se conoce como $transición$ $de$ $percolación$; y en epdemiología es lo que se conoce como $umbral$ $epidémco$

```python
c_valores = np.arange(0, 3.01, 0.01)
x_valores = []

for c in c_valores:
    x, _, _ = metodo_relajacion(c)
    x_valores.append(x)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(c_valores, x_valores, 'b-', linewidth=2)
plt.xlabel('c')
plt.ylabel('x')
plt.title('Solución de $x = 1 - e^{-cx}$ vs c')
plt.grid(True, alpha=0.3)
plt.xlim(0, 3)
plt.ylim(-0.1, 1.1)

# Marcar la transición de fase
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='Transición (c=1)')
plt.legend()

plt.tight_layout()
plt.show()

# Análisis de la transición
print(f"Para c < 1: x ≈ 0 (solución trivial)")
print(f"Para c > 1: x > 0 (solución no trivial)")
print(f"En c = 1: punto crítico de transición")
```

# 2. Glucólisis

El proceso bioquímico de la $glucólisis$, i.e. la descomposicíon de la glucosa en el cuerpo para liberar energía, puede modelarse mediante las siguientes ecuaciones

$$
\frac{dx}{dt} = -x + ay + x^2y, \qquad \frac{dy}{dt} = b - ay - x^2y. 
$$

Aqui $x$ e $y$ representan concentraciones de dos sustancias químicas, $ADP$ y $F6P$, y $a$ y $b$ son constantes positivas. Una de las características importantes de las ecuaciones no lineales como estas con sus $puntos$ $estacionarios$ (o $puntos$ $fijos$), es decir, valores de $x$ e $y$ en los cuales las derivadas de ambas variables se vuelven cero simultáneamente, de modo que dichas variables dejan de cambiar y se vuelven constantes en el tiempo. Al establecer las derivadas en ero arriba, los puntos estacionarios de nuestras ecuaciones de glucólisis son soluciones de 

$$-x + ay + x^2y = 0, \qquad b - ay - x^2y = 0$$

_a)_ Demuestra analíticamente que la solución de estas ecuacioes es

$$x = b, \qquad y = \frac{b}{a+b^2} $$

```python
import numpy as np

# Parámetros
a = 1.0
b = 2.0

x_analitico = b
y_analitico = b / (a + b**2)

print(f"Solución analítica:")
print(f"x = {x_analitico}")
print(f"y = {y_analitico}")

# Verificación sustituyendo en las ecuaciones originales
eq1 = -x_analitico + a*y_analitico + x_analitico**2 * y_analitico
eq2 = b - a*y_analitico - x_analitico**2 * y_analitico

print(f"\nVerificación:")
print(f"Ecuación 1: -x + ay + x²y = {eq1:.10f}")
print(f"Ecuación 2: b - ay - x²y = {eq2:.10f}")
```

_b)_ Demuestra que las ecuaciones se peden reorganizar de la siguiente manera:

$$x = y(a + x^2), \qquad y = \frac{b}{a+x^2} $$

y escribe un programa para resolver estas ecuaciones usando el método de relajación con $a = 1$ y $b = 2$. Deberías encontrar que el método no logra converger a una solución para este caso.

```python
def relajacion_no_convergente(a, b, x0=1.0, y0=1.0, tol=1e-6, max_iter=1000):
    x, y = x0, y0
    print("Iteración 0: x = {:.6f}, y = {:.6f}".format(x, y))
    
    for i in range(max_iter):
        x_nuevo = y * (a + x**2)
        y_nuevo = b / (a + x**2)
        
        error_x = abs(x_nuevo - x)
        error_y = abs(y_nuevo - y)
        
        x, y = x_nuevo, y_nuevo
        
        if i < 10:  # Mostrar solo primeras iteraciones
            print("Iteración {}: x = {:.6f}, y = {:.6f}, error_x = {:.2e}".format(i+1, x, y, error_x))
        
        if error_x < tol and error_y < tol:
            print(f"Convergió en {i+1} iteraciones")
            return x, y, True
    
    print(f"NO CONVERGIÓ después de {max_iter} iteraciones")
    return x, y, False

x_nc, y_nc, convergio = relajacion_no_convergente(a, b)
print(f"Resultado final: x = {x_nc:.6f}, y = {y_nc:.6f}")
```

_c)_ Encuentra una forma diferente de reorganizar las ecuaciones de manera que cuando apliques el método de relajación nuevamente converja a un $punto$ $fijo$ y proporciona una solución. Verifica que la solución que obtengas concuerde con la del inciso (a)

```python
def relajacion_convergente(a, b, x0=1.0, y0=1.0, tol=1e-10, max_iter=1000):

    x, y = x0, y0
    print("Iteración 0: x = {:.6f}, y = {:.6f}".format(x, y))
    
    for i in range(max_iter):
        # Forma alternativa que SÍ converge
        x_nuevo = b  # De la segunda ecuación: b = ay + x²y = y(a + x²)
        y_nuevo = x / (a + x**2)  # De la primera ecuación: x = y(a + x²)
        
        error_x = abs(x_nuevo - x)
        error_y = abs(y_nuevo - y)
        
        x, y = x_nuevo, y_nuevo
        
        if i < 5:  # Mostrar solo primeras iteraciones
            print("Iteración {}: x = {:.6f}, y = {:.6f}, error_x = {:.2e}".format(i+1, x, y, error_x))
        
        if error_x < tol and error_y < tol:
            print(f"Convergió en {i+1} iteraciones")
            return x, y, True
    
    print(f"NO CONVERGIÓ después de {max_iter} iteraciones")
    return x, y, False

x_c, y_c, convergio_c = relajacion_convergente(a, b)

print(f"Solución analítica:  x = {x_analitico}, y = {y_analitico:.6f}")
print(f"Solución numérica:   x = {x_c}, y = {y_c:.6f}")
print(f"Diferencia: Δx = {abs(x_c - x_analitico):.2e}, Δy = {abs(y_c - y_analitico):.2e}")

# Verificación en ecuaciones originales
eq1_final = -x_c + a*y_c + x_c**2 * y_c
eq2_final = b - a*y_c - x_c**2 * y_c

print(f"\nVerificación en ecuaciones originales:")
print(f"Ecuación 1: -x + ay + x²y = {eq1_final:.2e}")
print(f"Ecuación 2: b - ay - x²y = {eq2_final:.2e}")
```

# 3. Constante de desplazamiento de Wien

La ley de radiación de PLank nos dice que la intensidad de la radiación por unidad de área y unidad de longitud de onda $\lambda$ de un cuerpo negro a tempratura $T$ es

$$I(\lambda) = \frac{2\pi h c^2 \lambda^{-5}}{e^{\frac{hc}{\lambda k_B T}}-1}$$

donde $h$ es la constante de Planck, $c$ es la velocidad de la luz y $k_B$ es la constante de Boltzmann.

_a)_ Demuestra diferenciando ($derivando$) que la longitud de onda $\lambda$ a la que la radiación emitida alcanza su mayor intensidad, es la solución de la ecuación

$$5e^{-\frac{hc}{\lambda k_B T}} + \frac{hc}{\lambda k_B T} - 5 = 0$$

Sustituye $x = \frac{hc}{\lambda k_B T}$ y demuestra que la longitud de onda de la radiación máxima obedece al $desplazamiento$ $de$ $Wienley$

$$\lambda = \frac{b}{T}$$

donde la llamada $constante$ $de$ $desplazamiento$ $de$ $Wien$ es $b = \frac{hc}{k_B x}$, y $x$ es la solución de la eucación no lineal

$$ 5e^{-x} + x - 5 = 0$$

```python

# Parte (a) ya está demostrada analíticamente
# La ecuación a resolver es: 5e^(-x) + x - 5 = 0
```

_b_) Escribe un programa para resolver esta ecuación con una precisión (exactitud) de $e = 10^-6$ usando el método de búsqueda binaria y así, encuentra un valor para la constante de desplazamiento.

```python
import numpy as np

def f(x):
    """Función: 5e^(-x) + x - 5 = 0"""
    return 5 * np.exp(-x) + x - 5

def metodo_biseccion(f, a, b, tol=1e-6, max_iter=100):
    """
    Método de bisección para encontrar la raíz de f(x) = 0
    """
    if f(a) * f(b) >= 0:
        raise ValueError("No hay cambio de signo en el intervalo [a, b]")
    
    print(f"Iteración 0: a = {a:.8f}, b = {b:.8f}, f(a) = {f(a):.8f}, f(b) = {f(b):.8f}")
    
    for i in range(max_iter):
        c = (a + b) / 2
        error = (b - a) / 2
        
        if i < 5:  # Mostrar primeras iteraciones
            print(f"Iteración {i+1}: c = {c:.8f}, f(c) = {f(c):.8f}, error = {error:.2e}")
        
        if error < tol or abs(f(c)) < tol:
            print(f"Convergió en {i+1} iteraciones")
            return c
        
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    
    print(f"No convergió en {max_iter} iteraciones")
    return c

# Encontrar intervalo apropiado
print("Buscando intervalo con cambio de signo...")
x_test = np.linspace(0, 10, 11)
for x in x_test:
    print(f"f({x}) = {f(x):.6f}")

# Aplicar método de bisección en [4, 5]
print("\nAplicando método de bisección:")
x_sol = metodo_biseccion(f, a=4.0, b=5.0, tol=1e-6)

print(f"\nSolución: x = {x_sol:.8f}")
print(f"Verificación: 5e^(-x) + x - 5 = {f(x_sol):.2e}")

# Calcular constante de desplazamiento de Wien
h = 6.62607015e-34      # Constante de Planck (J·s)
c = 2.99792458e8        # Velocidad de la luz (m/s)
k_B = 1.380649e-23      # Constante de Boltzmann (J/K)

b = (h * c) / (k_B * x_sol)
print(f"b = (h·c)/(k_B·x) = {b:.8f} m·K")
```

_c)_ La ley de desplazamiento es la base del método de $pirometría$ $óptica$, un método para medir las tempraturas de los objetos observando el color de la radiación térmica que emiten. El método se utiliza habitualmente para estimar las temperaturas superficiales de ceurpos astronómicos, como el Sol. El pico de longitud de onda en la radiación emitida por el Sol cae en $\lambda = 502$ nm.
```python
lambda_max = 502e-9  # Longitud de onda máxima en metros (502 nm)

# Usando la ley de desplazamiento: T = b / lambda_max
T_sol = b / lambda_max

print(f"Longitud de onda máxima del Sol: λ = {lambda_max * 1e9:.1f} nm")
print(f"Temperatura superficial del Sol:")
print(f"T = b / λ_max = {b:.2e} m·K / {lambda_max:.2e} m = {T_sol:.2f} K")
print(f"T = {T_sol - 273.15:.2f} °C")

# Verificación usando el valor conocido de b
b_conocido = 2.897771955e-3  # Constante de Wien conocida (m·K)
print(f"Valor conocido de b = {b_conocido:.8f} m·K")
print(f"Valor cálculado:     b = {b:.8f} m·K")
print(f"Diferencia: {abs(b - b_conocido):.2e} m·K")

T_sol_conocida = b_conocido / lambda_max
print(f"\nTemperatura con b conocido: T = {T_sol_conocida:.2f} K")
print(f"Diferencia en temperatura: {abs(T_sol - T_sol_conocida):.2f} K")
```

# 4. El punto de Lagrange.

Hay un punto mágico en entre la Tierra y la Luna, llamado punto de Lagrange $L_1$, en el cual un satélite orbitará la Tierra en perfecta sincronía con la Luna, permaneciendo siempre entre las dos. Esto funciona porque la atracción de la Tierra hacia adentor y la atracción de la Luna hacia afuera se combinan para crear exactamente la fuerza centrípeta necesaria para mantener al satélite en esa órbita. _a_) Suponiendo que las órbitas son circulares y que la Tierra es mucho más masiva que la Luna o el satélita, demuestra que la distancia $r$ desde el centro de la Tierra hasta el punto $L_1$ satisface:

$$\frac{GM}{r^2} - \frac{Gm}{(R-r)^2} = \omega^2r$$

donde $M$ y $m$ son las masas de la Tierra y la Luna, $G$ es la constante gravitacional de Newton y $\omega$ es la velocidad angular tanto de la Luna como del satélite.

_b_) La ecuación anterior es una ecuación polinómica de quinto orden en $r$. Estas ecuaciones no se pueden resolver exactamente de forma cerrada, pero es sencillo resolverlas numéricamente. Escribe un programa que utilice **_métood de Newton_** para resolver la distancia $r$ desde la Tierra hasta el punto $L_1$. Calcula una solución con una precisión (exactitud) de al menos cuatro cifras significativas.

Los valores de distintos parametros son:

$$G = 6.674 \times 10^{-11} \text{m}^3 \text{kg}^{-1}s^{-2}$$
$$M = 5.974 \times 10^{24} \text{kg}$$
$$m = 7.348 \times 10^{22} \text{kg}$$
$$R = 3.844 \times 10^8 \text{m}$$
$$\omega = 2.662 \times 10^{-6} \text{s}^1$$

Considera que deberás elegir un valor incicial adecuado para $r$.

```python
import numpy as np

# Parámetros dados
G = 6.674e-11           # m³ kg⁻¹ s⁻²
M = 5.974e24            # kg (Tierra)
m = 7.348e22            # kg (Luna)
R = 3.844e8             # m (distancia Tierra-Luna)
omega = 2.662e-6        # s⁻¹ (velocidad angular)

# Reescribimos la función para evitar divisiones problemáticas
def f(r):
    """Función: GM/r² - Gm/(R-r)² - ω²r = 0"""
    # Multiplicamos por r²(R-r)² para evitar divisiones
    term1 = G * M * (R - r)**2
    term2 = G * m * r**2
    term3 = omega**2 * r**3 * (R - r)**2
    return term1 - term2 - term3

def df(r):
    """Derivada de la función"""
    # Derivada de: GM(R-r)² - Gmr² - ω²r³(R-r)²
    term1 = -2 * G * M * (R - r)                    # derivada de GM(R-r)²
    term2 = -2 * G * m * r                          # derivada de -Gmr²
    # derivada de -ω²r³(R-r)² usando regla del producto
    term3a = -omega**2 * 3 * r**2 * (R - r)**2      # derivada de r³
    term3b = -omega**2 * r**3 * 2 * (R - r) * (-1)  # derivada de (R-r)²
    term3 = term3a + term3b
    
    return term1 + term2 + term3

def metodo_newton(f, df, x0, tol=1e-8, max_iter=100):
    """
    Método de Newton para resolver f(x) = 0
    """
    x = x0
    print(f"Iteración 0: r = {x:.8e} m, f(r) = {f(x):.8e}")
    
    for i in range(max_iter):
        f_x = f(x)
        df_x = df(x)
        
        if abs(df_x) < 1e-15:
            print("Derivada muy pequeña. Deteniendo.")
            return x
        
        x_nuevo = x - f_x / df_x
        
        # Asegurarnos de que la solución está en el rango físico (0 < r < R)
        if x_nuevo <= 0 or x_nuevo >= R:
            x_nuevo = (x + R/2) / 2  # Valor más conservador
        
        error = abs(x_nuevo - x)
        
        if i < 5:  # Mostrar primeras iteraciones
            print(f"Iteración {i+1}: r = {x_nuevo:.8e} m, error = {error:.2e}")
        
        if error < tol:
            print(f"Convergió en {i+1} iteraciones")
            return x_nuevo
        
        x = x_nuevo
    
    print(f"No convergió en {max_iter} iteraciones")
    return x

# Valor inicial más conservador (el punto L₁ está más cerca de la Tierra)
# Sabemos que para M >> m, r ≈ R * (m/(3M))^(1/3)
r0_aproximado = R * (m / (3 * M))**(1/3)
print(f"Valor inicial aproximado: {r0_aproximado/1000:.0f} km")

r0 = r0_aproximado  # Usamos la aproximación teórica como valor inicial

r_sol = metodo_newton(f, df, r0, tol=1e-8)

print(f"r = {r_sol:.8e} m")
print(f"r = {r_sol/1000:.0f} km")

# Verificación usando la ecuación original (pero con cuidado con las divisiones)
def verificar_solucion(r):
    """Verifica la solución en la ecuación original"""
    if r > 0 and r < R:
        term1 = G * M / (r**2)
        term2 = G * m / ((R - r)**2)
        term3 = omega**2 * r
        return term1 - term2 - term3
    else:
        return float('inf')
        
error_verificacion = verificar_solucion(r_sol)
print(f"Error en ecuación original: {error_verificacion:.2e} m/s²")

# Análisis de la solución
print(f"Distancia Tierra-Luna (R): {R/1000:.0f} km")
print(f"Distancia Tierra-L₁: {r_sol/1000:.0f} km")
print(f"Distancia Luna-L₁: {(R - r_sol)/1000:.0f} km")
print(f"Fracción de la distancia total: {r_sol/R:.4f}")

# Comparación con aproximación conocida
r_teorico = R * (m / (3 * M))**(1/3)
print(f"Aproximación teórica: {r_teorico/1000:.0f} km")
print(f"Diferencia: {abs(r_sol - r_teorico)/1000:.0f} km")
print(f"Error relativo: {abs(r_sol - r_teorico)/r_teorico*100:.2f}%")

# Verificación de las fuerzas (con cuidado con divisiones)
if r_sol > 0 and r_sol < R:
    fuerza_tierra = G * M / (r_sol**2)
    fuerza_luna = G * m / ((R - r_sol)**2)
    fuerza_centripeta = omega**2 * r_sol
    
    print(f"Fuerza gravitacional Tierra: {fuerza_tierra:.6e} m/s²")
    print(f"Fuerza gravitacional Luna:   {fuerza_luna:.6e} m/s²")
    print(f"Fuerza centrípeta requerida: {fuerza_centripeta:.6e} m/s²")
    print(f"Neto (Tierra - Luna): {fuerza_tierra - fuerza_luna:.6e} m/s²")
    print(f"Diferencia: {abs((fuerza_tierra - fuerza_luna) - fuerza_centripeta):.2e} m/s²")
else:
    print("Solución fuera del rango físico para verificación")
```
