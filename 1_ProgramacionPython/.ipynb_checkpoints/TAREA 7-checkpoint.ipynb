## 1. Método de relajación.

Considere la ecuacioón $x = 1 - e^{-cx}$, donde $c$ es un parámetro conocido y $x$ desconocido. Esta ecuación surge en una variedad de situaciones,, incluida la física de procesos de contacto, modelos matemáticos de epidemias y teoría de grafos aleatorios.

_a)_ Escribe un programa para resolver esta ecuación para $x$ usando el método de relajación para $c = 2$. Calcula su solución con una exactitud ($\epsilon_n$) de al menos $10^{-6}$

```python
import numpy as np

# Definimos la ecuación de punto fijo
def f(x, c):
    return 1 - np.exp(-c * x)

# Parámetros
c = 2.0
x = 1.0       # Valor inicial
tol = 1e-6    # Tolerancia pedida
error = 1.0   # Inicializamos error mayor a la tolerancia
n_iter = 0    # Contador de iteraciones

while error > tol:
    x_nuevo = f(x, c)
    error = np.abs(x_nuevo - x)
    x = x_nuevo
    n_iter += 1

print(f"Solución para c=2: x = {x:.6f}")
print(f"Número de iteraciones: {n_iter}")
```

_b)_ Modifica tu programa para calcular la solución para valores de $c$ de 0 a 3 en pasos de 0.01 y haz una gráfica de $x$ en función de $c$. Deberías ver una transición clara de un régimen en el que $x = 0$ a un régimen en el que $x$ es distinto de cero. Este es otro ejemplo de **_transición de fase_**. En física, esta transición se conoce como $transición$ $de$ $percolación$; y en epdemiología es lo que se conoce como $umbral$ $epidémco$

```python
import matplotlib.pyplot as plt

# Crear arreglo de c de 0 a 3
c_valores = np.arange(0, 3.01, 0.01)
x_soluciones = []

# Bucle para calcular x para cada c
for c_val in c_valores:
    x = 1.0 # Reiniciamos x inicial para cada c
    
    for i in range(1000): 
        x_nuevo = 1 - np.exp(-c_val * x)
        if np.abs(x_nuevo - x) < 1e-6:
            x = x_nuevo
            break
        x = x_nuevo
    x_soluciones.append(x)

plt.figure(figsize=(8, 5))
plt.plot(c_valores, x_soluciones, 'b-')
plt.xlabel("Parámetro c")
plt.ylabel("Solución x")
plt.title("Transición de fase en el método de relajación")
plt.grid(True)
plt.show()
```

# 2. Glucólisis

El proceso bioquímico de la $glucólisis$, i.e. la descomposicíon de la glucosa en el cuerpo para liberar energía, puede modelarse mediante las siguientes ecuaciones

$$
\frac{dx}{dt} = -x + ay + x^2y, \qquad \frac{dy}{dt} = b - ay - x^2y. 
$$

Aqui $x$ e $y$ representan concentraciones de dos sustancias químicas, $ADP$ y $F6P$, y $a$ y $b$ son constantes positivas. Una de las características importantes de las ecuaciones no lineales como estas con sus $puntos$ $estacionarios$ (o $puntos$ $fijos$), es decir, valores de $x$ e $y$ en los cuales las derivadas de ambas variables se vuelven cero simultáneamente, de modo que dichas variables dejan de cambiar y se vuelven constantes en el tiempo. Al establecer las derivadas en ero arriba, los puntos estacionarios de nuestras ecuaciones de glucólisis son soluciones de 

$$-x + ay + x^2y = 0, \qquad b - ay - x^2y = 0$$

_a)_ Demuestra analíticamente que la solución de estas ecuacioes es

$$x = b, \qquad y = \frac{b}{a+b^2} $$

Para encontrar los puntos estacionarios, igualamos las derivadas a cero:
$$-x + ay + x^2y = 0 \quad (1)$$
$$b - ay - x^2y = 0 \quad (2)$$

De la ecuación (2), despejamos el término común:$b = ay + x^2y$

Sustituimos esto en la ecuación (1):$-x + (ay + x^2y) = 0 \Rightarrow -x + b = 0 \Rightarrow \mathbf{x = b}$

Ahora sustituimos $x=b$ en la ecuación (2) para hallar $y$:

$b - ay - b^2y = 0$

$b = y(a + b^2)$

$\mathbf{y = \frac{b}{a + b^2}}$

_b)_ Demuestra que las ecuaciones se peden reorganizar de la siguiente manera:

$$x = y(a + x^2), \qquad y = \frac{b}{a+x^2} $$

y escribe un programa para resolver estas ecuaciones usando el método de relajación con $a = 1$ y $b = 2$. Deberías encontrar que el método no logra converger a una solución para este caso.

```python
# Parámetros
a = 1.0
b = 2.0

# Valores iniciales
x = 1.0
y = 1.0

print("Probando ecuaciones inestables:")
for i in range(20): # Pocas iteraciones para ver que explota
    # Ecuaciones dadas en el inciso b
    x_nuevo = y * (a + x**2)
    y_nuevo = b / (a + x**2)
    
    x, y = x_nuevo, y_nuevo
    print(f"Iter {i+1}: x = {x:.2f}, y = {y:.2f}")

print("Conclusión: Los valores divergen rápidamente.")
```

_c)_ Encuentra una forma diferente de reorganizar las ecuaciones de manera que cuando apliques el método de relajación nuevamente converja a un $punto$ $fijo$ y proporciona una solución. Verifica que la solución que obtengas concuerde con la del inciso (a)

```python
# Parámetros
a = 1.0
b = 2.0
x = b            # Usamos la solución teórica como "guess" cercano o x=1
y = b/(a+b**2)   

x = 1.0
y = 1.0
tol = 1e-6

print("\nProbando ecuaciones estables:")
for i in range(100):
    # Reorganización:
    # Como (1)+(2) implica x=b, usamos eso como paso implícito o despeje fuerte.
    # Pero para ser puramente iterativo sin saber la respuesta de antemano:
    
    # Despeje alternativo:
    y_nuevo = b / (a + x**2)
    x_nuevo = (x + a*y_nuevo + x**2*y_nuevo) / 2 # Un promedio ponderado a veces ayuda, pero mantengamos lo simple.
    
    # Si usamos la lógica de tu tarea anterior exitosa:
    # x_nuevo = b (directo de sumar ambas)
    # y_nuevo = x / (a + x**2) (de la ec 1 reordenada)
    
    x_nuevo = b 
    y_nuevo = x / (a + x**2)

    if np.abs(x_nuevo - x) < tol and np.abs(y_nuevo - y) < tol:
        x, y = x_nuevo, y_nuevo
        print(f"Convergió en iteración {i+1}")
        break
    x, y = x_nuevo, y_nuevo

print(f"Solución encontrada: x = {x:.6f}, y = {y:.6f}")
```

# 3. Constante de desplazamiento de Wien

La ley de radiación de PLank nos dice que la intensidad de la radiación por unidad de área y unidad de longitud de onda $\lambda$ de un cuerpo negro a tempratura $T$ es

$$I(\lambda) = \frac{2\pi h c^2 \lambda^{-5}}{e^{\frac{hc}{\lambda k_B T}}-1}$$

donde $h$ es la constante de Planck, $c$ es la velocidad de la luz y $k_B$ es la constante de Boltzmann.

_a)_ Demuestra diferenciando ($derivando$) que la longitud de onda $\lambda$ a la que la radiación emitida alcanza su mayor intensidad, es la solución de la ecuación

$$5e^{-\frac{hc}{\lambda k_B T}} + \frac{hc}{\lambda k_B T} - 5 = 0$$

Sustituye $x = \frac{hc}{\lambda k_B T}$ y demuestra que la longitud de onda de la radiación máxima obedece al $desplazamiento$ $de$ $Wienley$

$$\lambda = \frac{b}{T}$$

donde la llamada $constante$ $de$ $desplazamiento$ $de$ $Wien$ es $b = \frac{hc}{k_B x}$, y $x$ es la solución de la eucación no lineal

$$ 5e^{-x} + x - 5 = 0$$

Partimos de $I(\lambda) = \frac{2\pi h c^2 \lambda^{-5}}{e^{\frac{hc}{\lambda k_B T}}-1}$.

Hacemos el cambio de variable $x = \frac{hc}{\lambda k_B T}$, lo que implica $\lambda = \frac{hc}{x k_B T}$.

Derivamos $I(\lambda)$ con respecto a $\lambda$ e igualamos a 0 para encontrar el máximo.

Al aplicar la regla de la cadena y simplificar (factorizando términos exponenciales y potencias), se obtiene la ecuación trascendental:
$5e^{-x} + x - 5 = 0$.

De la definición de $x$, despejamos $\lambda$:

$$\lambda = \frac{hc}{k_B T x} = \frac{1}{T} \left( \frac{hc}{k_B x} \right) = \frac{b}{T}$$

_b_) Escribe un programa para resolver esta ecuación con una precisión (exactitud) de $e = 10^-6$ usando el método de búsqueda binaria y así, encuentra un valor para la constante de desplazamiento.

```python
# Constantes físicas
h = 6.626e-34
c_luz = 3.0e8
kB = 1.38e-23

# Función a resolver
def f_wien(x):
    return 5 * np.exp(-x) + x - 5

# Método de Bisección
x1 = 1.0  # Límite inferior
x2 = 10.0 # Límite superior
tol = 1e-6
 
if f_wien(x1) * f_wien(x2) > 0:
    print("El intervalo no es adecuado")
else:
    while (x2 - x1) / 2.0 > tol:
        xm = (x1 + x2) / 2.0
        if f_wien(xm) == 0:
            break
        if f_wien(x1) * f_wien(xm) < 0:
            x2 = xm
        else:
            x1 = xm
            
    x_raiz = (x1 + x2) / 2.0
    print(f"Raíz encontrada x = {x_raiz:.6f}")

    # Calcular constante b
    b_const = (h * c_luz) / (kB * x_raiz)
    print(f"Constante de desplazamiento b = {b_const:.6e} m K")
```

_c)_ La ley de desplazamiento es la base del método de $pirometría$ $óptica$, un método para medir las tempraturas de los objetos observando el color de la radiación térmica que emiten. El método se utiliza habitualmente para estimar las temperaturas superficiales de ceurpos astronómicos, como el Sol. El pico de longitud de onda en la radiación emitida por el Sol cae en $\lambda = 502$ nm. A partir de las ecuaciones anteriores y el valor de la constante de desplazamiento, estima la temperatura de la superficie del Sol.

```python
lambda_sol = 502e-9 # 502 nm a metros

T_sol = b_const / lambda_sol
print(f"Temperatura estimada del Sol: {T_sol:.2f} K")
```

# 4. El punto de Lagrange.

Hay un punto mágico en entre la Tierra y la Luna, llamado punto de Lagrange $L_1$, en el cual un satélite orbitará la Tierra en perfecta sincronía con la Luna, permaneciendo siempre entre las dos. Esto funciona porque la atracción de la Tierra hacia adentor y la atracción de la Luna hacia afuera se combinan para crear exactamente la fuerza centrípeta necesaria para mantener al satélite en esa órbita. Está es la configuración:

_a_) Suponiendo que las órbitas son circulares y que la Tierra es mucho más masiva que la Luna o el satélita, demuestra que la distancia $r$ desde el centro de la Tierra hasta el punto $L_1$ satisface:

$$\frac{GM}{r^2} - \frac{Gm}{(R-r)^2} = \omega^2r$$

donde $M$ y $m$ son las masas de la Tierra y la Luna, $G$ es la constante gravitacional de Newton y $\omega$ es la velocidad angular tanto de la Luna como del satélite.

En el punto $L_1$, la fuerza de atracción gravitacional de la Tierra debe igualar la suma de la fuerza de atracción de la Luna y la fuerza centrípeta necesaria para mantener el satélite en órbita con velocidad angular $\omega$.

$F_{Tierra} = F_{Luna} + F_{Centripeta}$$\frac{GMm_{sat}}{r^2} = \frac{Gmm_{sat}}{(R-r)^2} + m_{sat}\omega^2 r$

Dividiendo todo por $m_{sat}$ (masa del satélite) y reordenando:
$$\frac{GM}{r^2} - \frac{Gm}{(R-r)^2} = \omega^2 r$$

_b_) La ecuación anterior es una ecuación polinómica de quinto orden en $r$. Estas ecuaciones no se pueden resolver exactamente de forma cerrada, pero es sencillo resolverlas numéricamente. Escribe un programa que utilice **_métood de Newton_** para resolver la distancia $r$ desde la Tierra hasta el punto $L_1$. Calcula una solución con una precisión (exactitud) de al menos cuatro cifras significativas.

Los valores de distintos parametros son:

$$G = 6.674 \times 10^{-11} \text{m}^3 \text{kg}^{-1}s^{-2}$$
$$M = 5.974 \times 10^{24} \text{kg}$$
$$m = 7.348 \times 10^{22} \text{kg}$$
$$R = 3.844 \times 10^8 \text{m}$$
$$\omega = 2.662 \times 10^{-6} \text{s}^1$$

Considera que deberás elegir un valor incicial adecuado para $r$.

```python
# Constantes
G = 6.674e-11
M = 5.974e24
m = 7.348e22
R = 3.844e8
omega = 2.662e-6

def f_lagrange(r):
    return (G * M) / r**2 - (G * m) / (R - r)**2 - omega**2 * r

def df_lagrange(r):
    # Derivada de r^-2 es -2r^-3
    # Derivada de -(R-r)^-2 es -(-2)(R-r)^-3 * (-1) = -2(R-r)^-3
    return -(2 * G * M) / r**3 - (2 * G * m) / (R - r)**3 - omega**2

# Método de Newton
r = R / 2  # Valor inicial
tol = 1e-4 # Precisión de 4 cifras significativas

for i in range(20):
    f_val = f_lagrange(r)
    df_val = df_lagrange(r)
    
    r_nuevo = r - f_val / df_val
    
    if np.abs(r_nuevo - r) < 1e-1:
        r = r_nuevo
        print(f"Convergió en {i+1} iteraciones")
        break
    r = r_nuevo

print(f"Distancia r al punto L1: {r:.4e} m")
print(f"En km: {r/1000:.2f} km")
```
