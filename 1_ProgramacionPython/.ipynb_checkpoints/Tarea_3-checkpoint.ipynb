## 1. Gráfica de datos experimentales

El archivo $\texttt{manchasolares.txt}$ (adjunto), contiene el número observado de manchas solares en el Sol en cada mes desde enero de 1749. El archivo contiene dos columnas de números, la primera es el mes y la segunda el número de manchas solares

_a)_ Escribe un programa que lea los datos y haga una gráfica de las manchas solares en funcion del tiempo.

_b)_ Modifica tu programa para mostrar solo los primeros 1000 datos (experimentales)en la gráfica.

_c)_ Modifica nuevamente tu programa para calcular y graficar la _media (promedio) móvil_ de los datos, definida por:

$$
Y_k = \frac{1}{2r+1} \sum_{m=-r}^{r} y_{k+m},
$$

donde $r = 5$ (en este caso) y $y_k$ son los números de manchas solares. El programa debe graficar tanto los datos originales como la _media móvil_ en el mismo gráfico, sólo sobre los primeros 1000 datos.

```python
import numpy as np
import matplotlib.pyplot as plt

# a) Leer los datos y hacer gráfica completa
def leer_datos(archivo):
    """Lee el archivo de datos de manchas solares"""
    datos = np.loadtxt(archivo)
    tiempo = datos[:, 0]  # primera columna: tiempo/meses
    manchas = datos[:, 1]  # segunda columna: número de manchas
    return tiempo, manchas

def graficar_datos_completos(tiempo, manchas):
    """Grafica todos los datos experimentales"""
    plt.figure(figsize=(12, 6))
    plt.plot(tiempo, manchas, 'b-', alpha=0.7, linewidth=0.8, label='Datos originales')
    plt.xlabel('Tiempo (meses desde enero 1749)')
    plt.ylabel('Número de manchas solares')
    plt.title('Manchas solares desde 1749')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

# b) Mostrar solo los primeros 1000 datos
def graficar_primeros_1000(tiempo, manchas):
    """Grafica solo los primeros 1000 datos"""
    tiempo_1000 = tiempo[:1000]
    manchas_1000 = manchas[:1000]
    
    plt.figure(figsize=(12, 6))
    plt.plot(tiempo_1000, manchas_1000, 'b-', alpha=0.7, linewidth=0.8, label='Primeros 1000 datos')
    plt.xlabel('Tiempo (meses desde enero 1749)')
    plt.ylabel('Número de manchas solares')
    plt.title('Primeros 1000 datos de manchas solares')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    return tiempo_1000, manchas_1000

# c) Calcular y graficar media móvil
def media_movil(datos, r=5):
    """Calcula la media móvil con radio r"""
    n = len(datos)
    media_mov = np.zeros(n)
    
    for k in range(n):
        # Definir los índices para la suma
        inicio = max(0, k - r)
        fin = min(n, k + r + 1)
        ventana = datos[inicio:fin]
        media_mov[k] = np.mean(ventana)
    
    return media_mov

def graficar_media_movil(tiempo, manchas, r=5):
    """Grafica datos originales y media móvil"""
    # Calcular media móvil
    media_mov = media_movil(manchas, r)
    
    plt.figure(figsize=(12, 6))
    
    # Graficar datos originales
    plt.plot(tiempo, manchas, 'b-', alpha=0.5, linewidth=0.8, label='Datos originales')
    
    # Graficar media móvil
    plt.plot(tiempo, media_mov, 'r-', linewidth=2, label=f'Media móvil (r={r})')
    
    plt.xlabel('Tiempo (meses desde enero 1749)')
    plt.ylabel('Número de manchas solares')
    plt.title('Manchas solares y media móvil (primeros 1000 datos)')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Programa principal
def main():
    archivo = 'manchasolares.txt'
    
    try:
        # Leer datos
        tiempo, manchas = leer_datos(archivo)
        print(f"Datos leídos: {len(tiempo)} puntos")
        
        # a) Gráfica de todos los datos
        print("Generando gráfica de todos los datos...")
        graficar_datos_completos(tiempo, manchas)
        
        # b) Gráfica de primeros 1000 datos
        print("Generando gráfica de primeros 1000 datos...")
        tiempo_1000, manchas_1000 = graficar_primeros_1000(tiempo, manchas)
        
        # c) Gráfica con media móvil
        print("Generando gráfica con media móvil...")
        graficar_media_movil(tiempo_1000, manchas_1000, r=5)
        
    except FileNotFoundError:
        print(f"Error: No se pudo encontrar el archivo '{archivo}'")
        print("Asegúrate de que el archivo esté en el mismo directorio")
    except Exception as e:
        print(f"Error: {e}")

# Versión alternativa más compacta
def programa_completo():
    """Versión compacta del programa completo"""
    try:
        # Leer datos
        datos = np.loadtxt('manchasolares.txt')
        tiempo = datos[:, 0]
        manchas = datos[:, 1]
        
        # Primeros 1000 datos
        t_1000 = tiempo[:1000]
        m_1000 = manchas[:1000]
        
        # Calcular media móvil
        r = 5
        media_mov = np.zeros(1000)
        for k in range(1000):
            inicio = max(0, k - r)
            fin = min(1000, k + r + 1)
            media_mov[k] = np.mean(m_1000[inicio:fin])
        
        # Graficar
        plt.figure(figsize=(12, 8))
        
        plt.subplot(2, 1, 1)
        plt.plot(tiempo, manchas, 'b-', alpha=0.7, linewidth=0.5)
        plt.xlabel('Tiempo (meses)')
        plt.ylabel('Manchas solares')
        plt.title('Todos los datos (1749-presente)')
        plt.grid(True, alpha=0.3)
        
        plt.subplot(2, 1, 2)
        plt.plot(t_1000, m_1000, 'b-', alpha=0.5, linewidth=0.8, label='Datos originales')
        plt.plot(t_1000, media_mov, 'r-', linewidth=2, label=f'Media móvil (r={r})')
        plt.xlabel('Tiempo (meses)')
        plt.ylabel('Manchas solares')
        plt.title('Primeros 1000 datos con media móvil')
        plt.grid(True, alpha=0.3)
        plt.legend()
        
        plt.tight_layout()
        plt.show()
        
    except FileNotFoundError:
        print("Error: Archivo 'manchasolares.txt' no encontrado")

# Ejecutar el programa
if __name__ == "__main__":
    # Ejecutar la versión completa
    main()
    
    # O ejecutar la versión compacta
    # programa_completo()
```

## 2. La gráfica de Feigenbaum (caos determinista)

En un especial _halloween_ de los Simpsons, Homero viaja al pasado a la era Jurásica y por un pequeño descuido cambia el curso de toda la historia en el futuro. Lo anterior es una referencia al cuento _A Sound of Thunder_ de 1952 del escritor de ciencia ficción Ray Bradbury en donde lo que se destruye en el pasado es una mariposa. Esta idea fue retomada y popularizada por el físico Edward Lorenz en 1972 cuando dio una conferencia en la Asociación Estadounidense para el Avance de la Ciencia (_American Association for the Advancement of Science_) titulada "¿El aleteo de una mariposa en Brasil puede provocar un tornado en Texas?". Dando paso al concepto del _efecto mariposa_, del cual probablemente hayas oído hablar y que es el ejemplo clásico de _caos determinista_ en sisteas climáticos. El caos determinista tambien aparece en muchos sistemas físicos más complejos, incluyendo especialmente la dinámica de fluidos.
```python
import numpy as np
import matplotlib.pyplot as plt

# a) Programa básico para el diagrama de Feigenbaum
def diagrama_feigenbaum_basico():
    """Versión básica del diagrama de Feigenbaum"""
    # Parámetros
    r_min, r_max, r_step = 1.0, 4.0, 0.01
    iteraciones_transitorio = 1000  # Iteraciones para estabilizar
    iteraciones_grafica = 1000      # Iteraciones para graficar
    
    # Generar valores de r
    r_values = np.arange(r_min, r_max + r_step, r_step)
    
    # Listas para almacenar puntos a graficar
    puntos_r = []
    puntos_x = []
    
    print("Calculando diagrama de Feigenbaum (versión básica)...")
    
    # Iterar sobre cada valor de r
    for r in r_values:
        x = 0.5  # Valor inicial
        
        # Iteraciones de transitorio (estabilización)
        for _ in range(iteraciones_transitorio):
            x = r * x * (1 - x)
        
        # Iteraciones para graficar
        for _ in range(iteraciones_grafica):
            x = r * x * (1 - x)
            puntos_r.append(r)
            puntos_x.append(x)
    
    # Graficar
    plt.figure(figsize=(12, 8))
    plt.plot(puntos_r, puntos_x, ',', color='blue', markersize=0.5, alpha=0.8)
    plt.xlabel('Parámetro r')
    plt.ylabel('x∞ (valor estable)')
    plt.title('Diagrama de Feigenbaum - Mapeo Logístico')
    plt.grid(True, alpha=0.3)
    plt.xlim(r_min, r_max)
    plt.ylim(0, 1)
    plt.show()
    
    return puntos_r, puntos_x

# c) Versión optimizada con arreglos
def diagrama_feigenbaum_optimizado():
    """Versión optimizada usando operaciones con arreglos"""
    # Parámetros
    r_min, r_max, r_step = 1.0, 4.0, 0.01
    iteraciones_transitorio = 1000
    iteraciones_grafica = 1000
    
    # Crear arreglos
    r_values = np.arange(r_min, r_max + r_step, r_step)
    x_values = np.full_like(r_values, 0.5)  # Todos los valores iniciales en 0.5
    
    print("Calculando diagrama de Feigenbaum (versión optimizada)...")
    
    # Iteraciones de transitorio
    for _ in range(iteraciones_transitorio):
        x_values = r_values * x_values * (1 - x_values)
    
    # Preparar arrays para graficar
    todos_r = []
    todos_x = []
    
    # Iteraciones para graficar
    for _ in range(iteraciones_grafica):
        x_values = r_values * x_values * (1 - x_values)
        todos_r.extend(r_values)
        todos_x.extend(x_values)
    
    # Graficar
    plt.figure(figsize=(12, 8))
    plt.plot(todos_r, todos_x, ',', color='red', markersize=0.5, alpha=0.8)
    plt.xlabel('Parámetro r')
    plt.ylabel('x∞ (valor estable)')
    plt.title('Diagrama de Feigenbaum - Versión Optimizada')
    plt.grid(True, alpha=0.3)
    plt.xlim(r_min, r_max)
    plt.ylim(0, 1)
    plt.show()
    
    return todos_r, todos_x

# Versión aún más optimizada (scatter plot)
def diagrama_feigenbaum_muy_optimizado():
    """Versión muy optimizada para mejor rendimiento"""
    # Parámetros
    r_min, r_max, r_step = 1.0, 4.0, 0.01
    iteraciones_transitorio = 1000
    iteraciones_grafica = 100
    
    # Crear arreglos
    r_values = np.arange(r_min, r_max + r_step, r_step)
    x_values = np.full_like(r_values, 0.5)
    
    print("Calculando diagrama de Feigenbaum (versión muy optimizada)...")
    
    # Transitorio
    for _ in range(iteraciones_transitorio):
        x_values = r_values * x_values * (1 - x_values)
    
    # Para cada iteración de graficado, almacenar resultados
    resultados_x = np.zeros((len(r_values), iteraciones_grafica))
    
    for i in range(iteraciones_grafica):
        x_values = r_values * x_values * (1 - x_values)
        resultados_x[:, i] = x_values
    
    # Preparar datos para scatter plot
    r_para_grafica = np.repeat(r_values, iteraciones_grafica)
    x_para_grafica = resultados_x.flatten()
    
    # Graficar con scatter para mejor rendimiento
    plt.figure(figsize=(12, 8))
    plt.scatter(r_para_grafica, x_para_grafica, s=0.1, alpha=0.5, color='green')
    plt.xlabel('Parámetro r')
    plt.ylabel('x∞ (valor estable)')
    plt.title('Diagrama de Feigenbaum - Scatter Plot Optimizado')
    plt.grid(True, alpha=0.3)
    plt.xlim(r_min, r_max)
    plt.ylim(0, 1)
    plt.show()
    
    return r_para_grafica, x_para_grafica

# Función para analizar el punto de transición al caos
def analizar_transicion_caos():
    """Analiza el valor de r donde comienza el comportamiento caótico"""
    # Parámetros con mayor resolución cerca de la transición
    r_min, r_max, r_step = 3.0, 3.6, 0.001
    iteraciones_transitorio = 2000
    iteraciones_grafica = 100
    
    r_values = np.arange(r_min, r_max + r_step, r_step)
    x_values = np.full_like(r_values, 0.5)
    
    # Transitorio
    for _ in range(iteraciones_transitorio):
        x_values = r_values * x_values * (1 - x_values)
    
    # Obtener valores únicos para cada r después de estabilizar
    valores_unicos_por_r = []
    
    for _ in range(iteraciones_grafica):
        x_values = r_values * x_values * (1 - x_values)
    
    # Para cada r, ver cuántos valores únicos tiene (indicador de caos)
    for i, r in enumerate(r_values):
        x_temp = x_values[i]
        # Hacer algunas iteraciones más para este r específico
        for _ in range(100):
            x_temp = r * x_temp * (1 - x_temp)
        
        # Verificar si tiene múltiples valores (caos)
        if len(set(np.round([x_temp] * 10, 6))) > 4:  # Más de 4 valores únicos
            print(f"Posible transición al caos alrededor de r = {r:.3f}")
            return r
    
    return None

# Programa principal
def main():
    print("=== DIAGRAMA DE FEIGENBAUM - MAPEO LOGÍSTICO ===")
    print("xₙ₊₁ = r * xₙ * (1 - xₙ)")
    print()
    
    # Ejecutar diferentes versiones
    version = input("Selecciona versión (1=básica, 2=optimizada, 3=scatter): ") or "2"
    
    if version == "1":
        diagrama_feigenbaum_basico()
    elif version == "2":
        diagrama_feigenbaum_optimizado()
    else:
        diagrama_feigenbaum_muy_optimizado()
    
    # b) Analizar transición al caos
    print("\nAnalizando transición al comportamiento caótico...")
    r_transicion = analizar_transicion_caos()
    
    if r_transicion:
        print(f"✓ El sistema transita al caos alrededor de r ≈ {r_transicion:.3f}")
        print("  (Este valor puede variar ligeramente dependiendo de la precisión)")
    else:
        print("No se pudo determinar claramente el punto de transición")
    
    print("\nObservaciones:")
    print("- Para r < 3: Comportamiento ordenado (punto fijo)")
    print("- 3 ≤ r < 3.45: Duplicación de período (órbitas periódicas)")
    print("- r ≈ 3.57: Comienzo del comportamiento caótico")
    print("- r > 3.57: Caos determinista completo")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# a) Versión básica: conjunto de Mandelbrot en blanco y negro
def mandelbrot_basico(N=100, max_iter=100):
    """
    Calcula el conjunto de Mandelbrot básico (blanco y negro)
    N: resolución de la cuadrícula
    max_iter: número máximo de iteraciones
    """
    print(f"Calculando conjunto de Mandelbrot ({N}x{N})...")
    
    # Crear cuadrícula en el plano complejo
    x = np.linspace(-2, 2, N)
    y = np.linspace(-2, 2, N)
    X, Y = np.meshgrid(x, y)
    C = X + 1j * Y  # Puntos del plano complejo
    
    # Inicializar array para resultados
    mandelbrot_set = np.zeros((N, N), dtype=bool)
    z = np.zeros_like(C)
    
    # Iterar para cada punto
    for i in range(max_iter):
        z = z**2 + C
        # Puntos que han escapado (|z| > 2)
        escaped = np.abs(z) > 2
        # Actualizar solo los puntos que no han escapado aún
        mandelbrot_set |= escaped
        # Para puntos que escaparon, no seguir iterando
        z[escaped] = np.nan
    
    # Invertir: True para puntos que NO están en el conjunto
    return ~mandelbrot_set

# b) Versión avanzada con colores según iteraciones
def mandelbrot_colorido(N=100, max_iter=100, log_scale=True):
    """
    Calcula el conjunto de Mandelbrot con colores según iteraciones
    """
    print(f"Calculando Mandelbrot colorido ({N}x{N}, {max_iter} iteraciones)...")
    
    # Crear cuadrícula
    x = np.linspace(-2, 2, N)
    y = np.linspace(-2, 2, N)
    X, Y = np.meshgrid(x, y)
    C = X + 1j * Y
    
    # Array para contar iteraciones hasta escape
    iterations = np.zeros((N, N), dtype=int)
    z = np.zeros_like(C)
    
    # Máscara de puntos que aún no han escapado
    active = np.ones_like(C, dtype=bool)
    
    for i in range(max_iter):
        # Solo calcular para puntos activos
        z[active] = z[active]**2 + C[active]
        
        # Verificar qué puntos han escapado en esta iteración
        escaped = (np.abs(z) > 2) & active
        
        # Registrar iteración de escape
        iterations[escaped] = i
        
        # Desactivar puntos que escaparon
        active[escaped] = False
        
        # Si todos los puntos han escapado, terminar antes
        if not np.any(active):
            break
    
    # Para puntos que nunca escaparon, asignar max_iter
    iterations[active] = max_iter
    
    if log_scale:
        # Usar logaritmo para mejor visualización de detalles
        iterations = np.log(iterations + 1)
    
    return iterations

# Función para visualizar resultados
def visualizar_mandelbrot(resultado, tipo='basico', cmap='hot'):
    """
    Visualiza el conjunto de Mandelbrot
    """
    plt.figure(figsize=(12, 10))
    
    if tipo == 'basico':
        plt.imshow(resultado, extent=[-2, 2, -2, 2], cmap='binary', origin='lower')
        plt.title('Conjunto de Mandelbrot', fontsize=16)
    else:
        im = plt.imshow(resultado, extent=[-2, 2, -2, 2], cmap=cmap, origin='lower')
        plt.colorbar(im, label='Iteraciones (log scale)' if 'log' in str(resultado.dtype) else 'Iteraciones')
        plt.title('Conjunto de Mandelbrot - Versión Colorida', fontsize=16)
    
    plt.xlabel('Re(c)')
    plt.ylabel('Im(c)')
    plt.grid(False)
    plt.tight_layout()
    plt.show()

# Versión optimizada con vectorización completa
def mandelbrot_vectorizado(N=100, max_iter=100):
    """
    Versión completamente vectorizada para mejor rendimiento
    """
    print(f"Calculando Mandelbrot vectorizado ({N}x{N})...")
    
    x = np.linspace(-2, 2, N)
    y = np.linspace(-2, 2, N)
    c = x[:, np.newaxis] + 1j * y[np.newaxis, :]
    
    # Array para resultados
    output = np.zeros(c.shape, dtype=int)
    z = np.zeros(c.shape, dtype=complex)
    
    for i in range(max_iter):
        # Aplicar la fórmula a todos los puntos
        z = z**2 + c
        
        # Encontrar puntos que han escapado
        escaped = (np.abs(z) > 2) & (output == 0)
        output[escaped] = i
    
    # Puntos que nunca escaparon
    output[output == 0] = max_iter
    
    return output

# Función para explorar diferentes regiones
def mandelbrot_zoom(center_x=0, center_y=0, width=4, height=4, N=100, max_iter=100):
    """
    Calcula el conjunto de Mandelbrot en una región específica
    """
    x_min = center_x - width/2
    x_max = center_x + width/2
    y_min = center_y - height/2
    y_max = center_y + height/2
    
    print(f"Zoom: [{x_min:.3f}, {x_max:.3f}] x [{y_min:.3f}, {y_max:.3f}]")
    
    x = np.linspace(x_min, x_max, N)
    y = np.linspace(y_min, y_max, N)
    X, Y = np.meshgrid(x, y)
    C = X + 1j * Y
    
    iterations = np.zeros((N, N), dtype=int)
    z = np.zeros_like(C)
    
    for i in range(max_iter):
        z = z**2 + C
        escaped = (np.abs(z) > 2) & (iterations == 0)
        iterations[escaped] = i
    
    iterations[iterations == 0] = max_iter
    
    return iterations, (x_min, x_max, y_min, y_max)

# Programa principal
def main():
    print("=== CONJUNTO DE MANDELBROT ===")
    print("zₙ₊₁ = zₙ² + c")
    print()
    
    # Seleccionar modo
    print("Modos disponibles:")
    print("1. Básico (blanco y negro)")
    print("2. Colorido (iteraciones)")
    print("3. Zoom en regiones interesantes")
    print("4. Alta resolución")
    
    modo = input("Selecciona modo (1-4): ") or "2"
    
    # Parámetros según el modo
    if modo == "1":
        # Modo básico
        resultado = mandelbrot_basico(N=200, max_iter=100)
        visualizar_mandelbrot(resultado, tipo='basico')
        
    elif modo == "2":
        # Modo colorido
        resultado = mandelbrot_colorido(N=300, max_iter=100, log_scale=True)
        visualizar_mandelbrot(resultado, tipo='colorido', cmap='jet')
        
    elif modo == "3":
        # Zoom en regiones interesantes
        regiones = [
            (-0.75, 0.1, 0.5, 0.5),  # Región del "valle"
            (-0.16, 1.04, 0.1, 0.1), # Mini-Mandelbrot
            (-1.77, 0, 0.1, 0.1)     # Borde izquierdo
        ]
        
        for i, (cx, cy, w, h) in enumerate(regiones):
            resultado, extent = mandelbrot_zoom(cx, cy, w, h, N=200, max_iter=200)
            plt.figure(figsize=(8, 8))
            plt.imshow(np.log(resultado + 1), extent=extent, cmap='hot', origin='lower')
            plt.title(f'Zoom {i+1}: Mandelbrot en [{extent[0]:.3f}, {extent[1]:.3f}]')
            plt.colorbar()
            plt.show()
            
    else:
        # Alta resolución (puede tomar más tiempo)
        resultado = mandelbrot_vectorizado(N=500, max_iter=200)
        plt.figure(figsize=(12, 10))
        plt.imshow(np.log(resultado + 1), extent=[-2, 2, -2, 2], 
                  cmap='nipy_spectral', origin='lower')
        plt.colorbar(label='log(Iteraciones + 1)')
        plt.title('Conjunto de Mandelbrot - Alta Resolución', fontsize=16)
        plt.show()

# Información sobre el conjunto de Mandelbrot
def info_mandelbrot():
    """Proporciona información sobre el conjunto de Mandelbrot"""
    print("\n" + "="*50)
    print("INFORMACIÓN SOBRE EL CONJUNTO DE MANDELBROT")
    print("="*50)
    print("• Descubierto por Benoît Mandelbrot (1980)")
    print("• Uno de los fractales más famosos")
    print("• Definido por: zₙ₊₁ = zₙ² + c")
    print("• z₀ = 0, c ∈ ℂ")
    print("• Punto en el conjunto si |zₙ| no diverge")
    print()
    print("Características notables:")
    print("✓ Auto-similaridad (estructuras que se repiten)")
    print("✓ Frontera infinitamente compleja")
    print("✓ Mini-Mandelbrots en diferentes escalas")
    print("✓ Conexión con la teoría del caos")

if __name__ == "__main__":
    main()
    info_mandelbrot()
```
