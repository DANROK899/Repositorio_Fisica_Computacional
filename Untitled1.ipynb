```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constantes
a0 = 1.0  # Radio de Bohr en unidades atómicas
P1 = 1e-5  # Primera superficie de probabilidad
P2 = 1e-3  # Segunda superficie de probabilidad

def r_from_P(P, theta, phi):
    """Calcula r para P constante dado theta y phi"""
    A = 32 * np.pi * a0**5 * P
    sin2_theta = np.sin(theta)**2
    cos2_phi = np.cos(phi)**2
    
    # Evitar división por cero
    denominator = sin2_theta * cos2_phi
    denominator = np.where(denominator < 1e-10, 1e-10, denominator)
    
    rhs = A / denominator
    
    # Resolver r^2 * exp(-r/a0) = rhs numéricamente
    # Usamos el método de Newton
    r_guess = a0 * np.log(rhs / (a0**2) + 1) if rhs > 0 else a0
    
    r = r_guess
    for _ in range(10):
        f = r**2 * np.exp(-r/a0) - rhs
        dfdr = (2*r - r**2/a0) * np.exp(-r/a0)
        r = r - f/dfdr
    
    return np.real(r)

# Generar malla de ángulos
theta = np.linspace(0, np.pi, 50)
phi = np.linspace(0, 2*np.pi, 50)
theta, phi = np.meshgrid(theta, phi)

# Calcular superficies para P1 y P2
r1 = np.zeros_like(theta)
r2 = np.zeros_like(theta)

for i in range(theta.shape[0]):
    for j in range(theta.shape[1]):
        r1[i,j] = r_from_P(P1, theta[i,j], phi[i,j])
        r2[i,j] = r_from_P(P2, theta[i,j], phi[i,j])

# Convertir a coordenadas cartesianas
x1 = r1 * np.sin(theta) * np.cos(phi)
y1 = r1 * np.sin(theta) * np.sin(phi)
z1 = r1 * np.cos(theta)

x2 = r2 * np.sin(theta) * np.cos(phi)
y2 = r2 * np.sin(theta) * np.sin(phi)
z2 = r2 * np.cos(theta)

# Graficar
fig = plt.figure(figsize=(12, 6))

# Superficie P = 1e-5
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(x1, y1, z1, alpha=0.7, cmap='Blues')
ax1.set_title(f'Superficie P = {P1}')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')

# Superficie P = 1e-3
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(x2, y2, z2, alpha=0.7, cmap='Reds')
ax2.set_title(f'Superficie P = {P2}')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')

plt.tight_layout()
plt.show()

# Guardar datos para Excel/Origin (primer cuadrante)
# Reducir número de puntos para manejar en Excel
theta_red = np.linspace(0, np.pi, 20)
phi_red = np.linspace(0, 2*np.pi, 20)
theta_red, phi_red = np.meshgrid(theta_red, phi_red)

r1_red = np.zeros_like(theta_red)
r2_red = np.zeros_like(theta_red)

for i in range(theta_red.shape[0]):
    for j in range(theta_red.shape[1]):
        r1_red[i,j] = r_from_P(P1, theta_red[i,j], phi_red[i,j])
        r2_red[i,j] = r_from_P(P2, theta_red[i,j], phi_red[i,j])

x1_red = r1_red * np.sin(theta_red) * np.cos(phi_red)
y1_red = r1_red * np.sin(theta_red) * np.sin(phi_red)
z1_red = r1_red * np.cos(theta_red)

x2_red = r2_red * np.sin(theta_red) * np.cos(phi_red)
y2_red = r2_red * np.sin(theta_red) * np.sin(phi_red)
z2_red = r2_red * np.cos(theta_red)

# Exportar a CSV
import pandas as pd

df1 = pd.DataFrame({
    'x1': x1_red.flatten(),
    'y1': y1_red.flatten(), 
    'z1': z1_red.flatten()
})

df2 = pd.DataFrame({
    'x2': x2_red.flatten(),
    'y2': y2_red.flatten(),
    'z2': z2_red.flatten()
})

df1.to_csv('superficie_P1e-5.csv', index=False)
df2.to_csv('superficie_P1e-3.csv', index=False)

print("Datos exportados para Excel/Origin")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constantes físicas (en unidades atómicas)
a0 = 1.0  # radio de Bohr

# Función radial R_{21}(r)
def R21(r):
    return (1/(2*np.sqrt(6))) * (1/a0**1.5) * (r/a0) * np.exp(-r/(2*a0))

# Función de onda completa para p_x
def psi_px(x, y, z):
    r = np.sqrt(x**2 + y**2 + z**2)
    # Usar coordenadas cartesianas directamente para evitar problemas con phi
    # ψ_px ∝ x * exp(-r/2a0) 
    radial = R21(r)
    # En coordenadas cartesianas: sinθ cosφ = x/r
    angular = np.sqrt(3/(4*np.pi)) * (x / np.where(r == 0, 1, r))
    
    return radial * angular

# Densidad de probabilidad |ψ|²
def probability_density(x, y, z):
    psi = psi_px(x, y, z)
    return np.abs(psi)**2

# Visualización simplificada usando cortes 2D
def plot_probability_contours():
    """Visualización mediante cortes 2D en diferentes planos"""
    
    # Crear grid 2D para diferentes planos
    x = np.linspace(-10, 10, 200)
    z = np.linspace(-10, 10, 200)
    X, Z = np.meshgrid(x, z)
    
    # Plano y=0 (corte xz)
    Y_zero = np.zeros_like(X)
    P_xz = probability_density(X, Y_zero, Z)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # P = 10^-5
    contour1 = ax1.contour(X, Z, P_xz, levels=[1e-5], colors='blue', linewidths=2)
    ax1.set_title('Contorno P = $10^{-5}$ (plano xz, y=0)', fontsize=14)
    ax1.set_xlabel('x (a₀)', fontsize=12)
    ax1.set_ylabel('z (a₀)', fontsize=12)
    ax1.grid(True, alpha=0.3)
    ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    ax1.set_aspect('equal')
    
    # P = 10^-3  
    contour2 = ax2.contour(X, Z, P_xz, levels=[1e-3], colors='red', linewidths=2)
    ax2.set_title('Contorno P = $10^{-3}$ (plano xz, y=0)', fontsize=14)
    ax2.set_xlabel('x (a₀)', fontsize=12)
    ax2.set_ylabel('z (a₀)', fontsize=12)
    ax2.grid(True, alpha=0.3)
    ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    ax2.set_aspect('equal')
    
    plt.tight_layout()
    plt.show()
    
    return contour1, contour2

# Visualización 3D alternativa usando scatter plot
def plot_3d_probability_surface():
    """Visualización 3D usando puntos scatter"""
    
    print("Calculando superficie 3D... Esto puede tomar un momento...")
    
    # Crear grid 3D más pequeño para mejor rendimiento
    x = np.linspace(-8, 8, 40)
    y = np.linspace(-8, 8, 40) 
    z = np.linspace(-8, 8, 40)
    X, Y, Z = np.meshgrid(x, y, z, indexing='ij')
    
    P = probability_density(X, Y, Z)
    P_max = np.max(P)
    
    # Encontrar puntos cercanos a los valores de probabilidad deseados
    prob_level_1 = 1e-5
    prob_level_2 = 1e-3
    
    mask1 = (P > prob_level_1 * 0.8) & (P < prob_level_1 * 1.2)
    mask2 = (P > prob_level_2 * 0.8) & (P < prob_level_2 * 1.2)
    
    fig = plt.figure(figsize=(15, 6))
    
    # P = 10^-5
    ax1 = fig.add_subplot(121, projection='3d')
    points1 = mask1.nonzero()
    if len(points1[0]) > 0:
        ax1.scatter(X[points1], Y[points1], Z[points1], 
                   alpha=0.6, s=1, color='blue')
    ax1.set_title('Superficie P ≈ $10^{-5}$ para orbital $p_x$', fontsize=12)
    ax1.set_xlabel('x (a₀)')
    ax1.set_ylabel('y (a₀)')
    ax1.set_zlabel('z (a₀)')
    
    # P = 10^-3
    ax2 = fig.add_subplot(122, projection='3d')
    points2 = mask2.nonzero()
    if len(points2[0]) > 0:
        ax2.scatter(X[points2], Y[points2], Z[points2], 
                   alpha=0.6, s=1, color='red')
    ax2.set_title('Superficie P ≈ $10^{-3}$ para orbital $p_x$', fontsize=12)
    ax2.set_xlabel('x (a₀)')
    ax2.set_ylabel('y (a₀)')
    ax2.set_zlabel('z (a₀)')
    
    plt.tight_layout()
    plt.show()

# Función para analizar la distribución de probabilidad
def analyze_probability_distribution():
    """Analiza la distribución de probabilidad del orbital p_x"""
    
    x = np.linspace(-10, 10, 100)
    z = np.linspace(-10, 10, 100)
    X, Z = np.meshgrid(x, z)
    Y = np.zeros_like(X)
    
    P = probability_density(X, Y, Z)
    
    print("=== ANÁLISIS DEL ORBITAL p_x ===")
    print(f"Probabilidad máxima: {np.max(P):.2e}")
    print(f"Probabilidad mínima: {np.min(P):.2e}")
    print(f"Rango de probabilidad: {np.max(P)/np.min(P):.2e}")
    
    # Encontrar posición del máximo
    max_idx = np.argmax(P)
    max_x = X.flatten()[max_idx]
    max_z = Z.flatten()[max_idx]
    print(f"Posición del máximo: x = {max_x:.2f} a₀, z = {max_z:.2f} a₀")
    
    # Calcular volumen donde P > 1e-5
    volume_fraction_1e5 = np.sum(P > 1e-5) / P.size
    volume_fraction_1e3 = np.sum(P > 1e-3) / P.size
    print(f"Fracción volumétrica con P > 1e-5: {volume_fraction_1e5:.3f}")
    print(f"Fracción volumétrica con P > 1e-3: {volume_fraction_1e3:.3f}")

# Ejecutar todas las visualizaciones
if __name__ == "__main__":
    # Configurar estilo de plots
    plt.rcParams.update({'font.size': 10})
    
    print("Visualización del orbital p_x del átomo de hidrógeno")
    print("=" * 50)
    
    # Análisis de la distribución
    analyze_probability_distribution()
    
    print("\nGenerando gráficos 2D...")
    # Gráficos 2D (más rápidos y estables)
    contour1, contour2 = plot_probability_contours()
    
    print("\nGenerando gráfico 3D...")
    # Gráfico 3D (opcional - puede ser lento)
    plot_3d_probability_surface()
    
    print("\n" + "=" * 50)
    print("RESUMEN:")
    print("• El orbital p_x tiene simetría bilateral a lo largo del eje x")
    print("• Presenta un plano nodal en y=0 (donde la probabilidad es cero)")
    print("• La superficie P = 10⁻³ es más pequeña que P = 10⁻⁵")
    print("• La forma característica es de 'lóbulo doble' orientado en x")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# --- Constantes físicas ---
a0 = 1.0  # Radio de Bohr en unidades atómicas

# --- Definición de la función de onda p_x ---
def psi_px(x, y, z):
    r = np.sqrt(x**2 + y**2 + z**2)
    return (1 / (4 * np.sqrt(2 * np.pi) * a0**(5/2))) * x * np.exp(-r / (2 * a0))

# --- Malla 3D ---
lim = 6 * a0
N = 80
x = np.linspace(-lim, lim, N)
y = np.linspace(-lim, lim, N)
z = np.linspace(-lim, lim, N)
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')

# --- Densidad de probabilidad ---
Psi = psi_px(X, Y, Z)
P = np.abs(Psi)**2

# --- Niveles de probabilidad a graficar ---
levels = [1e-5, 1e-3]
colors = ['blue', 'red']

# --- Gráfica de superficies isovalor ---
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

for lvl, color in zip(levels, colors):
    verts, faces, normals, values = measure.marching_cubes(P, lvl)
    verts = verts / N * 2 * lim - lim  # escalar al espacio físico
    mesh = Poly3DCollection(verts[faces], alpha=0.4, facecolor=color)
    ax.add_collection3d(mesh)

ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)
ax.set_xlabel('x / a0')
ax.set_ylabel('y / a0')
ax.set_zlabel('z / a0')
ax.set_title('Superficies de probabilidad constante del orbital p_x')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_orbital_px(P_constant, a0=1.0, num_points=50, plot_range=6):
    """
    Grafica la superficie de probabilidad constante para el orbital p_x.

    Args:
        P_constant (float): El valor constante de la densidad de probabilidad (P = |Psi|^2)
                            para la superficie a graficar.
        a0 (float): Radio de Bohr (por simplicidad, lo fijamos en 1 unidad).
        num_points (int): Número de puntos en cada dimensión para la rejilla.
        plot_range (float): Rango del plot en cada eje (-plot_range a +plot_range).
    """

    # La constante de normalización N para el orbital 2px es:
    # N = 1 / (4 * sqrt(2 * pi)) * (1/a0)^(5/2)
    # Sin embargo, para graficar una superficie de probabilidad constante P,
    # el valor exacto de N se "absorbe" en la constante P.
    # Lo que necesitamos es resolver la ecuación:
    # P_constant = N^2 * x^2 * exp(-r/a0)
    # Para la gráfica, es más fácil trabajar con la forma
    # x^2 * exp(-r/a0) = P_constant / N^2 = Constante_Efectiva

    # Creamos una rejilla de puntos en el espacio 3D
    x = np.linspace(-plot_range, plot_range, num_points)
    y = np.linspace(-plot_range, plot_range, num_points)
    z = np.linspace(-plot_range, plot_range, num_points)
    X, Y, Z = np.meshgrid(x, y, z)

    # Calculamos r para cada punto de la rejilla
    R = np.sqrt(X**2 + Y**2 + Z**2)

    # Calculamos la densidad de probabilidad (ignorando la constante de normalización N^2
    # ya que buscamos un valor P_constant. Se puede ajustar P_constant para que tenga sentido)
    # La parte esencial es x^2 * exp(-R/a0)
    density = (X**2) * np.exp(-R / a0)

    # Creamos la figura y los ejes 3D
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Usamos Marching Cubes o similar para extraer la isosuperficie.
    # El método contourf no es adecuado para 3D.
    # Una forma de visualizar esto sin librerías externas como scikit-image
    # es pintar puntos discretos donde la densidad sea aproximadamente P_constant.
    # Sin embargo, eso no dará una superficie suave.

    # Una mejor aproximación sin librerías externas de isosuperficies es:
    # Si tenemos una función f(x,y,z) = C, podemos rebanar el espacio.
    # Pero para una superficie de probabilidad constante, necesitamos una librería
    # de isosuperficies como scikit-image para un resultado óptimo.
    # Como no puedo usar scikit-image aquí, simularé la forma coloreando puntos
    # o mostrando las "rebanadas" donde la densidad se aproxima a P_constant.

    # Una forma de visualizar la forma sin isosuperficies directas:
    # Estableceremos umbrales y haremos un plot de dispersión
    # (esto no es una superficie continua, pero muestra la forma)
    threshold = P_constant # Este es el valor de la "constante efectiva" P_constant / N^2

    # Encuentra los índices donde la densidad está cerca del umbral
    # Para una superficie, buscamos puntos donde density ≈ threshold.
    # La forma más común de representar esto sin librerías avanzadas
    # es utilizando el concepto de isovalores.
    # Aquí, ajustamos la escala de P_constant para que sea visible en nuestra escala de plot_range.
    # El valor P_constant debe ser ajustado experimentalmente o mediante cálculo de N^2.

    # Para un mejor resultado visual, podríamos usar una constante de referencia.
    # Si N=1 (ignorando la normalización real para ver la forma base),
    # entonces P_constant es el valor de x^2 * exp(-r/a0) que queremos graficar.

    # Vamos a crear una función de "isosuperficie" para un rango de valores de P
    # usando una librería externa para ello, en un entorno real de Python
    # aquí voy a usar una aproximación con puntos para mostrar la idea.

    # Generación de la superficie con np.where para puntos cercanos a la constante
    # Esta es una aproximación para entornos sin marching_cubes.
    # Para una visualización más precisa, se necesitaría 'skimage.measure.marching_cubes'.

    # Vamos a ajustar P_constant para que se adapte al rango de 'density'
    # Obtenemos los valores máximo y mínimo de density
    max_density = np.max(density)
    min_density = np.min(density)
    print(f"Rango de densidad calculada: [{min_density:.2e}, {max_density:.2e}]")
    print(f"Graficando para P_constant = {P_constant}")

    # Si P_constant está fuera del rango de densidad calculada, la gráfica estará vacía
    if not (min_density <= P_constant <= max_density):
        print(f"Advertencia: P_constant ({P_constant:.2e}) está fuera del rango de densidad calculada.")
        print("Asegúrate de que P_constant esté dentro de los límites de la función de densidad para ver el gráfico.")
        print(f"Intenta un P_constant más cercano a {max_density/10:.2e} o {max_density/100:.2e}")
        return

    # Usaremos una aproximación simple de puntos para visualizar la isosuperficie
    # Debido a las limitaciones de un ambiente que no puede ejecutar librerías avanzadas
    # de visualización 3D como mayavi o scikit-image.
    # Esto no generará una superficie suave, sino una nube de puntos que delinea la forma.

    # Aumentar la densidad de puntos para una mejor visualización de la forma
    # Genera puntos de manera esférica para cubrir el espacio de manera más eficiente
    theta = np.linspace(0, np.pi, num_points)
    phi = np.linspace(0, 2 * np.pi, num_points)
    theta, phi = np.meshgrid(theta, phi)

    # Vamos a resolver la ecuación de forma implícita
    # P_0 = x^2 * exp(-r/a0)
    # Esto es mejor con métodos de isosuperficie.
    # Para una representación sencilla, lo que se suele hacer es plotear X,Y,Z
    # de manera que se cumpla la condición.

    # Voy a generar los puntos de la superficie manualmente, resolviendo para r
    # Esto requiere resolver una ecuación trascendental (r = -a0 * ln(P_constant / x^2)).
    # Es más sencillo usar las combinaciones lineales de los armónicos esféricos.
    # Recordamos que p_x es proporcional a sin(theta)cos(phi) * R(r)
    # Entonces |psi|^2 es proporcional a sin^2(theta)cos^2(phi) * R(r)^2

    # Vamos a usar la definición directamente en cartesianas
    # P_constant = x^2 * exp(-r/a0)  (simplificada, ignorando N^2)
    # Tenemos que encontrar r en función de x para cada P_constant

    # Esto es una manera más efectiva de graficar una superficie implicita
    # Se puede usar vtk o trimesh, pero para matplotlib puro es más complicado.
    # Aquí un intento de usar el método de contorno en 3D
    # https://matplotlib.org/stable/gallery/mplot3d/surface3d_2.html

    # Una forma de crear una "superficie" sin librerías como skimage.
    # Vamos a iterar sobre el rango para encontrar los puntos que cumplen la condición.
    xs, ys, zs = [], [], []
    step = plot_range / (num_points / 2)
    for i in range(num_points):
        for j in range(num_points):
            for k in range(num_points):
                curr_x = x[i]
                curr_y = y[j]
                curr_z = z[k]
                curr_r = np.sqrt(curr_x**2 + curr_y**2 + curr_z**2)

                # Evitar divisiones por cero o problemas con logaritmos de cero
                if curr_r == 0:
                    continue

                # Si la densidad está cerca de P_constant, añadir el punto
                # Usamos un pequeño margen para capturar puntos discretos.
                if P_constant * 0.95 <= (curr_x**2) * np.exp(-curr_r / a0) <= P_constant * 1.05:
                    xs.append(curr_x)
                    ys.append(curr_y)
                    zs.append(curr_z)

    if not xs:
        print(f"No se encontraron puntos para P_constant = {P_constant}. Intenta ajustar P_constant o el rango de búsqueda.")
        return

    ax.scatter(xs, ys, zs, s=0.5, c='blue', alpha=0.5, label=f'P = {P_constant:.0e}')
    ax.set_xlabel('x (a0)')
    ax.set_ylabel('y (a0)')
    ax.set_zlabel('z (a0)')
    ax.set_title(f'Superficie de Probabilidad Constante para orbital $p_x$')
    ax.legend()
    plt.tight_layout()
    plt.show()

# --- Graficar para los valores solicitados ---
print("Graficando para P = 10^-5")
plot_orbital_px(P_constant=1e-5, plot_range=6)

print("\nGraficando para P = 10^-3")
plot_orbital_px(P_constant=1e-3, plot_range=6)

# Nota: Los valores de P_constant son "efectivos" aquí.
# Para hacerlos corresponder a un P real, se necesitaría la constante N^2.
# Por ejemplo, si N^2 = 1, entonces P_constant es la densidad real.
# La forma del orbital es la misma, solo el tamaño cambia.

# Para obtener una visualización más precisa y con superficies continuas,
# en un entorno real de Jupyter, podrías usar la librería 'scikit-image'
# (pip install scikit-image) y su función 'marching_cubes'.
# Ejemplo de cómo se vería la lógica con marching_cubes:

# from skimage.measure import marching_cubes_lewiner
# # ... (código para X, Y, Z, R, density como arriba)
# # verts, faces, _, _ = marching_cubes_lewiner(density, level=P_constant, spacing=(dx, dy, dz))
# # ax.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2], cmap='viridis', lw=1)

# Pero dado que el entorno actual no permite librerías externas de este tipo,
# el método de scatter plot es la mejor aproximación para delinear la forma.
# Ajusta 'plot_range' y 'num_points' para un mejor resultado.
# Los valores de P_constant son relativos a la escala de 'density'.
# Aumentar P_constant hará que el orbital sea más pequeño.
# Disminuir P_constant hará que el orbital sea más grande.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage.measure import marching_cubes  # Esta es la función clave

def plot_px_isosurface(P_values, a0=1.0, num_points=80, plot_range=20):
    """
    Grafica múltiples superficies de probabilidad constante (isosuperficies)
    para el orbital p_x usando scikit-image (marching_cubes) y matplotlib.
    
    Args:
        P_values (list): Lista de tuplas (valor_P, color). 
                         P es el valor de la densidad de prob. efectiva x^2 * exp(-r/a0).
        a0 (float): Radio de Bohr (lo fijamos en 1.0).
        num_points (int): Resolución de la rejilla. 80 es un buen balance.
        plot_range (float): Rango del gráfico (-range a +range).
    """
    
    print(f"Generando rejilla {num_points}x{num_points}x{num_points} en un rango de +/- {plot_range} a0...")
    
    # 1. Crear la rejilla 3D
    x_lin = np.linspace(-plot_range, plot_range, num_points)
    y_lin = np.linspace(-plot_range, plot_range, num_points)
    z_lin = np.linspace(-plot_range, plot_range, num_points)
    X, Y, Z = np.meshgrid(x_lin, y_lin, z_lin)
    
    # 2. Calcular la distancia 'r' y la densidad de probabilidad
    R = np.sqrt(X**2 + Y**2 + Z**2)
    
    # Calculamos la parte variable de la densidad de probabilidad |Psi_px|^2
    # |Psi_px|^2 = N^2 * x^2 * exp(-r/a0)
    # Graficamos la isosuperficie de la parte variable:
    # density = x^2 * exp(-r/a0)
    # Evitamos r=0 para no tener exp(inf) si r es muy pequeño
    R = R + 1e-9  # Evitar división por cero si a0=0, aunque aquí R está en el exp
    density = (X**2) * np.exp(-R / a0)
    
    print(f"Rango de densidad calculada: [{density.min():.2e}, {density.max():.2e}]")
    
    # 3. Preparar el gráfico 3D
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111, projection='3d')

    # 4. Iterar sobre los valores de P y graficar cada isosuperficie
    for P_val, color in P_values:
        
        # 5. ¡La magia! Usar Marching Cubes para encontrar la superficie
        # Esta función calcula los vértices y las caras de la superficie 3D
        # donde la 'density' es igual a 'P_val'.
        try:
            verts, faces, _, _ = marching_cubes(
                density, 
                level=P_val, 
                spacing=(x_lin[1]-x_lin[0], y_lin[1]-y_lin[0], z_lin[1]-z_lin[0])
            )
        except ValueError:
            print(f"Advertencia: No se encontró superficie para P = {P_val:.2e}.")
            print("Intenta un valor dentro del rango de densidad.")
            continue

        # 6. Marching cubes devuelve vértices en coordenadas de la rejilla,
        # así que los re-escalamos a nuestro 'plot_range'
        verts = verts + np.array([x_lin[0], y_lin[0], z_lin[0]])

        # 7. Graficar la superficie 3D usando los triángulos (caras)
        ax.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2],
                        color=color, 
                        alpha=0.4,  # Transparencia para ver a través
                        label=f'P = {P_val:.0e}')
    
    # 8. Ajustes finales del gráfico
    ax.set_xlabel('x ($a_0$)', fontsize=12)
    ax.set_ylabel('y ($a_0$)', fontsize=12)
    ax.set_zlabel('z ($a_0$)', fontsize=12)
    ax.set_title('Superficies de Probabilidad Constante del orbital $p_x$', fontsize=16)
    
    # Establecer límites iguales para una proporción correcta
    ax.set_xlim([-plot_range, plot_range])
    ax.set_ylim([-plot_range, plot_range])
    ax.set_zlim([-plot_range, plot_range])
    
    ax.legend()
    plt.tight_layout()
    plt.show()

# --- Ejecutar la función ---
# Usamos los valores P = 10^-3 (más interna, roja) y P = 10^-5 (más externa, azul)
# (Estos son valores de "densidad efectiva", no la P absoluta, pero definen la forma)

# Nota: P=10^-3 es una densidad MÁS ALTA, por lo que la superficie es MÁS PEQUEÑA.
#       P=10^-5 es una densidad MÁS BAJA, por lo que la superficie es MÁS GRANDE.
valores_P = [
    (1e-3, 'red'),   # Superficie interna (como en tu imagen)
    (1e-5, 'blue')   # Superficie externa (como en tu imagen)
]

# Usamos un rango de 20 para ver bien la superficie de 10^-5
# Aumentar 'num_points' (ej. a 100) da más detalle pero tarda más.
plot_px_isosurface(valores_P, plot_range=20, num_points=80)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros
a0 = 1  # Radio de Bohr en unidades atómicas

# Parte radial R_{21}(r)
def R_21(r):
    return (1 / np.sqrt(3)) * (1 / (2 * a0)**(3/2)) * (r / a0) * np.exp(-r / (2 * a0))

# Función de onda para p_x
def psi_px(r, theta, phi):
    return R_21(r) * np.sin(theta) * np.cos(phi)

# Malla esférica
r = np.linspace(0.1, 10, 100)
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)
r, theta, phi = np.meshgrid(r, theta, phi)

# Coordenadas cartesianas
x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)

# Densidad de probabilidad
P = np.abs(psi_px(r, theta, phi))**2

# Umbrales de probabilidad
thresholds = [1e-5, 1e-3]
colors = ['blue', 'red']

# Crear figura
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Superficies de probabilidad constante para el orbital $p_x$")

# Graficar puntos que cumplen con cada umbral
for threshold, color in zip(thresholds, colors):
    mask = np.isclose(P, threshold, rtol=0.2)
    ax.scatter(x[mask], y[mask], z[mask], s=1, alpha=0.5, label=f'$|\Psi|^2 = {threshold}$', color=color)

# Estética
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()
plt.tight_layout()
plt.show()
```
