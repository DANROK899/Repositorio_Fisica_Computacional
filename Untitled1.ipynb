### Resumen

En esta práctica se estudiaron las propiedades mecánicas de un polímero ,en este caso, el chicle bajo diferentes condiciones de deformación y temperatura. Se preparó una muestra estándar de 1.5 cm de longitud inicial. Se aplicaron cuatro pruebas: estiramiento lento y rápido a temperatura ambiente, y estiramiento lento y rápido a baja temperatura (tras refrigeración durante 30 minutos). Los resultados demostraron que el material es altamente viscoelástico: presentó un comportamiento dúctil extremo bajo deformación lenta, una ductilidad reducida bajo deformación rápida, y un comportamiento completamente frágil a baja temperatura, fracturándose súbitamente. Se concluye que las propiedades mecánicas del chicle dependen críticamente de la velocidad de deformación y de la temperatura, exhibiendo una clara transición dúctil-frágil. Además, se destaca la importancia de utilizar equipo de medición y control adecuado para minimizar errores experimentales.

### Metodología

1. **Preparación del Material:** Se mascaron dos unidades de chicle hasta obtener una consistencia suave y homogénea.
2. **Selección de Muestra:** Se formaron óvalos de 1.5 cm y 5 cm. En pruebas preliminares, la muestra de 5 cm superó los 4 metros de elongación sin fracturarse, excediendo la capacidad de medición. Por lo tanto, se seleccionó la muestra de 1.5 cm de longitud inicial ($L_0$) para todos los experimentos.
3. **Prueba 1 (Deformación Lenta, Temp. Ambiente):** La muestra de 1.5 cm se estiró lentamente con las manos, midiendo la longitud final ($L_f$) en el punto de fractura con un flexómetro.
4. **Prueba 2 (Deformación Rápida, Temp. Ambiente):** La muestra se volvió a amasar y reformar a su $L_0$ de 1.5 cm. Se aplicó un estiramiento súbito ("de golpe") y se midió la $L_f$ en la fractura.
5. **Prueba 3 (Deformación Lenta, Baja Temp.):** La muestra se reformó a 1.5 cm y se introdujo en un refrigerador por 30 minutos. Inmediatamente después, se intentó un estiramiento lento, registrando las observaciones y la $L_f$ al momento de la fractura.
6. **Prueba 4 (Deformación Rápida, Baja Temp.):** Otra muestra similar a la anterior se introdujo al mismo tiempo. Inmediatamente después, se intentó un estiramiento rápido, registrando las observaciones y la $L_f$ al momento de la fractura.


Para todas las pruebas, se calculó el porcentaje de elongación máximo hasta la fractura usando la fórmula:

$$
\text{Elongación (\%)} = \left( \frac{L_f - L_0}{L_0} \right) \times 100
$$

### Discusión

Los resultados obtenidos demuestran de manera clara que el chicle es un polímero viscoelástico, lo que significa que sus propiedades mecánicas no son constantes, sino que dependen fundamentalmente de la velocidad de deformación y de la temperatura.

#### 1. Efecto de la Velocidad de Deformación (Tasa de Deformación)

La comparación entre la Prueba 1 (lenta) y la Prueba 2 (rápida) es notoria.

- **Prueba 1 (Lenta):** Con una elongación del 7233%, el material exhibe un comportamiento dúctil extremo. Al estirar lentamente, se da tiempo a que las cadenas poliméricas que forman el chicle se desenreden, deslicen unas sobre otras y se alineen en la dirección de la fuerza. Esto permite una enorme deformación plástica antes de la fractura.
- **Prueba 2 (Rápida):** Con una elongación de solo 700%, el comportamiento es significativamente más frágil. Al aplicar una fuerza súbita, las cadenas poliméricas no tienen tiempo de reacomodarse. La energía se concentra en los enlaces intermoleculares, provocando una fractura mucho más rápida y con menor deformación.

#### 2. Efecto de la Temperatura

Las Pruebas 3 y 4 son ejemplos claros de una transición dúctil-frágil.

- Al enfriar el chicle, se reduce la energía térmica del sistema. Las cadenas poliméricas pierden su movilidad y se "congelan" en una estructura amorfa y rígida. El material se encuentra por debajo de su temperatura de transición vítrea ($T_g$).
- Las observaciones de la fractura instantánea y la poca maleabilidad confirman que el material se comporta como un sólido vítreo y frágil. En lugar de generar una deformación elástica, la energía aplicada simplemente rompe los enlaces, resultando en una fractura súbita con una elongación casi nula. La dificultad para sujetarlo se debe a esta misma rigidez, que impide que el material se amolde a los dedos.

#### Consideraciones sobre el método experimental

Es notable que, al realizar este tipo de pruebas de forma manual, se introducen márgenes de error significativos. La imposibilidad de controlar con precisión la velocidad de deformación especialmente en el estiramiento rápido y la dificultad para medir la longitud exacta en el momento de la fractura, debido al movimiento de las manos y a la adherencia del material, afectaron la exactitud de los resultados. En condiciones ideales, estas pruebas deberían realizarse con equipos como extensómetros o máquinas de ensayo universal, que permiten aplicar cargas y velocidades controladas, así como medir deformaciones con mayor precisión. Sin embargo, en el caso de polímeros tan deformables como el chicle, incluso estos equipos pueden tener limitaciones si no están adaptados para mediciones extremas. La falta de un sistema de sujeción adecuado y la naturaleza adhesiva del material también complicaron las mediciones, especialmente en las pruebas a baja temperatura, donde el material se volvió resbaladizo y quebradizo.

#### Conclusión

Se concluye que el chicle es un material viscoelástico cuyo comportamiento mecánico es una función directa de las condiciones experimentales. A temperatura ambiente y baja velocidad de deformación, es un material altamente dúctil. Sin embargo, al aumentar la velocidad de deformación o al disminuir la temperatura por debajo de su $T_g$, su comportamiento cambia drásticamente a frágil. Los resultados validan los principios de la física de materiales poliméricos. Además, se destaca la importancia de utilizar equipos de medición y control adecuados para minimizar errores y obtener resultados más confiables y reproducibles.
