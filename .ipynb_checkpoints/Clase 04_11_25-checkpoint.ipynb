$$m = \tanh (\frac{C}{T} m)$$

Con $T$ arbitraria tal que si $T = 1 \Rightarrow m = \tanh(cm)$

tal que $m = f(m)$

Y resovemos $$C = \frac{MJ}{K_B} = 1 $$

$$\Rightarrow m = \tanh(m)$$
$m = f(m)$

```python
import numpy as np
import matplotlib.pyplot as plt

def metodo_relajacion(f, df, x0, tol=1e-8, max_iter=1000):
    """
    Mtodo de relajacin para resolver x = f(x)
    
    Parmetros:
    f: funcin f(x)
    df: derivada f'(x)
    x0: valor inicial
    tol: tolerancia
    max_iter: mximo nmero de iteraciones
    
    Retorna:
    x: solucin
    iteraciones: nmero de iteraciones realizadas
    """
    x_prev = x0
    iteraciones = 0
    
    for i in range(max_iter):
        x_next = f(x_prev)
        
        # Clculo del error segn el mtodo de relajacin
        if abs(df(x_next)) < 1:
            denominator = 1 - 1/df(x_next)
            if abs(denominator) > 1e-10:  # Evitar divisin por cero
                error = abs((x_next - x_prev) / denominator)
            else:
                error = abs(x_next - x_prev)
        else:
            error = abs(x_next - x_prev)
        
        x_prev = x_next
        iteraciones += 1
        
        if error < tol:
            break
    
    return x_prev, iteraciones

def f_m(m, T):
    """Funcin f(m) = tanh(m/T)"""
    return np.tanh(m / T)

def df_m(m, T):
    """Derivada de f(m) = tanh(m/T)"""
    return (1 / T) * (1 / np.cosh(m / T)**2)

def resolver_ecuacion_magnetizacion(T_values, m0=0.1):
    """
    Resuelve M = tanh(m/T) para diferentes valores de T
    
    Parmetros:
    T_values: array de valores de temperatura
    m0: valor inicial para el mtodo iterativo
    
    Retorna:
    m_solutions: soluciones de m para cada T
    """
    m_solutions = []
    
    for T in T_values:
        if T == 0:  # Evitar divisin por cero
            m_solutions.append(1.0)
            continue
            
        # Definir funciones para el T actual
        def f_current(m):
            return f_m(m, T)
            
        def df_current(m):
            return df_m(m, T)
        
        # Para T > 1, la nica solucin es m = 0
        if T > 1:
            m_sol, _ = metodo_relajacion(f_current, df_current, m0)
            # Verificar si es realmente cero (dentro de tolerancia)
            if abs(m_sol) < 1e-6:
                m_solutions.append(0.0)
            else:
                m_solutions.append(m_sol)
        else:
            # Para T < 1, hay soluciones positivas y negativas
            # Solucin positiva
            m_pos, _ = metodo_relajacion(f_current, df_current, m0)
            m_solutions.append(abs(m_pos))  # Tomamos el valor absoluto
    
    return np.array(m_solutions)

# Parmetros de la simulacin
T_min = 0.1
T_max = 2.0
num_points = 200

# Generar valores de temperatura
T_values = np.linspace(T_min, T_max, num_points)

# Resolver la ecuacin para cada temperatura
m_values = resolver_ecuacion_magnetizacion(T_values)

# Crear la grfica
plt.figure(figsize=(10, 6))
plt.plot(T_values, m_values, 'b-', linewidth=2, label='Solucin M = tanh(m/T)')
plt.axvline(x=1.0, color='r', linestyle='--', alpha=0.7, label='T = 1 (Temperatura crtica)')
plt.xlabel('Temperatura (T)', fontsize=12)
plt.ylabel('Magnetizacin (m)', fontsize=12)
plt.title('Solucin de M = tanh(m/T) usando Mtodo de Relajacin', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(T_min, T_max)
plt.ylim(-0.1, 1.1)

# Aadir informacin sobre el mtodo
plt.text(0.6, 0.8, f'Mtodo de Relajacin\nTolerancia: 1e-8\nPuntos: {num_points}', 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgray", alpha=0.7))

plt.tight_layout()
plt.show()

# Verificacin adicional para algunos puntos
print("Verificacin para algunos valores de T:")
print("T\tm\t\ttanh(m/T)")
print("-" * 35)
for T in [0.2, 0.5, 0.8, 1.0, 1.2, 1.5]:
    idx = np.argmin(np.abs(T_values - T))
    m = m_values[idx]
    tanh_val = np.tanh(m/T)
    print(f"{T:.1f}\t{m:.6f}\t{tanh_val:.6f}")

# Mostrar informacin sobre la convergencia
print(f"\nRango de T: [{T_min}, {T_max}]")
print(f"Nmero de puntos: {num_points}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# La ecuacin que queremos resolver es: m = tanh(m/T)
def ecuacion_magnetizacion(m, T):
    return np.tanh(m / T)
```

```python
def metodo_relajacion_simple(T, m_inicial=0.5, max_iteraciones=1000):

    m_actual = m_inicial
    
    for i in range(max_iteraciones):
        # Aplicamos la funcin: m_nuevo = tanh(m_actual/T)
        m_nuevo = ecuacion_magnetizacion(m_actual, T)
        
        # Si el cambio es muy pequeo, paramos
        if abs(m_nuevo - m_actual) < 0.000001:
            break
            
        m_actual = m_nuevo
    
    return m_nuevo

# Probemos el mtodo para una temperatura
T_prueba = 0.5
m_solucion = metodo_relajacion_simple(T_prueba)
print(f"Para T = {T_prueba}, la solucin es m = {m_solucion:.4f}")
print(f"Verificacin: tanh({m_solucion:.4f}/{T_prueba}) = {np.tanh(m_solucion/T_prueba):.4f}")
```

```python
# Crear un rango de temperaturas desde muy fro hasta muy caliente
temperaturas = np.linspace(0.1, 2.0, 50)  # 50 puntos entre 0.1 y 2.0
magnetizaciones = []

for T in temperaturas:
    # Para temperaturas altas, empezamos con un valor pequeo
    if T > 1.0:
        m_inicial = 0.1
    else:
        m_inicial = 0.8  # Para temperaturas bajas, empezamos ms alto
    
    m_sol = metodo_relajacion_simple(T, m_inicial)
    magnetizaciones.append(m_sol)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(temperaturas, magnetizaciones, 'bo-', linewidth=2, markersize=4)
plt.xlabel('Temperatura (T)')
plt.ylabel('Magnetizacin (m)')
plt.title('Soluciones de m = tanh(m/T)')
plt.grid(True, alpha=0.3)

# Lnea vertical en T=1 (temperatura crtica)
#plt.axvline(x=1.0, color='red', linestyle='--', label='T = 1 (crtica)')
#plt.legend()

plt.show()
```

```python
# Podemos probar diferentes temperaturas manualmente
def probar_temperatura_manual():
    T = float(input("Ingresa una temperatura T: "))
    m = metodo_relajacion_simple(T)
    
    print(f"\nPara T = {T}:")
    print(f" Solucin: m = {m:.6f}")
    print(f" Verificacin: tanh({m:.6f}/{T}) = {np.tanh(m/T):.6f}")
    print(f" Diferencia: {abs(m - np.tanh(m/T)):.8f}")

probar_temperatura_manual()
```

# Mtodo de Biseccin (Bsqueda Binaria)

El mtodo de biseccin es un algoritmo simple pero poderoso para encontrar races de funciones. Funciona dividiendo repetidamente un intervalo a la mitad y seleccionando el subintervalo que contiene la raz.

1. **Necesitamos una funcin f(x) que cambie de signo** en un intervalo [a,b]
2. **f(a) y f(b) deben tener signos opuestos** (uno positivo, otro negativo)
3. **Calculamos el punto medio** c = (a + b)/2
4. **Miramos el signo de f(c)**:
   - Si f(c) tiene mismo signo que f(a)  nuevo intervalo [c, b]
   - Si f(c) tiene mismo signo que f(b)  nuevo intervalo [a, c]
5. **Repetimos** hasta que el intervalo sea muy pequeo

```python
def metodo_biseccion(f, a, b, max_iter=100, tolerancia=0.0001):

    # Verificar cambio de signo
    if f(a) * f(b) > 0:
        print(f"ERROR: f({a}) y f({b}) tienen mismo signo")
        return None
    
    print(f"Bsqueda en intervalo [{a}, {b}]")
    print(f"f({a}) = {f(a):.4f}, f({b}) = {f(b):.4f}")
    print()
    
    iteracion = 0
    a_actual, b_actual = a, b
    
    while iteracion < max_iter:
        # Calcular punto medio
        c = (a_actual + b_actual) / 2
        f_c = f(c)
        
        print(f"Iteracin {iteracion + 1}:")
        print(f"   Intervalo: [{a_actual:.6f}, {b_actual:.6f}]")
        print(f"   Punto medio: {c:.6f}")
        print(f"   f(c) = {f_c:.6f}")
        
        # Verificar si encontramos la raz
        if abs(f_c) < tolerancia:
            print(f"   ENCONTRADO! f({c:.6f})  0")
            return c
        
        # Decidir nuevo intervalo
        if f(a_actual) * f_c < 0:
            b_actual = c
            print(f"   Raz entre [{a_actual:.6f}, {c:.6f}]")
        else:
            a_actual = c
            print(f"   Raz entre [{c:.6f}, {b_actual:.6f}]")
        
        error = b_actual - a_actual
        print(f"   Error: {error:.6f}")
        print()
        
        if error < tolerancia:
            print("Precisin alcanzada")
            return c
            
        iteracion += 1
    
    print("Mximo de iteraciones alcanzado")
    return c

# Ejecutar bsqueda
raiz = metodo_biseccion(f, 1, 3)
print(f"Raz encontrada: x = {raiz}")
print(f"f({raiz}) = {f(raiz)}")
```
