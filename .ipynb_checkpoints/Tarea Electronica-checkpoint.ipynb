### **1. ¿Qué es un Amplificador Operacional (Op-Amp)?**

**Definición:**
Un amplificador operacional es un **circuito integrado analógico** de alta ganancia que amplifica la diferencia de voltaje entre sus dos entradas.

**Características Ideales:**
- **Ganancia infinita**
- **Impedancia de entrada infinita** (no consume corriente de entrada)
- **Impedancia de salida cero**
- **Ancho de banda infinito**
- **Respuesta lineal**

**Símbolo y Terminales:**
```
          +Vcc
           |
   (-) ---| |--- Salida
          | |
   (+) ---| |--- 
           |
          -Vcc
```

**Funcionamiento Básico:**
- **Entrada No Inversora (+):** La señal sale en **misma fase**
- **Entrada Inversora (-):** La señal sale en **fase opuesta** (180°)
- La salida es: `Vout = Aol × (V+ - V-)` donde Aol es la ganancia en lazo abierto (muy alta, típicamente > 100,000)

**Configuraciones Típicas:**
- **Amplificador Inversor**
- **Amplificador No Inversor**  
- **Seguidor de Voltaje**
- **Sumador, Restador, Integrador, Derivador**


### **4. Diferencia entre Amplificador Operacional y Comparador**

Aunque se parecen físicamente, tienen **propósitos y comportamientos diferentes**:

#### **AMPLIFICADOR OPERACIONAL**

**Función Principal:** **Amplificar** señales analógicas manteniendo la relación lineal entrada-salida.

**Modo de Operación:**
- Casi siempre se usa con **realimentación negativa**
- Opera en la **región lineal**
- La salida es proporcional a la diferencia de entradas

**Características:**
- Diseñado para operar en **rango lineal**
- Tiempo de respuesta limitado (slew rate)
- Puede tener compensación interna para estabilidad
- Salida analógica proporcional a la entrada

**Ejemplo de Aplicación:**
```Amplificar una señal de audio de 0.1V a 1V con ganancia = 10```

#### **COMPARADOR**

**Función Principal:** **Comparar** dos voltajes y producir una salida digital indicando cuál es mayor.

**Modo de Operación:**
- Opera **sin realimentación** o con realimentación positiva
- Trabaja en **saturación** (todo o nada)
- La salita es digital (alto o bajo)

**Características:**
- Optimizado para **conmutación rápida**
- Sin compensación interna (más rápido)
- Puede tener salida compatible con lógica digital
- Diseñado para operar en saturación

**Ejemplo de Aplicación:**
```Detectar cuando una temperatura supera los 25°C y encender un ventilador```

| Característica | **Amplificador Operacional** | **Comparador** |
|----------------|------------------------------|----------------|
| **Función** | Amplificar señales | Comparar voltajes |
| **Tipo de Salida** | Analógica (proporcional) | Digital (alto/bajo) |
| **Modo Operación** | Lineal (con realimentación) | Saturación (sin realimentación) |
| **Velocidad** | Moderada (limitada por slew rate) | Muy rápida |
| **Realimentación** | Negativa (generalmente) | Positiva o ninguna |
| **Salida en Reposo** | Mitad del voltaje de alimentación | Alto o bajo definido |
| **Aplicación Típica** | Audio, instrumentación, filtros | Detección de umbral, conversión A/D |
