```python
import numpy as np
import matplotlib.pyplot as plt

k = 8.9875e9
L = 0.1
q0 = 100

x = np.linspace(-L/2, L/2, q0)
y = np.linspace(-L/2, L/2, q0)
X, Y = np.meshgrid(x, y)

V = np.zeros((30, 30))  # Potencial elctrico
Ex = np.zeros((30, 30)) # Componente x del campo
Ey = np.zeros((30, 30)) # Componente y del campo

# Calcular el potencial punto por punto
for i in range(30):
    for j in range(30):
        # Punto donde calculamos el potencial
        x0 = X[i,j]
        y0 = Y[i,j]
        
        potencial_total = 0
        
        # Sumar contribucin de pequeos pedazos de carga
        for m in range(20): 
            for n in range(20):
                # Posicin
                x_carga = -L/2 + (m * L/20)
                y_carga = -L/2 + (n * L/20)
                
                # Densidad de carga en ese punto
                densidad = q0 * np.sin(2*np.pi*x_carga/L) * np.sin(2*np.pi*y_carga/L)
                
                area_pedacito = (L/20) * (L/20)
                carga_pedacito = densidad * area_pedacito
                
                # Distancia al punto donde calculamos
                distancia = np.sqrt((x0 - x_carga)**2 + (y0 - y_carga)**2)
                
                if distancia > 0.001:  # Evitar divisin por cero
                    potencial_total += k * carga_pedacito / distancia
        
        V[i,j] = potencial_total

# Calcular campo elctrico = -gradiente del potencial
for i in range(1, 29):  # Evitar bordes
    for j in range(1, 29):
        # Derivada en x: (V[i+1,j] - V[i-1,j]) / (2*dx)
        dx = x[1] - x[0]  # Espaciado entre puntos
        Ex[i,j] = -(V[i+1,j] - V[i-1,j]) / (2*dx)
        
        # Derivada en y: (V[i,j+1] - V[i,j-1]) / (2*dy)
        dy = y[1] - y[0]  # Espaciado entre puntos
        Ey[i,j] = -(V[i,j+1] - V[i,j-1]) / (2*dy)
        
E_magnitud = np.sqrt(Ex**2 + Ey**2)

plt.figure(figsize=(8, 5))
x_carga = np.linspace(-0.5, 0.5, q0) 
y_carga = np.linspace(-0.5, 0.5, q0)
Xc, Yc = np.meshgrid(x_carga, y_carga)
densidad = q0 * np.sin(2*np.pi*Xc/L) * np.sin(2*np.pi*Yc/L)
plt.contourf(Xc, Yc, densidad, levels=20, cmap="seismic")
plt.colorbar(label='Densidad (C/m)')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Densidad de Carga Original')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
print(f"Potencial mximo: {np.max(V):.2e} V")
print(f"Campo mximo: {np.max(E_magnitud):.2e} V/m")
```
